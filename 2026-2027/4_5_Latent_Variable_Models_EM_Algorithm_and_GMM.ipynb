### Модели со скрытыми переменными: мотивация, определение и идея EM-алгоритма

Многие задачи машинного обучения и статистики естественно формулируются в терминах ненаблюдаемых (латентных) переменных — величин, которые не содержатся в данных, но существенно упрощают описание модели. Примерами могут служить метки кластеров в задаче кластеризации, индикаторы компонент смеси распределений, пропущенные значения в таблицах или скрытые факторы в факторном анализе. Если бы эти переменные были известны, оценивание параметров модели часто сводилось бы к простым вычислительным процедурам. EM-алгоритм (Expectation‑Maximization) — это итеративный метод, позволяющий находить оценки максимального правдоподобия в присутствии латентных переменных, не требуя их фактического знания.

#### 1. Мотивация: зачем нужны латентные переменные?

Рассмотрим общую вероятностную модель, в которой наблюдаемые данные $\mathbf{X} = \{\mathbf{x}_1,\dots,\mathbf{x}_n\}$ порождаются совместно с ненаблюдаемыми переменными $\mathbf{Z} = \{\mathbf{z}_1,\dots,\mathbf{z}_n\}$. Полное правдоподобие (complete data likelihood) $p(\mathbf{X}, \mathbf{Z} \mid \boldsymbol{\theta})$ предполагается принадлежащим некоторому параметрическому семейству и легко максимизируемым, если бы $\mathbf{Z}$ были известны. Однако на практике доступно лишь **наблюдаемое правдоподобие** (incomplete data likelihood), получаемое маргинализацией по $\mathbf{Z}$:

$$
L(\boldsymbol{\theta}) = p(\mathbf{X} \mid \boldsymbol{\theta}) = \int p(\mathbf{X}, \mathbf{Z} \mid \boldsymbol{\theta}) \, d\mathbf{Z}. \tag{1}
$$

Прямая максимизация $\log L(\boldsymbol{\theta})$ часто оказывается аналитически сложной или вычислительно неэффективной из-за интеграла (или суммы) внутри логарифма:

$$
\log p(\mathbf{X} \mid \boldsymbol{\theta}) = \log \sum_{\mathbf{Z}} p(\mathbf{X}, \mathbf{Z} \mid \boldsymbol{\theta}).
$$

Ключевая идея EM заключается в том, чтобы обойти трудность, работая не с наблюдаемым логарифмическим правдоподобием, а с его **нижней оценкой**, построенной с помощью условного распределения латентных переменных. Это позволяет на каждой итерации гарантированно улучшать истинное правдоподобие.

#### 2. Формулировка EM-алгоритма

Пусть $\boldsymbol{\theta}^{(t)}$ — текущая оценка параметров. EM-алгоритм состоит из двух шагов:

- **E-шаг (Expectation).** Вычисляется условное распределение латентных переменных при текущих параметрах: $p(\mathbf{Z} \mid \mathbf{X}, \boldsymbol{\theta}^{(t)})$. Затем строится функция $Q(\boldsymbol{\theta}, \boldsymbol{\theta}^{(t)})$ — математическое ожидание полного логарифмического правдоподобия по этому распределению:

$$
Q(\boldsymbol{\theta}, \boldsymbol{\theta}^{(t)}) = \mathbb{E}_{\mathbf{Z} \mid \mathbf{X}, \boldsymbol{\theta}^{(t)}} \bigl[ \log p(\mathbf{X}, \mathbf{Z} \mid \boldsymbol{\theta}) \bigr]. \tag{2}
$$

- **M-шаг (Maximization).** Ищется новое значение параметров $\boldsymbol{\theta}^{(t+1)}$, максимизирующее $Q$:

$$
\boldsymbol{\theta}^{(t+1)} = \arg\max_{\boldsymbol{\theta}} Q(\boldsymbol{\theta}, \boldsymbol{\theta}^{(t)}). \tag{3}
$$

Итерации повторяются до сходимости: изменения параметров или наблюдаемого правдоподобия становятся меньше заданного порога.

Таким образом, сложная задача максимизации $\log p(\mathbf{X} \mid \boldsymbol{\theta})$ заменяется последовательностью более простых задач максимизации ожидаемого полного логарифмического правдоподобия.

#### 3. Монотонное увеличение наблюдаемого правдоподобия

Фундаментальное свойство EM-алгоритма — на каждой итерации наблюдаемое правдоподобие не убывает:

$$
p(\mathbf{X} \mid \boldsymbol{\theta}^{(t+1)}) \ge p(\mathbf{X} \mid \boldsymbol{\theta}^{(t)}).
$$

Докажем это, используя неравенство Йенсена. Для любого распределения $q(\mathbf{Z})$ на латентных переменных справедливо следующее разложение:

$$
\begin{aligned}
\log p(\mathbf{X} \mid \boldsymbol{\theta}) &= \log \sum_{\mathbf{Z}} p(\mathbf{X}, \mathbf{Z} \mid \boldsymbol{\theta}) \\
&= \log \sum_{\mathbf{Z}} q(\mathbf{Z}) \frac{p(\mathbf{X}, \mathbf{Z} \mid \boldsymbol{\theta})}{q(\mathbf{Z})} \\
&\ge \sum_{\mathbf{Z}} q(\mathbf{Z}) \log \frac{p(\mathbf{X}, \mathbf{Z} \mid \boldsymbol{\theta})}{q(\mathbf{Z})} \qquad \text{(неравенство Йенсена)} \\
&= \mathcal{L}(q, \boldsymbol{\theta}).
\end{aligned}
$$

Функция $\mathcal{L}(q, \boldsymbol{\theta})$ является нижней оценкой логарифмического правдоподобия. EM-алгоритм выбирает $q(\mathbf{Z}) = p(\mathbf{Z} \mid \mathbf{X}, \boldsymbol{\theta}^{(t)})$ на E-шаге, делая оценку точной в точке $\boldsymbol{\theta}^{(t)}$:

$$
\mathcal{L}\bigl(p(\mathbf{Z} \mid \mathbf{X}, \boldsymbol{\theta}^{(t)}), \boldsymbol{\theta}^{(t)}\bigr) = \log p(\mathbf{X} \mid \boldsymbol{\theta}^{(t)}).
$$

Затем на M-шаге параметры $\boldsymbol{\theta}^{(t+1)}$ максимизируют $\mathcal{L}$, откуда

$$
\log p(\mathbf{X} \mid \boldsymbol{\theta}^{(t+1)}) \ge \mathcal{L}\bigl(p(\mathbf{Z} \mid \mathbf{X}, \boldsymbol{\theta}^{(t)}), \boldsymbol{\theta}^{(t+1)}\bigr) \ge \mathcal{L}\bigl(p(\mathbf{Z} \mid \mathbf{X}, \boldsymbol{\theta}^{(t)}), \boldsymbol{\theta}^{(t)}\bigr) = \log p(\mathbf{X} \mid \boldsymbol{\theta}^{(t)}).
$$

Таким образом, наблюдаемое правдоподобие монотонно возрастает. Это свойство не гарантирует сходимости к глобальному максимуму, но обеспечивает, что каждая итерация не ухудшает решение.

#### 4. Сходимость EM-алгоритма

EM-алгоритм сходится к стационарной точке наблюдаемого правдоподобия, т.е. к точке, где градиент $\nabla_{\boldsymbol{\theta}} \log p(\mathbf{X} \mid \boldsymbol{\theta}) = \mathbf{0}$. В общем случае это может быть локальный максимум, локальный минимум (редко) или седловая точка. Типичная скорость сходимости линейна и определяется наибольшим собственным значением матрицы Якоби отображения EM-шага. Вблизи оптимума EM может сходиться медленно, особенно в случае плохо разделённых компонент смеси.

Для практики важно:

- EM следует запускать из нескольких начальных точек, чтобы снизить риск застревания в плохом локальном максимуме.
- Существуют ускоренные версии (например, Aitken’s acceleration) и гибридные методы, переключающиеся на градиентный спуск вблизи сходимости.

> **Углублённый взгляд: доказательство монотонности через разложение на Q и H.**  
> Представим логарифмическое правдоподобие в виде
> $$
> \log p(\mathbf{X} \mid \boldsymbol{\theta}) = Q(\boldsymbol{\theta}, \boldsymbol{\theta}^{(t)}) + H(\boldsymbol{\theta}, \boldsymbol{\theta}^{(t)}),
> $$
> где
> $$
> \begin{aligned}
> Q(\boldsymbol{\theta}, \boldsymbol{\theta}^{(t)}) &= \mathbb{E}_{\mathbf{Z} \mid \mathbf{X}, \boldsymbol{\theta}^{(t)}} \bigl[ \log p(\mathbf{X}, \mathbf{Z} \mid \boldsymbol{\theta}) \bigr], \\
> H(\boldsymbol{\theta}, \boldsymbol{\theta}^{(t)}) &= -\mathbb{E}_{\mathbf{Z} \mid \mathbf{X}, \boldsymbol{\theta}^{(t)}} \bigl[ \log p(\mathbf{Z} \mid \mathbf{X}, \boldsymbol{\theta}) \bigr].
> \end{aligned}
> $$
> Функция $H$ является условной энтропией распределения $p(\mathbf{Z} \mid \mathbf{X}, \boldsymbol{\theta}^{(t)})$ относительно $p(\mathbf{Z} \mid \mathbf{X}, \boldsymbol{\theta})$. По неравенству Гиббса, для любого $\boldsymbol{\theta}$ имеем $H(\boldsymbol{\theta}^{(t)}, \boldsymbol{\theta}^{(t)}) \le H(\boldsymbol{\theta}, \boldsymbol{\theta}^{(t)})$, причём равенство достигается при $\boldsymbol{\theta} = \boldsymbol{\theta}^{(t)}$.  
> M-шаг увеличивает $Q$, а изменение $H$ не может уменьшить правдоподобие, так как $\log p(\mathbf{X} \mid \boldsymbol{\theta}^{(t+1)}) - \log p(\mathbf{X} \mid \boldsymbol{\theta}^{(t)}) = [Q(\boldsymbol{\theta}^{(t+1)}, \boldsymbol{\theta}^{(t)}) - Q(\boldsymbol{\theta}^{(t)}, \boldsymbol{\theta}^{(t)})] + [H(\boldsymbol{\theta}^{(t+1)}, \boldsymbol{\theta}^{(t)}) - H(\boldsymbol{\theta}^{(t)}, \boldsymbol{\theta}^{(t)})] \ge 0$. Это даёт альтернативное доказательство монотонности.

#### 5. Практические замечания

EM-алгоритм широко применяется в задачах кластеризации (гауссовские смеси), оценивания пропущенных значений, байесовских сетях и факторном анализе. Его основное преимущество — концептуальная простота и гарантированное улучшение правдоподобия на каждом шаге. Недостатки — возможность застревания в локальных оптимумах и медленная сходимость. Эти проблемы мотивируют развитие расширений, таких как стохастический EM, вариационный EM и обобщённый EM, которые будут рассмотрены в последующих разделах.



#### Визуализация EM-алгоритма: подгонка смеси двух гауссиан

Чтобы проиллюстрировать работу EM, сгенерируем данные из смеси двух двумерных нормальных распределений и запустим EM на несколько итераций. На графиках показаны точки, текущие центры (кресты) и эллипсы ковариаций компонент. Видно, как ответственности перераспределяются, а параметры компонент постепенно «захватывают» свои кластеры.

```python
# @title Демонстрация EM для смеси двух гауссиан
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from matplotlib.patches import Ellipse

# Генерация данных из смеси двух нормальных распределений
np.random.seed(42)
n_samples = 300
true_means = [np.array([-2, 0]), np.array([2, 1])]
true_covs = [np.array([[1, 0.5], [0.5, 1]]), np.array([[1, -0.3], [-0.3, 1]])]
weights = [0.4, 0.6]

# Сэмплируем метки компонент и сами точки
components = np.random.choice(2, size=n_samples, p=weights)
X = np.array([np.random.multivariate_normal(true_means[c], true_covs[c]) for c in components])

# Начальные параметры для EM (одна из возможных инициализаций)
np.random.seed(1)
k = 2
mu = X[np.random.choice(n_samples, k, replace=False)]
cov = [np.cov(X.T) for _ in range(k)]
pi = np.ones(k) / k

max_iter = 10
log_likelihoods = []
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.flatten()
selected_iters = [0, 1, 2, 4, 7, 9]

def plot_ellipse(ax, mean, cov, color, alpha=0.3):
    """Вспомогательная функция для отрисовки эллипса по ковариационной матрице."""
    v, w = np.linalg.eigh(cov)
    v = 2.0 * np.sqrt(2.0) * np.sqrt(v)
    u = w[0] / np.linalg.norm(w[0])
    angle = np.arctan2(u[1], u[0])
    ell = Ellipse(xy=mean, width=v[0], height=v[1], angle=np.degrees(angle),
                  edgecolor=color, fc=color, alpha=alpha, lw=2)
    ax.add_patch(ell)

for it in range(max_iter):
    # E-шаг: вычисление ответственностей
    resp = np.zeros((n_samples, k))
    for j in range(k):
        resp[:, j] = pi[j] * multivariate_normal.pdf(X, mean=mu[j], cov=cov[j])
    resp /= resp.sum(axis=1, keepdims=True)

    # M-шаг: пересчёт параметров
    nk = resp.sum(axis=0)
    pi = nk / n_samples
    for j in range(k):
        mu[j] = (resp[:, j] @ X) / nk[j]
        diff = X - mu[j]
        cov[j] = (diff.T * resp[:, j]) @ diff / nk[j]

    # Логарифмическое правдоподобие
    ll = np.sum(np.log(np.sum([pi[j] * multivariate_normal.pdf(X, mean=mu[j], cov=cov[j]) for j in range(k)], axis=1)))
    log_likelihoods.append(ll)

    # Отрисовка выбранных итераций
    if it in selected_iters:
        ax = axes[selected_iters.index(it)]
        ax.scatter(X[:, 0], X[:, 1], c=resp[:, 1], cmap='viridis', alpha=0.6)
        colors = ['blue', 'green']
        for j in range(k):
            ax.scatter(mu[j][0], mu[j][1], c=colors[j], marker='x', s=200, linewidths=3)
            plot_ellipse(ax, mu[j], cov[j], colors[j])
        ax.set_title(f'Итерация {it}')
        ax.set_xlabel('x1'); ax.set_ylabel('x2')

# График логарифмического правдоподобия
axes[-1].plot(range(len(log_likelihoods)), log_likelihoods, 'o-')
axes[-1].set_title('Логарифмическое правдоподобие')
axes[-1].set_xlabel('Итерация')
axes[-1].set_ylabel('Log-likelihood')
axes[-1].grid(True)

plt.tight_layout()
plt.show()
```

### EM-алгоритм для гауссовской смеси (GMM): вывод формул обновления

Общая теория EM-алгоритма, изложенная в предыдущем разделе, находит своё наиболее яркое и классическое воплощение в задаче разделения смеси гауссовских распределений (Gaussian Mixture Model, GMM). Здесь латентная переменная указывает, какой компонентой порождён каждый объект, а оцениваемые параметры включают веса компонент, их средние и ковариационные матрицы. Ниже мы подробно выведем E- и M-шаги для GMM.

#### 1. Модель гауссовской смеси

Пусть имеется выборка $\mathbf{X} = \{\mathbf{x}_1,\dots,\mathbf{x}_n\}$, где каждый $\mathbf{x}_i \in \mathbb{R}^d$ порождён независимо от распределения

$$
p(\mathbf{x}_i \mid \boldsymbol{\theta}) = \sum_{k=1}^K \pi_k \, \mathcal{N}(\mathbf{x}_i \mid \boldsymbol{\mu}_k, \boldsymbol{\Sigma}_k),
\tag{1}
$$

где $\boldsymbol{\theta} = \{\pi_1,\dots,\pi_K, \boldsymbol{\mu}_1,\dots,\boldsymbol{\mu}_K, \boldsymbol{\Sigma}_1,\dots,\boldsymbol{\Sigma}_K\}$ — совокупность параметров. Веса смеси удовлетворяют $\pi_k \ge 0$ и $\sum_{k=1}^K \pi_k = 1$. Каждое гауссовское ядро $\mathcal{N}(\mathbf{x} \mid \boldsymbol{\mu}, \boldsymbol{\Sigma})$ задаётся плотностью

$$
\mathcal{N}(\mathbf{x} \mid \boldsymbol{\mu}, \boldsymbol{\Sigma}) = \frac{1}{(2\pi)^{d/2} |\boldsymbol{\Sigma}|^{1/2}} \exp\!\Bigl( -\frac{1}{2}(\mathbf{x} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{x} - \boldsymbol{\mu}) \Bigr).
$$

Введём для каждого наблюдения $i$ латентную переменную $z_i \in \{1,\dots,K\}$, указывающую, из какой компоненты оно было взято. Если бы $z_i$ были известны, полное правдоподобие (complete data likelihood) имело бы вид

$$
p(\mathbf{x}_i, z_i \mid \boldsymbol{\theta}) = \pi_{z_i} \, \mathcal{N}(\mathbf{x}_i \mid \boldsymbol{\mu}_{z_i}, \boldsymbol{\Sigma}_{z_i}).
$$

Таким образом, совместное распределение всех наблюдаемых и латентных переменных факторизуется:

$$
p(\mathbf{X}, \mathbf{Z} \mid \boldsymbol{\theta}) = \prod_{i=1}^n \prod_{k=1}^K \Bigl[ \pi_k \, \mathcal{N}(\mathbf{x}_i \mid \boldsymbol{\mu}_k, \boldsymbol{\Sigma}_k) \Bigr]^{\mathbf{1}_{\{z_i = k\}}}.
$$

#### 2. E-шаг: вычисление ответственностей (responsibilities)

На E-шаге мы вычисляем условное распределение латентных переменных при текущем приближении параметров $\boldsymbol{\theta}^{(t)} = \{\pi_k^{(t)}, \boldsymbol{\mu}_k^{(t)}, \boldsymbol{\Sigma}_k^{(t)}\}$. Так как $z_i$ дискретны, достаточно найти апостериорные вероятности принадлежности каждого объекта к каждой компоненте:

$$
\gamma_{ik} = p(z_i = k \mid \mathbf{x}_i, \boldsymbol{\theta}^{(t)}) = \frac{p(z_i = k \mid \boldsymbol{\theta}^{(t)}) \, p(\mathbf{x}_i \mid z_i = k, \boldsymbol{\theta}^{(t)})}{p(\mathbf{x}_i \mid \boldsymbol{\theta}^{(t)})}.
$$

Подставляя выражения для априорной вероятности компоненты и условной плотности, получаем

$$
\boxed{ \gamma_{ik} = \frac{\pi_k^{(t)} \, \mathcal{N}\bigl(\mathbf{x}_i \mid \boldsymbol{\mu}_k^{(t)}, \boldsymbol{\Sigma}_k^{(t)}\bigr)}{\sum_{j=1}^K \pi_j^{(t)} \, \mathcal{N}\bigl(\mathbf{x}_i \mid \boldsymbol{\mu}_j^{(t)}, \boldsymbol{\Sigma}_j^{(t)}\bigr)} }, \qquad i=1,\dots,n,\; k=1,\dots,K. \tag{2}
$$

Величины $\gamma_{ik}$ называют **ответственностями** (responsibilities) — они показывают, насколько каждая компонента «отвечает» за данное наблюдение.

Теперь запишем ожидаемое полное логарифмическое правдоподобие (функцию $Q$). По определению

$$
Q(\boldsymbol{\theta}, \boldsymbol{\theta}^{(t)}) = \mathbb{E}_{\mathbf{Z} \mid \mathbf{X}, \boldsymbol{\theta}^{(t)}} \bigl[ \log p(\mathbf{X}, \mathbf{Z} \mid \boldsymbol{\theta}) \bigr].
$$

Поскольку $z_i$ независимы при условии $\mathbf{x}_i$, а $\log$ полного правдоподобия линеен по индикаторам $\mathbf{1}_{\{z_i = k\}}$, математическое ожидание сводится к замене индикаторов на $\gamma_{ik}$:

$$
Q(\boldsymbol{\theta}, \boldsymbol{\theta}^{(t)}) = \sum_{i=1}^n \sum_{k=1}^K \gamma_{ik} \Bigl[ \log \pi_k + \log \mathcal{N}(\mathbf{x}_i \mid \boldsymbol{\mu}_k, \boldsymbol{\Sigma}_k) \Bigr]. \tag{3}
$$

Эта функция зависит от $\boldsymbol{\theta}$ через $\pi_k, \boldsymbol{\mu}_k, \boldsymbol{\Sigma}_k$, а $\gamma_{ik}$ фиксированы (вычислены по $\boldsymbol{\theta}^{(t)}$).

#### 3. M-шаг: обновление параметров

На M-шаге мы максимизируем $Q$ по $\boldsymbol{\theta}$ при ограничении $\sum_{k=1}^K \pi_k = 1$. Задача распадается на независимую оптимизацию весов смеси и параметров каждой компоненты.

**Обновление весов $\pi_k$.** Выделим в $Q$ члены, зависящие от $\pi_k$:

$$
Q_\pi = \sum_{i=1}^n \sum_{k=1}^K \gamma_{ik} \log \pi_k.
$$

Максимизируем $Q_\pi$ по $\pi_k$ при условии $\sum_k \pi_k = 1$, используя метод множителей Лагранжа. Функция Лагранжа:

$$
\mathcal{L} = \sum_{i=1}^n \sum_{k=1}^K \gamma_{ik} \log \pi_k + \lambda \Bigl( \sum_{k=1}^K \pi_k - 1 \Bigr).
$$

Приравнивая производную по $\pi_k$ к нулю:

$$
\frac{\partial \mathcal{L}}{\partial \pi_k} = \frac{\sum_{i=1}^n \gamma_{ik}}{\pi_k} + \lambda = 0 \quad \Longrightarrow \quad \pi_k = -\frac{\sum_{i=1}^n \gamma_{ik}}{\lambda}.
$$

Суммируя по $k$ и используя ограничение $\sum_k \pi_k = 1$, получаем $\lambda = - \sum_{i=1}^n \sum_{k=1}^K \gamma_{ik} = -n$, поскольку $\sum_{k=1}^K \gamma_{ik} = 1$ для каждого $i$. Следовательно,

$$
\boxed{ \pi_k^{(t+1)} = \frac{1}{n} \sum_{i=1}^n \gamma_{ik} }. \tag{4}
$$

Новый вес компоненты равен средней ответственности, которую она имеет за все наблюдения.

**Обновление средних $\boldsymbol{\mu}_k$.** Член $Q$, зависящий от $\boldsymbol{\mu}_k$, имеет вид

$$
Q_{\boldsymbol{\mu}_k} = \sum_{i=1}^n \gamma_{ik} \left[ -\frac{1}{2} (\mathbf{x}_i - \boldsymbol{\mu}_k)^T \boldsymbol{\Sigma}_k^{-1} (\mathbf{x}_i - \boldsymbol{\mu}_k) \right] + \text{const}.
$$

Градиент по $\boldsymbol{\mu}_k$:

$$
\frac{\partial Q}{\partial \boldsymbol{\mu}_k} = \sum_{i=1}^n \gamma_{ik} \boldsymbol{\Sigma}_k^{-1} (\mathbf{x}_i - \boldsymbol{\mu}_k) = \mathbf{0}.
$$

Так как $\boldsymbol{\Sigma}_k^{-1}$ обратима, это эквивалентно

$$
\sum_{i=1}^n \gamma_{ik} (\mathbf{x}_i - \boldsymbol{\mu}_k) = \mathbf{0} \quad \Longrightarrow \quad \boxed{ \boldsymbol{\mu}_k^{(t+1)} = \frac{\sum_{i=1}^n \gamma_{ik} \mathbf{x}_i}{\sum_{i=1}^n \gamma_{ik}} }. \tag{5}
$$

Таким образом, новое среднее компоненты $k$ — это взвешенное среднее всех наблюдений, где веса равны $\gamma_{ik}$.

**Обновление ковариационных матриц $\boldsymbol{\Sigma}_k$.** Зависимость $Q$ от $\boldsymbol{\Sigma}_k$ сосредоточена в слагаемых

$$
Q_{\boldsymbol{\Sigma}_k} = \sum_{i=1}^n \gamma_{ik} \left[ -\frac{1}{2} \log |\boldsymbol{\Sigma}_k| - \frac{1}{2} (\mathbf{x}_i - \boldsymbol{\mu}_k)^T \boldsymbol{\Sigma}_k^{-1} (\mathbf{x}_i - \boldsymbol{\mu}_k) \right].
$$

Для максимизации по $\boldsymbol{\Sigma}_k$ продифференцируем по элементам матрицы (или используем матричное дифференцирование). Напомним, что $\frac{\partial}{\partial \boldsymbol{\Sigma}} \log |\boldsymbol{\Sigma}| = \boldsymbol{\Sigma}^{-T}$, а $\frac{\partial}{\partial \boldsymbol{\Sigma}} \mathbf{a}^T \boldsymbol{\Sigma}^{-1} \mathbf{a} = -\boldsymbol{\Sigma}^{-T} \mathbf{a} \mathbf{a}^T \boldsymbol{\Sigma}^{-T}$. Приравнивая производную к нулю, получаем

$$
-\frac{1}{2} \sum_{i=1}^n \gamma_{ik} \left( \boldsymbol{\Sigma}_k^{-1} - \boldsymbol{\Sigma}_k^{-1} (\mathbf{x}_i - \boldsymbol{\mu}_k)(\mathbf{x}_i - \boldsymbol{\mu}_k)^T \boldsymbol{\Sigma}_k^{-1} \right) = \mathbf{0}.
$$

Умножая слева и справа на $\boldsymbol{\Sigma}_k$, приходим к условию

$$
\sum_{i=1}^n \gamma_{ik} \Bigl( \boldsymbol{\Sigma}_k - (\mathbf{x}_i - \boldsymbol{\mu}_k)(\mathbf{x}_i - \boldsymbol{\mu}_k)^T \Bigr) = \mathbf{0},
$$

откуда

$$
\boxed{ \boldsymbol{\Sigma}_k^{(t+1)} = \frac{\sum_{i=1}^n \gamma_{ik} (\mathbf{x}_i - \boldsymbol{\mu}_k^{(t+1)})(\mathbf{x}_i - \boldsymbol{\mu}_k^{(t+1)})^T}{\sum_{i=1}^n \gamma_{ik}} }. \tag{6}
$$

Новая ковариационная матрица компоненты $k$ — это взвешенная выборочная ковариация относительно обновлённого среднего.

Итерации E- и M-шагов повторяются до тех пор, пока наблюдаемое логарифмическое правдоподобие

$$
\log p(\mathbf{X} \mid \boldsymbol{\theta}) = \sum_{i=1}^n \log \sum_{k=1}^K \pi_k \, \mathcal{N}(\mathbf{x}_i \mid \boldsymbol{\mu}_k, \boldsymbol{\Sigma}_k)
$$

не перестанет существенно возрастать.

#### 4. Инициализация и регуляризация

EM для GMM чувствителен к начальному приближению параметров. Неудачная инициализация может привести к сходимости в плохом локальном оптимуме. Распространённые стратегии:

- **Случайная инициализация** — выбираются случайные центры, ковариации полагаются единичными, веса равномерными.
- **Результат алгоритма k-средних** — центры кластеров k-means используются как начальные $\boldsymbol{\mu}_k$, ковариации оцениваются по получившимся кластерам.
- **Инициализация на основе иерархической кластеризации** — более устойчива, но вычислительно дороже.

Часто EM запускается несколько раз с разными начальными условиями, и выбирается решение с наибольшим правдоподобием.

Ещё одна практическая проблема — **вырождение компонент**. Если на каком-то шаге компонента $k$ объясняет лишь одно наблюдение (или несколько, лежащих очень близко), оценка ковариации $\boldsymbol{\Sigma}_k$ может стать почти вырожденной, а правдоподобие устремится к бесконечности. Чтобы этого избежать, применяют **регуляризацию**: к оценке (6) добавляют малую положительно определённую матрицу, например $\epsilon \mathbf{I}$:

$$
\boldsymbol{\Sigma}_k^{(t+1)} \leftarrow \boldsymbol{\Sigma}_k^{(t+1)} + \epsilon \mathbf{I}, \quad \epsilon > 0.
$$

Это гарантирует, что все ковариации остаются полного ранга, и предотвращает численную неустойчивость.


```python
# @title Демонстрация EM для GMM
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from matplotlib.patches import Ellipse

# Генерация данных
np.random.seed(42)
n = 300
true_means = [np.array([-2, 0]), np.array([2, 1])]
true_covs = [np.array([[1, 0.5], [0.5, 1]]), np.array([[1, -0.3], [-0.3, 1]])]
weights = [0.4, 0.6]
components = np.random.choice(2, size=n, p=weights)
X = np.array([np.random.multivariate_normal(true_means[c], true_covs[c]) for c in components])

# Инициализация
k = 2
mu = X[np.random.choice(n, k, replace=False)]
cov = [np.cov(X.T) for _ in range(k)]
pi = np.ones(k) / k

max_iter = 15
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.flatten()
selected = [0, 1, 2, 5, 10, 14]

def plot_ellipse(ax, mean, cov, color, alpha=0.3):
    v, w = np.linalg.eigh(cov)
    v = 2 * np.sqrt(2) * np.sqrt(v)
    u = w[0] / np.linalg.norm(w[0])
    angle = np.arctan2(u[1], u[0])
    ell = Ellipse(xy=mean, width=v[0], height=v[1], angle=np.degrees(angle),
                  edgecolor=color, fc=color, alpha=alpha, lw=2)
    ax.add_patch(ell)

for it in range(max_iter):
    # E-шаг
    resp = np.zeros((n, k))
    for j in range(k):
        resp[:, j] = pi[j] * multivariate_normal.pdf(X, mean=mu[j], cov=cov[j])
    resp /= resp.sum(axis=1, keepdims=True)

    # M-шаг
    nk = resp.sum(axis=0)
    pi = nk / n
    for j in range(k):
        mu[j] = (resp[:, j] @ X) / nk[j]
        diff = X - mu[j]
        cov[j] = (diff.T * resp[:, j]) @ diff / nk[j] + 1e-6 * np.eye(2)

    # Отрисовка выбранных итераций
    if it in selected:
        ax = axes[selected.index(it)]
        ax.scatter(X[:, 0], X[:, 1], c=resp[:, 1], cmap='viridis', alpha=0.6)
        colors = ['blue', 'green']
        for j in range(k):
            ax.scatter(mu[j][0], mu[j][1], c=colors[j], marker='x', s=200, linewidths=3)
            plot_ellipse(ax, mu[j], cov[j], colors[j])
        ax.set_title(f'Итерация {it}')
        ax.set_xlabel('x1'); ax.set_ylabel('x2')

plt.tight_layout()
plt.show()
```


#### 5. Углублённый взгляд: полный вывод формул M-шага

Проведём формальный вывод формул (4)–(6), используя матричное дифференцирование и метод Лагранжа.

**Веса $\pi_k$.**  
Функция Лагранжа $\mathcal{L}(\boldsymbol{\pi}, \lambda) = \sum_{k=1}^K \left( \sum_{i=1}^n \gamma_{ik} \right) \log \pi_k + \lambda \bigl( \sum_{k=1}^K \pi_k - 1 \bigr)$. Условия первого порядка:

$$
\frac{\partial \mathcal{L}}{\partial \pi_k} = \frac{N_k}{\pi_k} + \lambda = 0, \quad N_k = \sum_{i=1}^n \gamma_{ik}.
$$

Отсюда $\pi_k = -N_k / \lambda$. Просуммировав по $k$, находим $1 = \sum_k \pi_k = -n / \lambda$, следовательно $\lambda = -n$ и $\pi_k = N_k / n$, что совпадает с (4).

**Средние $\boldsymbol{\mu}_k$.**  
Игнорируя константы, запишем целевую функцию

$$
\tilde{Q}(\boldsymbol{\mu}_k) = -\frac{1}{2} \sum_{i=1}^n \gamma_{ik} (\mathbf{x}_i - \boldsymbol{\mu}_k)^T \boldsymbol{\Sigma}_k^{-1} (\mathbf{x}_i - \boldsymbol{\mu}_k).
$$

Градиент по $\boldsymbol{\mu}_k$:

$$
\nabla_{\boldsymbol{\mu}_k} \tilde{Q} = \sum_{i=1}^n \gamma_{ik} \boldsymbol{\Sigma}_k^{-1} (\mathbf{x}_i - \boldsymbol{\mu}_k).
$$

Приравнивание к нулю и умножение на $\boldsymbol{\Sigma}_k$ даёт $\sum_i \gamma_{ik} (\mathbf{x}_i - \boldsymbol{\mu}_k) = \mathbf{0}$, откуда сразу следует (5).

**Ковариации $\boldsymbol{\Sigma}_k$.**  
Целевой функционал для $\boldsymbol{\Sigma}_k$ с учётом всех наблюдений:

$$
\tilde{Q}(\boldsymbol{\Sigma}_k) = -\frac{1}{2} \sum_{i=1}^n \gamma_{ik} \Bigl( \log |\boldsymbol{\Sigma}_k| + (\mathbf{x}_i - \boldsymbol{\mu}_k)^T \boldsymbol{\Sigma}_k^{-1} (\mathbf{x}_i - \boldsymbol{\mu}_k) \Bigr).
$$

Обозначим $\mathbf{S}_k = \sum_{i=1}^n \gamma_{ik} (\mathbf{x}_i - \boldsymbol{\mu}_k)(\mathbf{x}_i - \boldsymbol{\mu}_k)^T$. Тогда второй член можно записать как $\operatorname{tr}\bigl( \boldsymbol{\Sigma}_k^{-1} \mathbf{S}_k \bigr)$, поскольку $(\mathbf{x}_i - \boldsymbol{\mu}_k)^T \boldsymbol{\Sigma}_k^{-1} (\mathbf{x}_i - \boldsymbol{\mu}_k) = \operatorname{tr}\bigl( \boldsymbol{\Sigma}_k^{-1} (\mathbf{x}_i - \boldsymbol{\mu}_k)(\mathbf{x}_i - \boldsymbol{\mu}_k)^T \bigr)$. Таким образом,

$$
\tilde{Q}(\boldsymbol{\Sigma}_k) = -\frac{1}{2} N_k \log |\boldsymbol{\Sigma}_k| - \frac{1}{2} \operatorname{tr}\bigl( \boldsymbol{\Sigma}_k^{-1} \mathbf{S}_k \bigr), \quad N_k = \sum_i \gamma_{ik}.
$$

Используем матричные производные: $\frac{\partial \log |\boldsymbol{\Sigma}|}{\partial \boldsymbol{\Sigma}} = \boldsymbol{\Sigma}^{-T}$, $\frac{\partial \operatorname{tr}(\boldsymbol{\Sigma}^{-1} \mathbf{S})}{\partial \boldsymbol{\Sigma}} = - \boldsymbol{\Sigma}^{-T} \mathbf{S}^T \boldsymbol{\Sigma}^{-T}$. Приравнивая производную к нулю (в симметричном случае, учитывая, что $\boldsymbol{\Sigma}$ симметрична, можно использовать симметризованную производную), получаем

$$
-\frac{1}{2} N_k \boldsymbol{\Sigma}_k^{-1} + \frac{1}{2} \boldsymbol{\Sigma}_k^{-1} \mathbf{S}_k \boldsymbol{\Sigma}_k^{-1} = \mathbf{0}.
$$

Умножаем слева и справа на $\boldsymbol{\Sigma}_k$: $N_k \boldsymbol{\Sigma}_k = \mathbf{S}_k$, что даёт $\boldsymbol{\Sigma}_k = \mathbf{S}_k / N_k$ — формула (6).

Заметим, что компонента с нулевой суммарной ответственностью ($N_k = 0$) не влияет на правдоподобие, и её параметры можно не обновлять или обнулить; обычно на практике такую компоненту удаляют или переинициализируют.

#### Связь с алгоритмом k-средних

Алгоритм k-means можно рассматривать как предельный случай EM для GMM, когда все ковариационные матрицы стремятся к $\epsilon \mathbf{I}$ с $\epsilon \to 0$, а веса смеси фиксированы и равны. В этом пределе $\gamma_{ik}$ становятся индикаторами принадлежности ближайшему центру (жёсткая кластеризация), а M-шаг сводится к пересчёту центров как средних точек в кластерах. Этот взгляд объясняет, почему EM даёт «мягкую» кластеризацию с вероятностными принадлежностями, в отличие от жёсткой кластеризации k-means.

### Инициализация, выбор числа компонент и проблемы EM для GMM

После того как выведены формулы E- и M-шагов для гауссовской смеси, на первый план выходят практические вопросы, от которых зависит успех применения алгоритма. К ним относятся: с каких параметров начинать итерации, как определить подходящее число компонент $K$, как избежать вырождения компонент и как ускорить медленную сходимость EM. Ниже мы рассмотрим эти аспекты, а затем обсудим байесовское расширение GMM, позволяющее автоматически выбирать $K$.

#### 1. Инициализация параметров

EM-алгоритм чувствителен к начальному приближению $\boldsymbol{\theta}^{(0)}$. Неудачный выбор может привести к попаданию в плохой локальный максимум правдоподобия или к крайне медленной сходимости. Рассмотрим основные стратегии инициализации.

**Случайная инициализация.** Простейший способ — выбрать $K$ наблюдений случайным образом в качестве начальных средних $\boldsymbol{\mu}_k^{(0)}$, задать равные веса $\pi_k^{(0)} = 1/K$, а ковариационные матрицы положить единичными, масштабированными на общую дисперсию данных: $\boldsymbol{\Sigma}_k^{(0)} = s^2 \mathbf{I}$, где $s^2$ — выборочная дисперсия. Этот метод быстр, но может давать нестабильные результаты.

**Инициализация на основе k-means.** Более надёжный подход — сначала выполнить жёсткую кластеризацию алгоритмом k-средних (особенно с инициализацией k-means++), а затем по полученным кластерам оценить параметры GMM:

$$
\boldsymbol{\mu}_k^{(0)} = \frac{1}{|\mathcal{C}_k|} \sum_{i \in \mathcal{C}_k} \mathbf{x}_i, \qquad
\boldsymbol{\Sigma}_k^{(0)} = \frac{1}{|\mathcal{C}_k|} \sum_{i \in \mathcal{C}_k} (\mathbf{x}_i - \boldsymbol{\mu}_k^{(0)})(\mathbf{x}_i - \boldsymbol{\mu}_k^{(0)})^T, \qquad
\pi_k^{(0)} = \frac{|\mathcal{C}_k|}{n}.
$$

Поскольку k-means сам по себе может застревать в локальных минимумах, его также запускают несколько раз и выбирают лучший результат.

**Многократные запуски.** Учитывая локальную природу EM, стандартной практикой является запуск алгоритма из $R$ различных случайных начальных приближений и выбор решения с наибольшим наблюдаемым логарифмическим правдоподобием:

$$
\boldsymbol{\theta}^* = \arg\max_{r=1,\dots,R} \log p(\mathbf{X} \mid \boldsymbol{\theta}^{(r)}).
$$

Значение $R$ обычно берут от 5 до 50 в зависимости от сложности задачи и вычислительных ресурсов.

#### 2. Выбор числа компонент $K$

Число компонент $K$ не известно заранее и должно быть определено на основе данных. Используют два основных подхода: информационные критерии и кросс-валидацию.

**Информационные критерии** основаны на компромиссе между качеством подгонки (логарифмическим правдоподобием) и сложностью модели (числом параметров). Для GMM число свободных параметров $p(K)$ зависит от типа ковариационных матриц:

- *Сферические* ($\boldsymbol{\Sigma}_k = \sigma_k^2 \mathbf{I}$): $p(K) = (K-1) + K d + K$,
- *Диагональные* ($\boldsymbol{\Sigma}_k = \operatorname{diag}(\sigma_{k1}^2,\dots,\sigma_{kd}^2)$): $p(K) = (K-1) + K d + K d$,
- *Полные* ($\boldsymbol{\Sigma}_k$ произвольные): $p(K) = (K-1) + K d + K \frac{d(d+1)}{2}$.

Здесь $(K-1)$ соответствует весам смеси (c учётом $\sum\pi_k=1$), $K d$ — средним, а последнее слагаемое — ковариациям.

**AIC (Akaike Information Criterion)** и **BIC (Bayesian Information Criterion)** определяются как

$$
\begin{aligned}
\operatorname{AIC}(K) &= -2 \log p(\mathbf{X} \mid \hat{\boldsymbol{\theta}}_K) + 2 p(K), \\
\operatorname{BIC}(K) &= -2 \log p(\mathbf{X} \mid \hat{\boldsymbol{\theta}}_K) + p(K) \log n,
\end{aligned}
$$

где $\hat{\boldsymbol{\theta}}_K$ — оценка максимального правдоподобия для $K$ компонент. Выбирается $K$, минимизирующее критерий. BIC штрафует сложность сильнее (множитель $\log n$), поэтому имеет тенденцию выбирать более простые модели и асимптотически состоятелен, в то время как AIC может переоценивать $K$.

**Кросс-валидация** — более прямой, но вычислительно затратный метод. Данные разбиваются на обучающую и валидационную части; для каждого $K$ на обучающей выборке обучается GMM, а на валидационной оценивается логарифмическое правдоподобие $\log p(\mathbf{X}_{\text{val}} \mid \hat{\boldsymbol{\theta}}_K)$. Процедура повторяется для нескольких разбиений, и выбирается $K$ с наилучшим средним значением.

#### 3. Проблема вырожденных компонент

Одна из известных проблем EM для GMM — возможность вырождения компонент. Если в процессе итераций некоторая компонента начинает описывать очень малое число наблюдений, лежащих почти в одной точке (или даже одно наблюдение), её оценённая ковариационная матрица может стать почти вырожденной, а соответствующая плотность — неограниченно большой в этой точке. В результате логарифмическое правдоподобие стремится к $+\infty$, что делает решение некорректным.

Математически, если для компоненты $k$ все $\gamma_{ik} \approx 0$, кроме нескольких точек, для которых $\mathbf{x}_i$ почти совпадают, то взвешенная выборочная ковариация (6) может иметь определитель, близкий к нулю. Тогда слагаемое $-\frac{1}{2}\log|\boldsymbol{\Sigma}_k|$ в $Q$ и, соответственно, в наблюдаемом правдоподобии, стремится к $+\infty$.

**Методы борьбы:**
- **Регуляризация ковариаций:** замена (6) на
  $$
  \boldsymbol{\Sigma}_k^{(t+1)} \leftarrow \boldsymbol{\Sigma}_k^{(t+1)} + \epsilon \mathbf{I},
  $$
  где $\epsilon > 0$ мало. Это гарантирует, что минимальное собственное значение ковариационной матрицы не меньше $\epsilon$, и правдоподобие остаётся конечным.
- **Ограничение структуры ковариаций:** использование диагональных или сферических ковариационных матриц уменьшает число параметров и снижает риск вырождения.
- **Байесовский подход:** введение априорных распределений (см. ниже) автоматически регуляризует решение.
- **Удаление компонент:** если эффективный размер компоненты $N_k = \sum_i \gamma_{ik}$ падает ниже порога (например, $N_k < 1$ или $N_k < d+1$), компонента исключается, а её вес распределяется между оставшимися.

#### 4. Ускорение сходимости EM

Хотя EM гарантирует монотонное увеличение правдоподобия, его сходимость часто бывает линейной и медленной вблизи оптимума. Существуют методы ускорения:

**Ускорение Эйткена (Aitken’s acceleration).** Пусть $\ell^{(t)} = \log p(\mathbf{X} \mid \boldsymbol{\theta}^{(t)})$ — значения логарифмического правдоподобия на итерациях. Предполагая, что последовательность $\ell^{(t)}$ сходится как геометрическая прогрессия, можно оценить предел и экстраполировать. Обновлённая оценка предела:

$$
\ell_{\infty} = \ell^{(t)} + \frac{\ell^{(t+1)} - \ell^{(t)}}{1 - c^{(t)}}, \quad c^{(t)} = \frac{\ell^{(t+1)} - \ell^{(t)}}{\ell^{(t)} - \ell^{(t-1)}}.
$$

Затем параметры могут быть скорректированы с учётом этой экстраполяции.

**Over-relaxed EM** — шаг M-шага заменяется на $\boldsymbol{\theta}^{(t+1)} = \boldsymbol{\theta}^{(t)} + \eta (\boldsymbol{\theta}^* - \boldsymbol{\theta}^{(t)})$, где $\boldsymbol{\theta}^*$ — результат стандартного M-шага, а $\eta > 1$ — коэффициент релаксации.

**Гибридные методы:** вблизи сходимости переключаются на методы второго порядка (Ньютон, квази-ньютоновские) или градиентный спуск с адаптивным шагом, так как EM в этой области может делать очень маленькие шаги.

Для больших данных разработаны стохастические варианты (online EM, incremental EM) и вариационный EM, которые аппроксимируют апостериорное распределение и ускоряют вычисления.

#### 5. Углублённый взгляд: байесовский GMM и автоматический выбор $K$

Байесовский подход к GMM позволяет не только регуляризовать параметры, но и автоматически определять эффективное число компонент. Вводится априорное распределение на параметры смеси:

- На веса $\boldsymbol{\pi}$ — симметричное распределение Дирихле:
  $$
  p(\boldsymbol{\pi}) = \operatorname{Dirichlet}(\boldsymbol{\pi} \mid \alpha_0, \dots, \alpha_0), \qquad \alpha_0 > 0.
  $$
  При $\alpha_0 < 1$ априорное распределение способствует разреженности весов, позволяя некоторым компонентам иметь пренебрежимо малые веса.
- На средние и ковариации — сопряжённое нормально-обратное-вишартовское распределение:
  $$
  p(\boldsymbol{\mu}_k, \boldsymbol{\Sigma}_k) = \mathcal{N}\bigl(\boldsymbol{\mu}_k \mid \mathbf{m}_0, \beta_0^{-1} \boldsymbol{\Sigma}_k\bigr) \cdot \operatorname{Inv-Wishart}(\boldsymbol{\Sigma}_k \mid \nu_0, \mathbf{W}_0).
  $$

Апостериорное распределение всех параметров и латентных переменных не имеет замкнутого вида, но может быть аппроксимировано **вариационным байесовским EM** (VBEM). В VBEM на E-шаге вычисляются приближённые апостериорные вероятности $\gamma_{ik}$, а на M-шаге обновляются гиперпараметры апостериорных распределений. Преимущество такого подхода: при $\alpha_0 < 1$ компоненты, не объясняющие данные, автоматически «выключаются» — их апостериорный вес стремится к нулю, и эффективное число компонент определяется данными. Такой метод известен как **вариационная байесовская смесь гауссовских распределений**.

Байесовский GMM устраняет необходимость явного выбора $K$ и проблему вырожденных ковариаций, поскольку априорное распределение действует как регуляризатор. Он является естественным мостом к непараметрическим байесовским моделям, таким как процесс Дирихле (Dirichlet Process GMM), где число компонент может расти с объёмом данных.

### Реализация EM для гауссовской смеси на Python с нуля и визуализация

Создадим класс `GaussianMixtureEM`, реализующий EM-алгоритм для модели гауссовской смеси. Все ключевые формулы E- и M-шагов будут воплощены непосредственно в коде, а численная устойчивость обеспечена переходом в логарифмическое пространство и регуляризацией ковариационных матриц.

Запустите следующую ячейку – она содержит полное определение класса.

```python
# @title Класс GaussianMixtureEM
import numpy as np
from scipy.stats import multivariate_normal
from sklearn.cluster import KMeans

class GaussianMixtureEM:
    def __init__(self, n_components=2, max_iter=200, tol=1e-4, reg=1e-6, random_state=None):
        self.K = n_components
        self.max_iter = max_iter
        self.tol = tol
        self.reg = reg
        self.random_state = random_state
        self.weights_ = None
        self.means_ = None
        self.covariances_ = None
        self.log_likelihood_ = []

    def _initialize(self, X):
        n, d = X.shape
        rng = np.random.RandomState(self.random_state)
        kmeans = KMeans(n_clusters=self.K, n_init=1, random_state=self.random_state)
        kmeans.fit(X)
        self.means_ = kmeans.cluster_centers_
        self.covariances_ = np.zeros((self.K, d, d))
        labels = kmeans.labels_
        for k in range(self.K):
            Xk = X[labels == k]
            if len(Xk) > 1:
                self.covariances_[k] = np.cov(Xk, rowvar=False) + self.reg * np.eye(d)
            else:
                self.covariances_[k] = np.eye(d)
        self.weights_ = np.ones(self.K) / self.K

    def _e_step(self, X):
        n, d = X.shape
        log_resp = np.zeros((n, self.K))
        for k in range(self.K):
            mvn = multivariate_normal(mean=self.means_[k], cov=self.covariances_[k], allow_singular=True)
            log_resp[:, k] = mvn.logpdf(X) + np.log(self.weights_[k] + 1e-300)
        log_resp_max = np.max(log_resp, axis=1, keepdims=True)
        log_resp -= log_resp_max
        resp = np.exp(log_resp)
        resp /= resp.sum(axis=1, keepdims=True)
        return resp

    def _m_step(self, X, resp):
        n, d = X.shape
        Nk = resp.sum(axis=0) + 1e-15
        self.weights_ = Nk / n
        self.means_ = (resp.T @ X) / Nk[:, np.newaxis]
        for k in range(self.K):
            diff = X - self.means_[k]
            cov_k = (diff * resp[:, k][:, np.newaxis]).T @ diff / Nk[k]
            cov_k += self.reg * np.eye(d)
            self.covariances_[k] = cov_k

    def _compute_log_likelihood(self, X):
        n = X.shape[0]
        log_lik = 0.0
        for i in range(n):
            prob = 0.0
            for k in range(self.K):
                prob += self.weights_[k] * multivariate_normal.pdf(
                    X[i], mean=self.means_[k], cov=self.covariances_[k], allow_singular=True)
            log_lik += np.log(prob + 1e-300)
        return log_lik

    def fit(self, X):
        self._initialize(X)
        self.log_likelihood_ = []
        prev_log_lik = -np.inf
        for it in range(self.max_iter):
            resp = self._e_step(X)
            self._m_step(X, resp)
            log_lik = self._compute_log_likelihood(X)
            self.log_likelihood_.append(log_lik)
            if np.abs(log_lik - prev_log_lik) < self.tol:
                break
            prev_log_lik = log_lik
        return self

    def predict(self, X):
        resp = self._e_step(X)
        return np.argmax(resp, axis=1)

    def predict_proba(self, X):
        return self._e_step(X)
```

#### Тестирование на синтетических данных, сравнение с sklearn и анализ сходимости

Следующая ячейка генерирует смесь двух гауссиан, обучает нашу модель, строит эллипсы компонент, сравнивает результат с `sklearn.mixture.GaussianMixture`, демонстрирует проблему вырождения и визуализирует рост логарифмического правдоподобия. Запустите её, чтобы увидеть все результаты сразу.

```python
# @title Эксперименты с GaussianMixtureEM
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
from sklearn.mixture import GaussianMixture as SklearnGMM

# ==============================================
# Вспомогательная функция для рисования эллипсов
# ==============================================
def plot_gmm(gmm, X, title=""):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[:, 0], X[:, 1], c=gmm.predict(X), cmap='viridis', alpha=0.6, s=20)
    for k in range(gmm.K):
        mean = gmm.means_[k]
        cov = gmm.covariances_[k]
        vals, vecs = np.linalg.eigh(cov)
        angle = np.degrees(np.arctan2(*vecs[:, 0][::-1]))
        width, height = 2 * np.sqrt(vals)
        ell = Ellipse(xy=mean, width=width, height=height, angle=angle,
                      edgecolor='red', facecolor='none', linewidth=2)
        plt.gca().add_patch(ell)
    plt.title(title)
    plt.grid(True)
    plt.show()

# ==============================================
# Генерация данных
# ==============================================
np.random.seed(42)
n_samples = 300
mean1 = [2, 2]
cov1 = [[1.5, 0.8], [0.8, 1.0]]
X1 = np.random.multivariate_normal(mean1, cov1, n_samples // 2)

mean2 = [-2, -2]
cov2 = [[1.0, -0.5], [-0.5, 1.5]]
X2 = np.random.multivariate_normal(mean2, cov2, n_samples // 2)

X = np.vstack([X1, X2])
y_true = np.array([0]*(n_samples//2) + [1]*(n_samples//2))

# ==============================================
# Обучение нашей модели
# ==============================================
gmm = GaussianMixtureEM(n_components=2, reg=1e-6, random_state=0)
gmm.fit(X)
plot_gmm(gmm, X, "Наша реализация GMM")

# ==============================================
# Сравнение с sklearn
# ==============================================
gmm_sk = SklearnGMM(n_components=2, covariance_type='full', reg_covar=1e-6, random_state=0)
gmm_sk.fit(X)

print("Средние (наши):\n", gmm.means_)
print("Средние (sklearn):\n", gmm_sk.means_)
print("Веса (наши):\n", gmm.weights_)
print("Веса (sklearn):\n", gmm_sk.weights_)
print("Лог. правдоподобие (наши):", gmm.log_likelihood_[-1])
print("Лог. правдоподобие (sklearn):", gmm_sk.lower_bound_)

# ==============================================
# Демонстрация проблемы вырождения и регуляризации
# ==============================================
np.random.seed(123)
X_small = np.random.multivariate_normal([5, 5], [[0.05, 0], [0, 0.05]], 5)
X_large = np.random.multivariate_normal([0, 0], [[2, 0.5], [0.5, 2]], 200)
X_bad = np.vstack([X_small, X_large])

# Без регуляризации
gmm_bad = GaussianMixtureEM(n_components=2, reg=0, max_iter=100, random_state=0)
gmm_bad.fit(X_bad)
print("Без регуляризации, последний log-lik:", gmm_bad.log_likelihood_[-1])

# С регуляризацией
gmm_reg = GaussianMixtureEM(n_components=2, reg=1e-4, max_iter=100, random_state=0)
gmm_reg.fit(X_bad)
print("С регуляризацией, последний log-lik:", gmm_reg.log_likelihood_[-1])

# ==============================================
# Визуализация сходимости
# ==============================================
plt.figure(figsize=(8,5))
plt.plot(gmm.log_likelihood_, marker='o')
plt.xlabel('Итерация')
plt.ylabel('Логарифмическое правдоподобие')
plt.title('Сходимость EM для GMM')
plt.grid(True)
plt.show()
```

### Вариации EM-алгоритма: жёсткая кластеризация, факторный анализ и обработка пропусков

Классический EM-алгоритм для смеси гауссовских распределений — лишь один из представителей обширного семейства методов, основанных на идее чередования ожидания и максимизации. Модифицируя модель или способ выполнения E-шага, мы получаем такие известные алгоритмы, как k-means, вероятностный PCA и процедуры заполнения пропущенных значений. В этом разделе мы рассмотрим эти вариации и покажем, как EM связывает их в единую концептуальную схему.

#### 1. Жёсткий EM и алгоритм k-средних

Рассмотрим предельный случай GMM, когда ковариационные матрицы всех компонент стремятся к нулю, оставаясь пропорциональными единичной: $\boldsymbol{\Sigma}_k = \epsilon \mathbf{I}$, $\epsilon \to 0$. При этом гауссовская плотность $\mathcal{N}(\mathbf{x}_i \mid \boldsymbol{\mu}_k, \epsilon \mathbf{I})$ концентрируется вокруг среднего $\boldsymbol{\mu}_k$, и в пределе $\epsilon \to 0$ логарифм плотности ведёт себя как $-\frac{1}{2\epsilon}\|\mathbf{x}_i - \boldsymbol{\mu}_k\|^2 + O(1)$. Тогда ответственности (2) из предыдущего раздела принимают вид

$$
\gamma_{ik} = \frac{\pi_k \exp\!\bigl(-\frac{1}{2\epsilon}\|\mathbf{x}_i - \boldsymbol{\mu}_k\|^2\bigr)}{\sum_j \pi_j \exp\!\bigl(-\frac{1}{2\epsilon}\|\mathbf{x}_i - \boldsymbol{\mu}_j\|^2\bigr)} \;\xrightarrow{\epsilon \to 0}\; \begin{cases}
1, & \text{если } k = \arg\min_j \|\mathbf{x}_i - \boldsymbol{\mu}_j\|, \\
0, & \text{иначе}.
\end{cases}
$$

Таким образом, мягкое присвоение (soft assignment) вырождается в жёсткое (hard assignment) — каждая точка относится ровно к одной компоненте, а именно к ближайшему центру. M-шаг при этом сводится к пересчёту центров как средних арифметических точек, приписанных к каждой компоненте. Это в точности алгоритм **k-средних** (k-means), где веса компонент $\pi_k$ остаются фиксированными (или пересчитываются как доли точек) и обычно полагаются равными.

Таким образом, k-means можно рассматривать как EM для GMM с вырожденными ковариациями. Преимущество GMM перед k-means — сохранение информации о неопределённости принадлежности (мягкое присвоение) и моделирование полной ковариационной структуры кластеров.

#### 2. EM для факторного анализа (вероятностный PCA)

Модель факторного анализа предполагает, что наблюдаемые данные $\mathbf{x} \in \mathbb{R}^d$ порождаются линейным преобразованием низкоразмерного латентного вектора $\mathbf{z} \sim \mathcal{N}(\mathbf{0}, \mathbf{I}_q)$ плюс гауссовский шум:

$$
\mathbf{x} = \mathbf{W} \mathbf{z} + \boldsymbol{\mu} + \boldsymbol{\varepsilon}, \qquad \boldsymbol{\varepsilon} \sim \mathcal{N}(\mathbf{0}, \sigma^2 \mathbf{I}_d),
$$

где $\mathbf{W} \in \mathbb{R}^{d \times q}$ — матрица нагрузок, $\boldsymbol{\mu} \in \mathbb{R}^d$ — среднее. Совместное распределение $(\mathbf{x}, \mathbf{z})$ является гауссовским, и наблюдаемое правдоподобие $p(\mathbf{x})$ также гауссовское: $\mathbf{x} \sim \mathcal{N}(\boldsymbol{\mu}, \mathbf{W}\mathbf{W}^T + \sigma^2 \mathbf{I}_d)$.

Прямая максимизация наблюдаемого правдоподобия по $\mathbf{W}, \sigma^2$ возможна аналитически (через собственные векторы), но EM предлагает простую итеративную альтернативу. На E-шаге вычисляется апостериорное распределение латентных переменных:

$$
p(\mathbf{z}_i \mid \mathbf{x}_i, \boldsymbol{\theta}) = \mathcal{N}\bigl( \mathbf{z}_i \mid \mathbf{M}^{-1} \mathbf{W}^T (\mathbf{x}_i - \boldsymbol{\mu}), \; \sigma^2 \mathbf{M}^{-1} \bigr), \qquad \mathbf{M} = \mathbf{W}^T \mathbf{W} + \sigma^2 \mathbf{I}_q.
$$

Достаточными статистиками являются $\mathbb{E}[\mathbf{z}_i]$ и $\mathbb{E}[\mathbf{z}_i \mathbf{z}_i^T]$. На M-шаге параметры обновляются как

$$
\begin{aligned}
\mathbf{W}_{\text{new}} &= \Bigl( \sum_i (\mathbf{x}_i - \boldsymbol{\mu}) \mathbb{E}[\mathbf{z}_i]^T \Bigr) \Bigl( \sum_i \mathbb{E}[\mathbf{z}_i \mathbf{z}_i^T] \Bigr)^{-1}, \\
\sigma^2_{\text{new}} &= \frac{1}{n d} \sum_i \Bigl( \|\mathbf{x}_i - \boldsymbol{\mu}\|^2 - 2 \mathbb{E}[\mathbf{z}_i]^T \mathbf{W}_{\text{new}}^T (\mathbf{x}_i - \boldsymbol{\mu}) + \operatorname{tr}\bigl( \mathbb{E}[\mathbf{z}_i \mathbf{z}_i^T] \mathbf{W}_{\text{new}}^T \mathbf{W}_{\text{new}} \bigr) \Bigr).
\end{aligned}
$$

Этот EM-алгоритм для вероятностного PCA численно устойчив и легко обобщается на смеси факторных анализаторов.

#### 3. EM для данных с пропущенными значениями

Пропуски в данных — естественная область применения EM. Если некоторые компоненты вектора $\mathbf{x}_i$ отсутствуют, их можно рассматривать как латентные переменные. В предположении многомерной нормальности (или GMM) полное правдоподобие имеет удобный вид, а на E-шаге для каждого наблюдения вычисляются условные ожидания пропущенных признаков при текущих параметрах. Пусть $\mathbf{x}_i = (\mathbf{x}_i^{\text{obs}}, \mathbf{x}_i^{\text{mis}})$. Тогда

$$
\mathbb{E}[\mathbf{x}_i^{\text{mis}} \mid \mathbf{x}_i^{\text{obs}}, \boldsymbol{\theta}] = \boldsymbol{\mu}_{\text{mis}} + \boldsymbol{\Sigma}_{\text{mis},\text{obs}} \boldsymbol{\Sigma}_{\text{obs},\text{obs}}^{-1} (\mathbf{x}_i^{\text{obs}} - \boldsymbol{\mu}_{\text{obs}}),
$$

а также вычисляются вторые моменты. На M-шаге по дополненным данным пересчитываются выборочное среднее и ковариационная матрица (или параметры GMM). Итерации продолжаются до сходимости. Этот метод, известный как EM-алгоритм для восстановления пропусков, является стандартным в статистике и даёт оценки максимального правдоподобия для параметров при MAR-механизме пропусков.

#### 4. Пример на Python: EM для дополнения пропусков в GMM

Приведём краткую реализацию функции, которая в рамках обученного GMM заменяет пропущенные значения (`NaN`) на условные ожидания по текущим параметрам. Для простоты будем считать, что GMM уже обучена на полных данных, но в тестовой точке есть пропуски.



### Вариации EM-алгоритма: жёсткая кластеризация, факторный анализ и обработка пропусков

Классический EM для смеси гауссиан – лишь одна из многих моделей, использующих попеременное ожидание и максимизацию. Ниже мы реализуем две важные вариации: **вероятностный PCA (факторный анализ)** и **дополнение пропущенных значений** с помощью EM. Для каждой из них приведён полный код с наглядными визуализациями.

```python
# @title EM для Probabilistic PCA (факторный анализ)
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv, norm

class ProbabilisticPCA:
    def __init__(self, n_components=2, max_iter=100, tol=1e-4):
        self.q = n_components   # размерность латентного пространства
        self.max_iter = max_iter
        self.tol = tol
        self.W = None           # матрица нагрузок (d x q)
        self.sigma2 = None      # дисперсия шума
        self.mu = None          # среднее

    def fit(self, X):
        n, d = X.shape
        self.mu = np.mean(X, axis=0)
        Xc = X - self.mu
        # инициализация случайной матрицей
        rng = np.random.RandomState(42)
        self.W = rng.randn(d, self.q) * 0.1
        self.sigma2 = np.var(Xc)
        prev_log_lik = -np.inf
        for it in range(self.max_iter):
            # E-шаг: вычисляем моменты латентных переменных
            M = self.W.T @ self.W + self.sigma2 * np.eye(self.q)
            Minv = inv(M)
            # E[z_i] и E[z_i z_i^T]
            Ez = Xc @ self.W @ Minv                 # n x q
            Ezz = self.sigma2 * Minv + Ez.T @ Ez / n  # q x q (усреднённое)

            # M-шаг: обновление W и sigma2
            self.W = Xc.T @ Ez @ inv(Ezz * n)        # d x q
            # Можно также использовать более эффективную формулу:
            # self.W = (Xc.T @ Ez) @ inv(Ezz * n)
            recon = Xc - Ez @ self.W.T
            self.sigma2 = np.mean(recon ** 2) + self.sigma2 / self.q * np.trace(self.W.T @ self.W)

            # Логарифмическое правдоподобие (приближённо)
            # Для проверки сходимости используем изменение sigma2 и W
            log_lik = -0.5 * n * (d * np.log(2 * np.pi) + d + np.log(np.linalg.det(self.W @ self.W.T + self.sigma2 * np.eye(d))))
            if np.abs(log_lik - prev_log_lik) < self.tol:
                break
            prev_log_lik = log_lik

    def transform(self, X):
        Xc = X - self.mu
        M = self.W.T @ self.W + self.sigma2 * np.eye(self.q)
        return Xc @ self.W @ inv(M)

    def reconstruct(self, X):
        Z = self.transform(X)
        return Z @ self.W.T + self.mu

# ========== Генерация данных и демонстрация ==========
from sklearn.datasets import make_low_rank_matrix
np.random.seed(42)
n, d, q = 200, 10, 2
# Создаём низкоранговую матрицу данных
true_W = np.random.randn(d, q)
Z = np.random.randn(n, q)
X = Z @ true_W.T + np.random.randn(n, d) * 0.5

ppca = ProbabilisticPCA(n_components=q, max_iter=200)
ppca.fit(X)
X_recon = ppca.reconstruct(X)

# Визуализация
fig, ax = plt.subplots(1, 2, figsize=(12,5))
ax[0].imshow(X[:50], aspect='auto', cmap='viridis')
ax[0].set_title('Исходные данные (первые 50 точек)')
ax[1].imshow(X_recon[:50], aspect='auto', cmap='viridis')
ax[1].set_title('Восстановленные (PPCA)')
plt.tight_layout()
plt.show()

print("Ошибка восстановления (MSE):", np.mean((X - X_recon)**2))
```

#### EM для дополнения пропусков с помощью GMM

Если часть признаков отсутствует, их можно восстановить, опираясь на условные ожидания, вычисленные по наблюдаемой части данных. Следующая функция принимает обученную модель GMM (наш класс `GaussianMixtureEM`) и массив с пропусками (NaN), а возвращает дополненные значения.

Для эксперимента мы искусственно внесём пропуски в синтетическую выборку и сравним восстановленные значения с истинными.

```python
# @title Дополнение пропусков с помощью GMM (EM) – исправленная версия
import numpy as np
from scipy.stats import multivariate_normal

def impute_missing_gmm(gmm, X_missing):
    """
    Дополняет пропуски (NaN) условными ожиданиями по GMM.
    Если все признаки отсутствуют, использует безусловное среднее смеси.
    """
    X_imputed = X_missing.copy()
    for i in range(len(X_missing)):
        missing_mask = np.isnan(X_missing[i])
        if not np.any(missing_mask):
            continue
        observed_mask = ~missing_mask
        x_obs = X_missing[i, observed_mask]

        # Случай: все признаки пропущены – заполняем средним смеси
        if len(x_obs) == 0:
            mean_all = np.zeros_like(X_missing[i])
            for k in range(gmm.K):
                mean_all += gmm.weights_[k] * gmm.means_[k]
            X_imputed[i, missing_mask] = mean_all[missing_mask]
            continue

        # Апостериорные вероятности компонент только по наблюдаемым
        log_resp = np.zeros(gmm.K)
        for k in range(gmm.K):
            mu_k_obs = gmm.means_[k][observed_mask]
            cov_k_obs = gmm.covariances_[k][observed_mask][:, observed_mask]
            # добавим небольшую регуляризацию для устойчивости
            cov_k_obs += 1e-6 * np.eye(cov_k_obs.shape[0])
            log_p = multivariate_normal.logpdf(x_obs, mean=mu_k_obs, cov=cov_k_obs, allow_singular=True)
            log_resp[k] = log_p + np.log(gmm.weights_[k] + 1e-300)

        # Устойчивое преобразование в вероятности
        log_resp_max = np.max(log_resp)
        resp = np.exp(log_resp - log_resp_max)
        resp /= resp.sum()

        # Условное среднее пропущенных признаков
        imputed_val = np.zeros(np.sum(missing_mask))
        for k in range(gmm.K):
            mu_k_mis = gmm.means_[k][missing_mask]
            mu_k_obs = gmm.means_[k][observed_mask]
            cov_k = gmm.covariances_[k]
            cov_oo = cov_k[observed_mask][:, observed_mask] + 1e-6 * np.eye(x_obs.shape[0])
            cov_mo = cov_k[missing_mask][:, observed_mask]
            cond_mean = mu_k_mis + cov_mo @ np.linalg.inv(cov_oo) @ (x_obs - mu_k_obs)
            imputed_val += resp[k] * cond_mean
        X_imputed[i, missing_mask] = imputed_val
    return X_imputed

# ========== Демонстрация ==========
# Генерация данных и обучение GMM (если ещё не сделано)
np.random.seed(42)
n_samples = 300
X1 = np.random.multivariate_normal([2, 2], [[1.5, 0.8], [0.8, 1.0]], n_samples//2)
X2 = np.random.multivariate_normal([-2, -2], [[1.0, -0.5], [-0.5, 1.5]], n_samples//2)
X_full = np.vstack([X1, X2])

# Обучаем GMM (убедитесь, что класс GaussianMixtureEM определён в предыдущей ячейке)
gmm_full = GaussianMixtureEM(n_components=2, reg=1e-6, random_state=0).fit(X_full)

# Вносим пропуски (20% значений)
X_miss = X_full.copy()
mask = np.random.rand(*X_miss.shape) < 0.2
X_miss[mask] = np.nan

# Восстанавливаем
X_imp = impute_missing_gmm(gmm_full, X_miss)

# Визуализация
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.scatter(X_full[mask], X_imp[mask], alpha=0.6, edgecolors='k')
plt.plot([-4,6], [-4,6], 'r--', label='Идеальное восстановление')
plt.xlabel('Истинное значение')
plt.ylabel('Восстановленное значение')
plt.title('Дополнение пропусков с помощью GMM')
plt.legend()
plt.grid(True)
plt.show()

print("Среднеквадратичная ошибка на пропущенных позициях:",
      np.mean((X_full[mask] - X_imp[mask])**2))
```

#### 5. Связь с байесовскими моделями и вариационным выводом

EM-алгоритм является частным случаем **вариационного вывода**. В вариационном байесовском подходе (VB) апостериорное распределение латентных переменных аппроксимируется некоторым вариационным семейством $q(\mathbf{Z})$. Для EM это семейство точечное: $q(\mathbf{Z}) = \delta_{\mathbf{Z} = \mathbf{Z}^*}$, а оптимизация по $\mathbf{Z}^*$ эквивалентна E-шагу. В более общем VB, в частности для байесовского GMM, вариационное распределение параметров и латентных переменных выбирается факторизованным, и оптимизация чередуется аналогично EM — это вариационный EM (VBEM). Преимущество VBEM в автоматическом выборе числа компонент через разреживающие априорные распределения и встроенную регуляризацию.

Таким образом, EM-алгоритм не только решает задачу максимизации правдоподобия в присутствии латентных переменных, но и служит мостом к целому спектру современных вероятностных методов, включая вариационный вывод и генеративные модели с глубокими латентными переменными (VAE). Изучение EM и его расширений формирует фундамент для понимания и разработки алгоритмов машинного обучения, основанных на правдоподобии.
