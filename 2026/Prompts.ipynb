### Промпт 1A (Основы и матричный аппарат)

```text
Напиши первую часть первой половины главы «Множественная линейная регрессия» для учебного пособия.
Стиль: основной текст интуитивный, с яркими аналогиями; для профессионалов — углублённые вставки «Углублённый взгляд».
Используй длинные тире (—) и кавычки-ёлочки («»). Обозначение дисперсии — $\mathrm{Var}$.

Начни с мостика: «В предыдущей главе модель $y = w_0 + w_1 x + \varepsilon$ описывала прямую. Теперь признаков много, и прямая превращается в гиперплоскость».

Разделы (строго по порядку):

1. Постановка задачи множественной линейной регрессии
   - Модель $t = w_0 + w_1 x_1 + \dots + w_D x_D + \varepsilon$.
   - Ошибка $\varepsilon \sim \mathcal{N}(0,\sigma^2)$, условное распределение $p(t|\mathbf{x},\mathbf{w}) = \mathcal{N}(w_0+\sum_{j=1}^D w_j x_j,\sigma^2)$.
   - Цель: по выборке $\{(\mathbf{x}^{(i)}, t^{(i)})\}_{i=1}^n$ найти $\mathbf{w} = (w_0, w_1, \dots, w_D)^T$.
   - Коротко упомяни интерпретируемость: знак и величина коэффициентов имеют прямой смысл, что важно для объяснения модели заказчику.

2. Матричная запись модели
   - Вектор $\mathbf{t}\in\mathbb{R}^n$, вектор $\mathbf{w}\in\mathbb{R}^{D+1}$.
   - Матрица плана $X$ размера $n \times (D+1)$ с единичным столбцом:
     $$
     X = \begin{pmatrix}
     1 & x_1^{(1)} & \dots & x_D^{(1)} \\
     \vdots & \vdots & \ddots & \vdots \\
     1 & x_1^{(n)} & \dots & x_D^{(n)}
     \end{pmatrix}.
     $$
   - Модель: $\mathbf{t} = X\mathbf{w} + \boldsymbol{\varepsilon}$.

3. Базисные функции
   - Идея: нелинейные преобразования $\phi_j(\mathbf{x})$ для улавливания нелинейных зависимостей при линейности по параметрам.
     Подчеркни: модель остаётся линейной по $\mathbf{w}$, но может описывать сложные нелинейные закономерности.
   - Модель: $y(\mathbf{x},\mathbf{w}) = \sum_{j=0}^{M-1} w_j \phi_j(\mathbf{x}) = \mathbf{w}^T \boldsymbol{\phi}(\mathbf{x})$, $\phi_0(\mathbf{x})=1$.
   - Матрица плана $\Phi$ размера $n \times M$.
   - Приведи три примера базисных функций:
     * Полиномиальные: $\phi_k(x) = x^k$.
     * Радиально-базисные функции (RBF): $\phi_j(\mathbf{x}) = \exp(-\|\mathbf{x}-\boldsymbol{\mu}_j\|^2/(2\sigma^2))$. Интуиция: «колоколы» вокруг центров $\boldsymbol{\mu}_j$, моделируют локальные особенности.
     * Практический пример с логарифмом и сигнум-функцией: если мы подозреваем, что целевая переменная зависит от $\log x_1$ и от того, разные ли знаки у $x_1$ и $x_2$, модель может быть записана как
       $$y \approx w_1 x_1 + w_2 x_2 + w_3 \log x_1 + w_4 \,\text{sgn}(x_1 x_2) + w_0.$$
       Так из двумерной нелинейной задачи получается четырёхмерная линейная регрессия.
   - Сделай короткий мостик к категориальным признакам: а как быть, если признак категориальный (например, порода питомца)? Его тоже можно рассматривать как набор индикаторных базисных функций, но есть тонкость с линейной зависимостью — подробнее в разделе о подготовке признаков.
   - Везде далее используется $\Phi$.

4. Метод наименьших квадратов и вероятностное обоснование
   - Функция потерь: $E(\mathbf{w}) = \frac{1}{2} \sum_{i=1}^n (t^{(i)} - \mathbf{w}^T \boldsymbol{\phi}(\mathbf{x}^{(i)}))^2$.
   - Вероятностная модель: $\varepsilon_i \sim \mathcal{N}(0,\sigma^2)$ независимы, тогда $t^{(i)}|\mathbf{x}^{(i)} \sim \mathcal{N}(\mathbf{w}^T \boldsymbol{\phi}(\mathbf{x}^{(i)}), \sigma^2)$.
   - Функция правдоподобия:
     $$
     L(\mathbf{w}) = \prod_{i=1}^n \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(t^{(i)} - \mathbf{w}^T \boldsymbol{\phi}(\mathbf{x}^{(i)}))^2}{2\sigma^2}\right).
     $$
   - Логарифм:
     $$
     \log L(\mathbf{w}) = -\frac{n}{2}\log(2\pi\sigma^2) - \frac{1}{2\sigma^2}\sum_{i=1}^n (t^{(i)} - \mathbf{w}^T \boldsymbol{\phi}(\mathbf{x}^{(i)}))^2.
     $$
   - Вывод: максимизация $\log L$ эквивалентна минимизации SSE. МНК-оценки совпадают с оценками максимального правдоподобия при нормальных ошибках.
   - **Углублённый взгляд:** MLE для $\sigma^2$. Дифференцируем $\log L$ по $\sigma^2$:
     $\hat{\sigma}^2_{\text{MLE}} = \frac{1}{n}\sum (t^{(i)} - \hat{t}^{(i)})^2$ — смещённая оценка.
     Несмещённая: $s^2 = \frac{SSE}{n-M}$, потеря $M$ степеней свободы (оценено $M$ параметров).

В конце этой части вопросов для самопроверки нет — они будут в следующей части. Текст без кода.
```

### Промпт 1B (Аналитическое решение и вычислительные аспекты)

```text
Продолжи первую часть главы «Множественная линейная регрессия» сразу после раздела о вероятностном обосновании (матрицы $\Phi$, функция потерь, MLE уже введены).
Стиль тот же: интуитивный текст с вставками «Углублённый взгляд».

Разделы (строго по порядку):

1. Градиент целевой функции
   - Функция потерь $E(\mathbf{w}) = \frac{1}{2} \sum_{i=1}^n (t_i - \mathbf{w}^T \boldsymbol{\phi}(\mathbf{x}_i))^2$ (для краткости пишем $t_i$).
   - Подробный вывод $\partial E/\partial w_k$:
     * Правило цепочки, производная квадрата: $\partial E/\partial w_k = \sum_i (t_i - \mathbf{w}^T \boldsymbol{\phi}(\mathbf{x}_i)) \cdot (-\phi_k(\mathbf{x}_i))$.
     * Упрощаем: $\partial E/\partial w_k = \sum_i (\mathbf{w}^T \boldsymbol{\phi}(\mathbf{x}_i) - t_i) \phi_k(\mathbf{x}_i)$.
   - Сборка в вектор градиента:
     $$
     \nabla_{\mathbf{w}}E = \begin{pmatrix}
     \sum_i (\mathbf{w}^T \boldsymbol{\phi}(\mathbf{x}_i) - t_i)\phi_0(\mathbf{x}_i) \\
     \vdots \\
     \sum_i (\mathbf{w}^T \boldsymbol{\phi}(\mathbf{x}_i) - t_i)\phi_{M-1}(\mathbf{x}_i)
     \end{pmatrix} = \sum_{i=1}^n (\mathbf{w}^T \boldsymbol{\phi}(\mathbf{x}_i) - t_i) \boldsymbol{\phi}(\mathbf{x}_i).
     $$
   - Переход к матричной форме: вводим $\Phi$, $\mathbf{t}$, предсказания $\hat{\mathbf{t}}=\Phi\mathbf{w}$, ошибки $\mathbf{e}=\Phi\mathbf{w}-\mathbf{t}$.
     Убеждаемся, что $\Phi^T \mathbf{e}$ даёт нужный вектор (компонента $j$: $\sum_i \phi_j(\mathbf{x}_i) e_i$). Приводим проверку на игрушечном примере $n=2, M=3$ (выпиши матрицы, векторы, покажи равенство явно).
   - Итог: $\boxed{\nabla_{\mathbf{w}} E = \Phi^T (\Phi \mathbf{w} - \mathbf{t})}$.

2. Нормальные уравнения
   - Приравниваем градиент к нулю: $\Phi^T \Phi \mathbf{w} = \Phi^T \mathbf{t}$.
   - Если $\Phi^T \Phi$ обратима, решение $\hat{\mathbf{w}} = (\Phi^T \Phi)^{-1} \Phi^T \mathbf{t}$.
   - **Углублённый взгляд (альтернативный вывод):** можно исходить из транспонированной формы градиента. Показать цепочку:
     $(\nabla_{\mathbf{w}} E)^T = \mathbf{w}^T \Phi^T \Phi - \mathbf{t}^T \Phi^T = 0 \;\Rightarrow\; \mathbf{w}^T = \mathbf{t}^T \Phi^T (\Phi^T \Phi)^{-1} \;\Rightarrow\; \hat{\mathbf{w}} = (\Phi^T \Phi)^{-1} \Phi^T \mathbf{t}$.
   - Обсуждение невырожденности: $\Phi^T \Phi$ вырождена ⇔ столбцы $\Phi$ линейно зависимы ⇔ совершенная мультиколлинеарность. Пример: «площадь в м²» и «площадь в гектарах».

3. Вычислительная сложность и проблемы «точного» решения
   - Сложность $O(M^2 n + M^3)$. Расшифровка: $M^2 n$ на $\Phi^T\Phi$, $M^3$ на обращение.
   - Плохая обусловленность: число обусловленности $\kappa(\Phi) = \sigma_{\max}/\sigma_{\min}$, большая $\kappa$ → неустойчивость.
   - Геометрическая интуиция: линии уровня $E(\mathbf{w})$ — эллипсоиды. При высокой обусловленности они вытянуты, минимум в «овраге» неустойчив. *[Вставить рисунок: эллипсоиды для хорошо и плохо обусловленной матрицы]*.
   - **Углублённый взгляд (два метода):**
     * QR-разложение: $\Phi = QR$ ($Q$ ортогональна, $R$ верхнетреугольная). Тогда $\hat{\mathbf{w}} = R^{-1} Q^T \mathbf{t}$. Устойчивее, не формируется $\Phi^T\Phi$.
     * Сингулярное разложение (SVD): $\Phi = U \Sigma V^T$. Решение через псевдообратную: $\hat{\mathbf{w}} = V \Sigma^{-1} U^T \mathbf{t}$. Работает и при вырожденной $\Phi^T\Phi$. Подводит к идее регуляризации.

4. Простой численный пример с двумя признаками (без кода)
   - Табличка: 3 наблюдения, признаки $x_1$ (возраст автомобиля), $x_2$ (пробег), цель $t$ (цена).
   - Пошагово: записать $\Phi$ (с единичным столбцом), вычислить $\Phi^T\Phi$, $\Phi^T\mathbf{t}$, найти $(\Phi^T\Phi)^{-1}$ (с явным вычислением определителя и матрицы алгебраических дополнений) и $\hat{\mathbf{w}}$.
   - Интерпретировать коэффициенты ($w_0$ — базовая цена, $w_1$ — эффект возраста, $w_2$ — эффект пробега). Предупредить об осторожности при экстраполяции.

В конце — два блока вопросов для самопроверки (охватывают материал обеих частей 1A и 1B):
- Для начинающих (8 вопросов): от записи модели до интерпретации коэффициентов примера, что такое базисные функции, почему МНК и MLE дают одно и то же, что такое мультиколлинеарность, роль числа обусловленности.
- Для профессионалов (6 вопросов): полный вывод градиента, связь обусловленности с устойчивостью, сравнение QR и SVD, почему $s^2$ имеет $n-M$ степеней свободы, что делать при вырожденной $\Phi^T\Phi$, альтернативный вывод нормальных уравнений.
```

### Промпт 2 (Численные методы оптимизации)

```text
Напиши вторую часть главы «Множественная линейная регрессия», продолжая сразу после аналитического решения.
Стиль: интуитивный текст с углублёнными вставками «Углублённый взгляд».
Используй длинные тире, кавычки‑ёлочки, $\mathrm{Var}$.
Требуется код на Python (numpy, matplotlib, time). Импорты — единым блоком в начале.

Разделы строго по порядку:

1. Потребность в итеративных методах
   - Когда $n$ и $M$ велики, прямое обращение $\Phi^T\Phi$ слишком затратно. Градиентный спуск не требует матричных обращений.
   - Напомни градиент: $\nabla E(\mathbf{w}) = \Phi^T(\Phi\mathbf{w} - \mathbf{t})$.

2. Пакетный градиентный спуск (Batch Gradient Descent)
   - Правило: $\mathbf{w}^{(k+1)} = \mathbf{w}^{(k)} - \alpha \nabla E(\mathbf{w}^{(k)})$, $\alpha$ — скорость обучения.
   - Аналогия: спуск с горы в тумане.
   - Поведение $\alpha$: маленький — медленно, большой — осцилляции/расходимость. Описать траектории на эллипсоидах.
   - Явная матричная форма: $\mathbf{w}^{(k+1)} = \mathbf{w}^{(k)} - \alpha \Phi^T(\Phi\mathbf{w}^{(k)} - \mathbf{t})$.
   - Псевдокод.
   - **Углублённый взгляд:** сходимость для выпуклых функций, влияние числа обусловленности $\kappa$.

3. Стохастический градиентный спуск (SGD) и Mini-batch
   - SGD: обновление по одному случайному объекту $i$:
     $\mathbf{w}^{(k+1)} = \mathbf{w}^{(k)} - \alpha (\mathbf{w}^T \boldsymbol{\phi}(\mathbf{x}_i) - t_i) \boldsymbol{\phi}(\mathbf{x}_i)$.
   - Плюсы: быстрота, меньшая память, стохастичность помогает выходить из локальных минимумов.
   - Минусы: шумный градиент, трудный подбор $\alpha$.
   - Mini-batch: компромисс с размером батча $B$.
   - Расписания $\alpha$: step decay, exponential decay, cosine annealing.
   - **Углублённый взгляд:** теорема сходимости SGD (условия $\sum \alpha_k = \infty$, $\sum \alpha_k^2 < \infty$).

4. Неградиентные методы (краткий обзор)
   - Метод сопряжённых градиентов: использует текущий градиент и предыдущее направление, не требует гессиана.
   - Квазиньютоновские методы (L‑BFGS): аппроксимируют обратный гессиан по истории градиентов, сверхлинейная сходимость, ограниченная память.
   - Рекомендации: SGD для огромных данных, L‑BFGS для средних размеров.

5. Реализация SGD на Python с нуля
   - Сгенерировать данные: $n=10000$, $M=3$, истинные коэффициенты (например, 5, -2, 0.5, 1), добавить нормальный шум.
   - Код: функции градиента MSE (полный и для одного объекта), batch SGD, mini-batch SGD.
   - График: функция потерь по итерациям для разных $\alpha$ и размеров батча. Сравнить скорость сходимости.
   - Сравнить итоговые коэффициенты с аналитическим решением (через `np.linalg.solve`). Прокомментировать, что SGD при правильном $\alpha$ сходится к близким значениям, но для столь малого $M$ аналитическое решение практически мгновенно; настоящее преимущество SGD раскроется при больших $M$ и $n$ (см. раздел 6).

6. Сравнение вычислительной эффективности
   - Эксперимент с данными, где $M$ достаточно велико: например, сгенерировать данные при фиксированном $n=10000$ и варьировать $M$ от 5 до 100; а также при фиксированном $M=20$ варьировать $n$ от 1000 до 100000.
   - Для каждого сочетания замерить время (усреднить по 3–5 запускам, брать медиану) для аналитического решения, batch SGD и mini-batch SGD ($B=100$).
   - Построить графики «время от $n$» и «время от $M$». Отметить точку перехода, когда SGD становится быстрее.
   - Обсудить память: аналитическое решение $O(M^2)$, SGD $O(M)$.

В конце — два блока вопросов для самопроверки:
- Для начинающих (5): отличие SGD от пакетного, роль $\alpha$, что такое mini-batch, когда аналитическое решение неприменимо, зачем уменьшают $\alpha$.
- Для профессионалов (5): влияние $\kappa$ на скорость SGD, суть L‑BFGS, выбор расписания $\alpha$, компромисс смещения и дисперсии в mini-batch, почему SGD работает для невыпуклых функций (кратко).
```

```text
Напиши раздел «Регуляризация в линейной регрессии», который органично продолжает главу.
Стиль: интуитивные объяснения с углублёнными вставками «Углублённый взгляд».
Используй длинные тире, кавычки‑ёлочки, $\mathrm{Var}$.
Требуется код на Python (sklearn, numpy, matplotlib). Импорты — единым блоком в начале.

Содержание:

1. Зачем нужна регуляризация?
   - Переобучение при $M \gg n$, мультиколлинеарность, неустойчивость МНК.
   - Идея: штраф за величину коэффициентов. Геометрически: минимизация в ограниченной области (шар или ромб).

2. Ridge регрессия (L2)
   - Целевая функция: $E_{\text{ridge}} = \frac{1}{2}\|\mathbf{t} - \Phi\mathbf{w}\|^2 + \frac{\lambda}{2} \sum_{j=1}^{M-1} w_j^2$. Свободный член $w_0$ не штрафуется (или модифицируется единичная матрица).
   - Аналитическое решение: $\hat{\mathbf{w}}_{\text{ridge}} = (\Phi^T\Phi + \lambda I')^{-1} \Phi^T \mathbf{t}$, где $I'$ — с нулём на месте $w_0$.
   - Влияние $\lambda$: $\lambda \to 0$ → МНК, $\lambda \to \infty$ → веса стремятся к нулю, $w_0 \to \bar{t}$.
   - Выбор $\lambda$: кросс-валидация (`RidgeCV`).
   - **Углублённый взгляд:** байесовская интерпретация — гауссовский априор $w_j \sim \mathcal{N}(0,\tau^2)$, $\lambda = \sigma^2/\tau^2$.

3. Lasso регрессия (L1)
   - $E_{\text{lasso}} = \frac{1}{2}\|\mathbf{t} - \Phi\mathbf{w}\|^2 + \lambda \sum_{j=1}^{M-1} |w_j|$.
   - Свойство: разреженность — некоторые $w_j$ становятся точно нулями. Автоматический отбор признаков.
   - Геометрическая интерпретация: эллипсоид SSE + ромб ограничения, решение часто на осях.
   - Нет аналитического решения; координатный спуск.
   - **Углублённый взгляд:** байесовская интерпретация — лапласовский априор.

4. Elastic Net
   - $E = \frac{1}{2}\|\mathbf{t} - \Phi\mathbf{w}\|^2 + \lambda_1 \sum |w_j| + \frac{\lambda_2}{2} \sum w_j^2$.
   - Параметры в sklearn: $\alpha = \lambda_1 + \lambda_2$, $l_1\_ratio = \lambda_1/(\lambda_1+\lambda_2)$.
   - Преимущества: объединяет отбор признаков Lasso и устойчивость Ridge при коррелированных признаках.

5. Реализация и сравнение на Python
   - Сгенерировать данные: $M=50$ признаков, только 10 истинно влияют, $n=200$.
   - Применить `RidgeCV`, `LassoCV`, `ElasticNetCV`. Подобрать параметры кросс-валидацией.
   - Построить regularization path для Lasso (`lasso_path`): по оси X — $\log\lambda$, по оси Y — значения коэффициентов, каждая линия — один признак.
   - Сравнить на тестовой выборке $R^2$, MSE, число ненулевых коэффициентов. Вывести, какие именно признаки отобраны Lasso и Elastic Net, и сопоставить с истинным набором значимых признаков.
   - Сделать выводы о качестве отбора.

6. Практические рекомендации
   - Стандартизация признаков обязательна.
   - Когда Ridge, Lasso, Elastic Net.
   - О стандартных ошибках: в регуляризации они нестандартны; для оценки неопределённости использовать кросс-валидацию или бутстреп (кратко пояснить идею бутстрепа для Lasso).

Вопросы для самопроверки:
- Для начинающих (5): зачем штраф, что такое разреженность и где возникает, влияние $\lambda$ в Ridge, зачем стандартизация, преимущество Elastic Net.
- Для профессионалов (5): байесовский смысл L2/L1, почему Lasso даёт нули (геометрически или KKT), как работает координатный спуск, регуляризация Тихонова и Ridge, оценка неопределённости в Lasso.
```

### Теория, вероятностная модель и функция потерь

```text
Напиши первую часть главы «Логистическая регрессия» для учебного пособия, адресованного одновременно начинающим и профессионалам.
Цель книги — дать уникальную глубокую математическую базу по алгоритмам ML.

Стиль: основной текст интуитивный, с яркими аналогиями; для профессионалов — углублённые вставки «Углублённый взгляд» со строгими выводами и доказательствами.
Используй длинные тире (—) и кавычки-ёлочки («»). Обозначение дисперсии — $\mathrm{Var}$.

Начни с мостика: «В предыдущих главах мы предсказывали непрерывную величину — цену квартиры или стоимость автомобиля. Но часто ответ — это "да" или "нет". Вернёт ли заёмщик кредит? Является ли письмо спамом? Уйдёт ли клиент к конкуренту? Для таких задач используется логистическая регрессия».

Разделы строго по порядку:

1. Постановка задачи бинарной классификации
   - Отклик $y \in \{0, 1\}$, пример с кредитом.
   - Почему линейная регрессия не подходит (выход за [0,1], гетероскедастичность).
   - Идея: моделировать вероятность $P(y=1 \mid \mathbf{x})$.
   - Сигмоида $\sigma(z) = \frac{1}{1+e^{-z}}$, её график, свойства. Подробно опиши: область значений (0,1), монотонность, значение в нуле, симметричность $\sigma(-z) = 1 - \sigma(z)$. Обязательно выведи и поясни свойство производной $\sigma'(z) = \sigma(z)(1-\sigma(z))$.
   - **Углублённый взгляд: строгий вывод связи логита и сигмоиды.**  
     Приведи доказательство: если $p = \sigma(z)$, то $z = \ln\frac{p}{1-p} = \text{logit}(p)$, и наоборот. Покажи, что сигмоида и логит — взаимно обратные функции. Приведи таблицу свойств обеих функций: область определения, область значений, симметричность, монотонность.
   - Модель: $P(y=1 \mid \mathbf{x}) = \sigma(\mathbf{w}^T \boldsymbol{\phi}(\mathbf{x}))$, логит (log-odds), матрица плана $\Phi$.

2. Отношение шансов и интерпретация коэффициентов
   - Шансы $\frac{p}{1-p}$, логарифм шансов $\log\frac{p}{1-p} = \mathbf{w}^T \boldsymbol{\phi}(\mathbf{x})$.
   - Интерпретация $w_j$: увеличение $\phi_j$ на 1 умножает шансы на $e^{w_j}$.
   - Приближение для малых $w_j$, важность интерпретируемости в медицине, кредитном скоринге.

3. Метод максимального правдоподобия
   - Наблюдения — испытания Бернулли: $y_i \sim \text{Bernoulli}(p_i)$.
   - Функция правдоподобия $L(\mathbf{w}) = \prod p_i^{y_i} (1-p_i)^{1-y_i}$.
   - Логарифм правдоподобия, кросс-энтропийная функция потерь $E(\mathbf{w}) = -\sum [y_i\log p_i + (1-y_i)\log(1-p_i)]$.
   - Вывод: нет аналитического решения, нужна итеративная оптимизация.

   - **Углублённый взгляд: строгое доказательство выпуклости кросс-энтропии.**  
     Гессиан $H = \Phi^T D \Phi$, $D = \text{diag}(\sigma_i(1-\sigma_i))$. Поскольку $0<\sigma_i<1$, $D$ положительно определена, $H$ положительно полуопределена; при полном ранге $\Phi$ — строгая выпуклость.

   - **Углублённый взгляд: численный пример вычисления кросс-энтропии.**  
     Приведи следующий пример. Даны 6 объектов с признаками $x_1, x_2, x_3$ (бинарные), истинными метками $y$ и начальными весами $\mathbf{w}_0 = (0.1, -0.1, 0.2), b_0=0$. Вычисли вручную: для каждого $i$ линейную комбинацию $z^{(i)}$, предсказанную вероятность $\hat{y}^{(i)}=\sigma(z^{(i)})$, индивидуальные потери $-\bigl[y^{(i)}\log\hat{y}^{(i)} + (1-y^{(i)})\log(1-\hat{y}^{(i)})\bigr]$ и общую кросс-энтропию. Результат представь в виде таблицы и покажи, что суммарная потеря $J = 4.374$. Это даст читателю осязаемое понимание функции потерь. Никакого кода, только ручной расчёт.

4. Градиент функции потерь
   - Подробный покомпонентный вывод: $\frac{\partial E}{\partial w_k} = \sum_i (\sigma(\mathbf{w}^T \boldsymbol{\phi}_i) - y_i) \phi_k(\mathbf{x}_i)$.
   - Матричная форма: $\nabla_{\mathbf{w}} E = \Phi^T (\boldsymbol{\sigma} - \mathbf{y})$.
   - Сравнение с линейной регрессией: ошибки $\boldsymbol{\sigma} - \mathbf{y}$ нелинейны по $\mathbf{w}$.

   - **Углублённый взгляд: проблема полной разделимости.**  
     При линейной разделимости норма $\mathbf{w}$ неограниченно растёт, $E(\mathbf{w}) \to 0$, решения без регуляризации нет.

В конце — вопросы для самопроверки: 6 для начинающих, 5 для профессионалов (сигмоида, логиты, кросс-энтропия, градиент, разделимость).
Никакого кода.
```

```text
Напиши третью часть главы «Логистическая регрессия», продолжая после метрик и выбора порога.
Стиль: строгий, с интуитивными пояснениями и углублёнными вставками. Код на Python только для калибровки, Пуассоновской регрессии и взвешенной логистической регрессии. Библиотеки уже импортированы ранее, не повторять.

Разделы:

1. Асимптотические свойства оценки максимального правдоподобия
   - Информационная матрица Фишера: $I(\mathbf{w}) = -E[\partial^2 \ell / \partial \mathbf{w} \partial \mathbf{w}^T] = \Phi^T D \Phi = H$ (совпадение благодаря канонической link-функции).
   - Состоятельность, асимптотическая нормальность: $\sqrt{n}(\hat{\mathbf{w}} - \mathbf{w}^*) \xrightarrow{d} \mathcal{N}(0, I_1(\mathbf{w}^*)^{-1})$.
   - Эффективность (достижение границы Крамера–Рао).

2. Тест отношения правдоподобия (LRT)
   - Вложенные модели, статистика $\lambda = 2(\ell(\hat{\mathbf{w}}_1) - \ell(\hat{\mathbf{w}}_0)) \xrightarrow{d} \chi^2_{\text{df}}$.
   - Пример использования для проверки значимости группы признаков.

3. Логистическая регрессия как обобщённая линейная модель (GLM)
   - Три компонента GLM: случайная (биномиальная), систематическая ($\eta = \mathbf{w}^T \boldsymbol{\phi}$), link-функция (logit).
   - Каноническая ссылка и её преимущества.
   - **Расширение: Пуассоновская регрессия.**  
     Для счётных данных, $y \sim \text{Poisson}(\lambda)$, link $\log(\lambda) = \eta$. Краткий пример кода: сгенерировать данные (число ДТП от погоды и трафика), обучить `PoissonRegressor` из sklearn или `GLM` из statsmodels, показать коэффициенты и интерпретацию.

4. Вероятностная калибровка
   - Brier score, калибровочная кривая (reliability diagram).
   - Методы: Platt scaling, изотоническая регрессия.
   - Код: на примере breast_cancer после L1-регуляризации показать калибровку с `CalibratedClassifierCV`, построить калибровочные кривые до и после.

5. Взвешенная логистическая регрессия для несбалансированных классов
   - Взвешенная кросс-энтропия: $E_{\text{w}} = -\sum [c_1 y_i \log p_i + c_0 (1-y_i) \log(1-p_i)]$.
   - Модификация градиента: $\nabla E_{\text{w}} = \Phi^T \mathbf{v}$, где $v_i = c_1$ для $y_i=1$, иначе $c_0$; выведи итоговую формулу подробно.
   - Пример с `LogisticRegression(class_weight='balanced')`, сравнение метрик.

В этой части вопросы для самопроверки не задаются — они будут в следующей.
```

### Линейный SVM: геометрия, максимальный зазор и двойственная задача

```text
Напиши первую часть главы «Метод опорных векторов (SVM)» для учебного пособия, которое даёт глубокую математическую базу по машинному обучению.  
Стиль: основной текст — строгий и ясный, с геометрической интуицией; для профессионалов — вставки «Углублённый взгляд» с доказательствами и дополнительными деталями.  
Используй длинные тире (—) и кавычки‑ёлочки («»). Обозначение нормы — $\|\cdot\|$, дисперсии — $\mathrm{Var}$.  
Начни с мостика: «Логистическая регрессия строит вероятностную разделяющую поверхность, но не предъявляет жёстких требований к ширине разделяющей полосы. Метод опорных векторов, напротив, ищет гиперплоскость, которая не просто разделяет классы, а отстоит от ближайших точек обоих классов на максимально возможное расстояние.»

**Разделы первой части (строго по порядку, без кода):**

1. **Линейная разделимость и зазор**
   - Бинарная классификация, метки $y_i \in \{-1, +1\}$, обоснование выбора $\pm 1$.
   - Гиперплоскость $f(\mathbf{x}) = \mathbf{w}^T \mathbf{x} + b$, решающее правило $\mathrm{sign}(f(\mathbf{x}))$.
   - Функциональный зазор $\gamma_i = y_i(\mathbf{w}^T\mathbf{x}_i+b)$ и его недостаток (зависимость от масштаба).
   - Геометрический зазор $\frac{\gamma_i}{\|\mathbf{w}\|}$, вывод расстояния от точки до гиперплоскости.
   - **Определение:** зазор классификации $\rho = \min_i \frac{y_i(\mathbf{w}^T\mathbf{x}_i+b)}{\|\mathbf{w}\|}$.
   - Идея максимизации минимального зазора — поиск «наиболее устойчивой» разделяющей гиперплоскости.

2. **Прямая задача hard‑margin SVM**
   - Нормировка: условие $y_i(\mathbf{w}^T\mathbf{x}_i+b) \ge 1$ без потери общности (объяснить, почему это не ограничивает класс гиперплоскостей).
   - Тогда геометрический зазор становится $\frac{1}{\|\mathbf{w}\|}$, а максимизация зазора эквивалентна минимизации $\frac{1}{2}\|\mathbf{w}\|^2$.
   - Прямая задача оптимизации:
     $$ \min_{\mathbf{w},b} \frac{1}{2}\|\mathbf{w}\|^2 $$
     $$ \text{s.t. } y_i(\mathbf{w}^T\mathbf{x}_i+b) \ge 1, \quad i=1,\dots,n. $$
   - Поясни, что это задача выпуклого квадратичного программирования с линейными ограничениями.

   - **Углублённый взгляд: единственность решения.**  
     Покажи строгую выпуклость целевой функции и замкнутость допустимого множества; сделай вывод, что решение $\mathbf{w}^*$ единственно, а $b^*$ может быть неединственным в вырожденных случаях (опорные точки на одинаковом расстоянии).

3. **Лагранжиан и переход к двойственной задаче**
   - Вводим множители Лагранжа $\alpha_i \ge 0$, функция Лагранжа:
     $$ L(\mathbf{w}, b, \boldsymbol{\alpha}) = \frac{1}{2}\|\mathbf{w}\|^2 - \sum_{i=1}^n \alpha_i \bigl(y_i(\mathbf{w}^T\mathbf{x}_i+b) - 1\bigr). $$
   - Двойственная функция $g(\boldsymbol{\alpha}) = \inf_{\mathbf{w},b} L(\mathbf{w},b,\boldsymbol{\alpha})$.
   - Находим минимум по $\mathbf{w}, b$:
     $$ \frac{\partial L}{\partial \mathbf{w}} = 0 \;\Rightarrow\; \mathbf{w} = \sum_{i=1}^n \alpha_i y_i \mathbf{x}_i, $$
     $$ \frac{\partial L}{\partial b} = 0 \;\Rightarrow\; \sum_{i=1}^n \alpha_i y_i = 0. $$
   - Подставляем эти выражения обратно в $L$, проводим все алгебраические упрощения до получения
     $$ g(\boldsymbol{\alpha}) = \sum_{i=1}^n \alpha_i - \frac{1}{2}\sum_{i,j} \alpha_i \alpha_j y_i y_j (\mathbf{x}_i^T \mathbf{x}_j). $$
   - Формулируем двойственную задачу:
     $$ \max_{\boldsymbol{\alpha}} \sum_{i=1}^n \alpha_i - \frac{1}{2}\sum_{i=1}^n\sum_{j=1}^n \alpha_i \alpha_j y_i y_j (\mathbf{x}_i^T \mathbf{x}_j) $$
     $$ \text{s.t. } \alpha_i \ge 0, \quad \sum_{i=1}^n \alpha_i y_i = 0. $$
   - Отмечаем, что задача зависит только от скалярных произведений $\mathbf{x}_i^T\mathbf{x}_j$ — это ключ к ядровому трюку.

4. **Условия Каруша–Куна–Таккера (KKT) и опорные векторы**
   - Выписываем полную систему KKT для прямой задачи:
     1. Стационарность: $\mathbf{w} = \sum \alpha_i y_i \mathbf{x}_i$, $\sum \alpha_i y_i = 0$.
     2. Прямая допустимость: $y_i(\mathbf{w}^T\mathbf{x}_i+b) \ge 1$.
     3. Двойственная допустимость: $\alpha_i \ge 0$.
     4. Условие дополняющей нежёсткости: $\alpha_i \bigl(y_i(\mathbf{w}^T\mathbf{x}_i+b) - 1\bigr) = 0$.
   - Анализируем условие доп. нежёсткости:
     * Если $\alpha_i > 0$, то $y_i(\mathbf{w}^T\mathbf{x}_i+b) = 1$ — точка лежит **на границе зазора** (опорный вектор).
     * Если $\alpha_i = 0$, то точка может находиться дальше границы и не влияет на решение.
   - Делаем вывод о **разреженности**: решение $\mathbf{w}$ выражается только через опорные векторы.

   - **Углублённый взгляд: сильная двойственность и квалификация Слейтера.**  
     Докажи, что для SVM выполняется условие Слейтера (существует точка, удовлетворяющая строгим неравенствам), откуда следует совпадение оптимальных значений прямой и двойственной задач (нулевой зазор двойственности).

5. **Восстановление решающего правила**
   - Смещение $b$ вычисляется из любого опорного вектора с $\alpha_s > 0$:
     $$ b = y_s - \sum_{m=1}^n \alpha_m y_m (\mathbf{x}_m^T \mathbf{x}_s). $$
     Для численной устойчивости берут усреднение по всем таким векторам.
   - Итоговое решающее правило:
     $$ f(\mathbf{x}) = \sum_{i=1}^n \alpha_i y_i (\mathbf{x}_i^T \mathbf{x}) + b, \quad \hat{y} = \mathrm{sign}(f(\mathbf{x})). $$

**В конце первой части приведи вопросы для самопроверки:**
- Для начинающих (6): определение геометрического зазора, почему максимизируем $1/\|\mathbf{w}\|$, вид прямой задачи, что дают множители Лагранжа, что такое опорные векторы, как вычислить $b$.
- Для профессионалов (5): доказательство единственности решения $\mathbf{w}$, полный вывод двойственной задачи, анализ KKT и разреженности, проверка условия Слейтера, связь двойственной задачи с квадратичным программированием.

**Важно:** никакого кода, только математика и геометрические пояснения.
```

## ПРОМПТ 2  
### Soft‑margin SVM: slack‑переменные, регуляризация и статистическая интерпретация

```text
Продолжи главу «Метод опорных векторов (SVM)» сразу после раздела о hard‑margin SVM и KKT.
Стиль: основной текст — строгий, с ясной интуицией; углублённые вставки «Углублённый взгляд» — с доказательствами и дополнительными деталями.
Используй длинные тире, кавычки‑ёлочки. Кода по‑прежнему нет.

**Разделы строго по порядку:**

1. **Потребность в soft‑margin**
   - Реальные данные почти никогда не являются линейно разделимыми без ошибок. Шум, выбросы, слабая разделимость.
   - Идея: разрешить нескольким точкам нарушать границу зазора, но штрафовать за это.
   - Введение slack‑переменных $\xi_i \ge 0$, ослабляющих ограничения:
     $$ y_i(\mathbf{w}^T\mathbf{x}_i+b) \ge 1 - \xi_i, \quad i=1,\dots,n. $$
     Если $0 < \xi_i \le 1$, точка лежит внутри зазора, но классифицируется правильно; если $\xi_i > 1$, точка классифицируется неверно.

2. **Прямая задача soft‑margin SVM**
   - Штраф за нарушения: добавляем в целевую функцию сумму $\sum \xi_i$ с весовым коэффициентом $C > 0$:
     $$ \min_{\mathbf{w},b,\boldsymbol{\xi}} \frac{1}{2}\|\mathbf{w}\|^2 + C \sum_{i=1}^n \xi_i $$
     $$ \text{s.t. } y_i(\mathbf{w}^T\mathbf{x}_i+b) \ge 1 - \xi_i, \quad \xi_i \ge 0, \quad i=1,\dots,n. $$
   - Параметр $C$ управляет компромиссом между максимизацией зазора и минимизацией суммарной ошибки:
     * $C \to 0$ — широкий зазор, много нарушений (недообучение).
     * $C \to \infty$ — стремится к hard‑margin (переобучение, если есть выбросы).

   - **Углублённый взгляд: связь с регуляризацией и структурным риском.**  
     Покажи, что задача эквивалентна минимизации $\sum \max(0, 1 - y_i f(\mathbf{x}_i)) + \lambda \|\mathbf{w}\|^2$ (hinge loss + $L_2$‑регуляризация), где $\lambda = \frac{1}{2C}$. Поясни, что это реализация принципа структурной минимизации риска (SRM): выбираем модель с малой VC‑размерностью и малым эмпирическим риском.

3. **Двойственная задача soft‑margin**
   - Строим лагранжиан с двумя наборами множителей: $\alpha_i \ge 0$ (для основных ограничений) и $\mu_i \ge 0$ (для $\xi_i \ge 0$):
     $$ L = \frac{1}{2}\|\mathbf{w}\|^2 + C\sum \xi_i - \sum \alpha_i \bigl(y_i(\mathbf{w}^T\mathbf{x}_i+b) - 1 + \xi_i\bigr) - \sum \mu_i \xi_i. $$
   - Дифференцируем по $\mathbf{w}, b, \xi_i$:
     $$ \frac{\partial L}{\partial \mathbf{w}} = 0 \Rightarrow \mathbf{w} = \sum \alpha_i y_i \mathbf{x}_i, $$
     $$ \frac{\partial L}{\partial b} = 0 \Rightarrow \sum \alpha_i y_i = 0, $$
     $$ \frac{\partial L}{\partial \xi_i} = 0 \Rightarrow C - \alpha_i - \mu_i = 0. $$
   - Из последнего условия и $\mu_i \ge 0$ следует $\alpha_i \le C$. Подстановка даёт ту же двойственную целевую функцию, что и в hard‑margin, но с новыми ограничениями:
     $$ \max_{\boldsymbol{\alpha}} \sum_{i=1}^n \alpha_i - \frac{1}{2}\sum_{i,j} \alpha_i \alpha_j y_i y_j (\mathbf{x}_i^T \mathbf{x}_j) $$
     $$ \text{s.t. } 0 \le \alpha_i \le C, \quad \sum \alpha_i y_i = 0. $$
   - **Интерпретация значений $\alpha_i$:**
     * $\alpha_i = 0$ — точка правильно классифицирована и находится далеко от границы.
     * $0 < \alpha_i < C$ — точка лежит **точно на границе зазора** (опорный вектор).
     * $\alpha_i = C$ — точка **внутри зазора** или неверно классифицирована (slack-переменная активна).

4. **Условия KKT для soft‑margin**
   - Выпиши полные условия KKT, акцентируя дополнительные условия для $\xi_i$ и $\mu_i$. Покажи, как из них следует классификация точек по величине $\alpha_i$.
   - **Углублённый взгляд: смещение $b$ в soft‑margin.**  
     Объясни, как вычислить $b$ устойчивым образом: берут среднее по всем точкам с $0 < \alpha_i < C$ (свободные опорные векторы):
       $$ b = \frac{1}{n_{\text{free}}} \sum_{i: 0<\alpha_i<C} \left( y_i - \sum_{j} \alpha_j y_j (\mathbf{x}_j^T \mathbf{x}_i) \right). $$

**Вопросы для самопроверки:**
- Для начинающих (5): зачем нужны slack‑переменные, что означает $\xi_i > 1$, роль параметра $C$, отличие ограничений в двойственной задаче, где находятся точки с $\alpha_i=C$.
- Для профессионалов (5): вывод двойственной задачи soft‑margin, доказательство ограничения $\alpha_i \le C$, связь с hinge loss, выбор $b$ по свободным опорным векторам, почему soft‑margin SVM соответствует SRM.
```

## ПРОМПТ 3  
### Ядровой трюк, теорема Мерсера и нелинейный SVM

```text
Продолжи главу «Метод опорных векторов (SVM)» сразу после раздела о soft‑margin.
Стиль: строгий научный текст с интуитивными пояснениями; «Углублённый взгляд» для формальных доказательств.
Кода по‑прежнему нет.

**Разделы:**

1. **Мотивация: от линейного к нелинейному разделению**
   - Пример данных, линейно неразделимых в исходном пространстве, но разделимых после нелинейного отображения.
   - Идея: отобразить данные в спрямляющее пространство $\mathcal{H}$ с помощью функции $\phi: \mathbb{R}^d \to \mathcal{H}$ и применить линейный SVM в $\mathcal{H}$.

2. **Ядровой трюк**
   - Двойственная задача и решающее правило содержат только скалярные произведения $\phi(\mathbf{x}_i)^T \phi(\mathbf{x}_j)$. Если мы можем вычислить их напрямую через функцию ядра $K(\mathbf{x}_i, \mathbf{x}_j) = \langle \phi(\mathbf{x}_i), \phi(\mathbf{x}_j) \rangle$, то нам не нужно явно строить $\phi$ или работать в $\mathcal{H}$.
   - Двойственная задача с ядром:
     $$ \max_{\boldsymbol{\alpha}} \sum \alpha_i - \frac{1}{2}\sum_{i,j} \alpha_i \alpha_j y_i y_j K(\mathbf{x}_i, \mathbf{x}_j) $$
     $$ \text{s.t. } 0 \le \alpha_i \le C, \quad \sum \alpha_i y_i = 0. $$
   - Решающее правило: $f(\mathbf{x}) = \sum \alpha_i y_i K(\mathbf{x}_i, \mathbf{x}) + b$.

3. **Теорема Мерсера и свойства ядер**
   - **Определение:** функция $K: \mathcal{X}\times\mathcal{X} \to \mathbb{R}$ называется положительно полуопределённым ядром, если для любого конечного набора точек $\mathbf{x}_1,\dots,\mathbf{x}_m$ матрица $K_{ij} = K(\mathbf{x}_i,\mathbf{x}_j)$ симметрична и положительно полуопределена.
   - **Теорема Мерсера (формулировка без доказательства):** если $K$ — симметричная непрерывная положительно полуопределённая функция на компактной области, то существует гильбертово пространство $\mathcal{H}$ и отображение $\phi$ такие, что $K(\mathbf{x},\mathbf{x}') = \langle \phi(\mathbf{x}), \phi(\mathbf{x}') \rangle_{\mathcal{H}}$.
   - Это гарантирует корректность использования ядра в SVM.

4. **Стандартные ядра и их геометрический смысл**
   - Линейное ядро: $K(\mathbf{x},\mathbf{x}') = \mathbf{x}^T\mathbf{x}'$.
   - Полиномиальное степени $d$: $K(\mathbf{x},\mathbf{x}') = (\gamma \mathbf{x}^T\mathbf{x}' + r)^d$. Обсуди, что признаки в $\mathcal{H}$ — все мономы степени не выше $d$.
   - Радиально-базисная функция (RBF): $K(\mathbf{x},\mathbf{x}') = \exp(-\gamma \|\mathbf{x} - \mathbf{x}'\|^2)$. Покажи интуитивно, что это ядро соответствует бесконечномерному пространству, а $\gamma$ контролирует «радиус влияния» точки.
   - Сигмоидальное ядро: $K(\mathbf{x},\mathbf{x}') = \tanh(\gamma \mathbf{x}^T\mathbf{x}' + r)$; не всегда удовлетворяет условию Мерсера, но используется на практике.

   - **Углублённый взгляд: построение спрямляющего пространства для полиномиального ядра.**  
     Выведи явный вид отображения $\phi$ для $d=2$, $\gamma=1, r=0$: $\phi(\mathbf{x}) = (x_1^2, \sqrt{2}x_1 x_2, x_2^2, \dots)$ и покажи, что $\langle \phi(\mathbf{x}), \phi(\mathbf{x}') \rangle = (\mathbf{x}^T\mathbf{x}')^2$.

5. **Выбор ядра и его параметров**
   - Линейное ядро — частный случай, когда данные приблизительно линейно разделимы.
   - RBF — «универсальное» ядро, способное моделировать сложные границы. Роль $\gamma$: при малом $\gamma$ — гладкая граница, при большом — склонность к переобучению (каждая точка становится «островком»).
   - Полиномиальное — когда есть основания полагать полиномиальную зависимость между признаками.
   - Понятие о кросс‑валидации для выбора $C$ и $\gamma$ (детали в практической части позже).

**Вопросы для самопроверки:**
- Для начинающих (5): что такое ядро, зачем нужен ядровой трюк, примеры ядер, как $\gamma$ в RBF влияет на границу, что даёт теорема Мерсера на практике.
- Для профессионалов (5): формальное определение положительно полуопределённого ядра, доказательство, что RBF-ядро соответствует бесконечномерному пространству, построение $\phi$ для полиномиального ядра, условия Мерсера для сигмоидального ядра, компромисс bias‑variance при выборе $\gamma$.
```

## ПРОМТ 4  
### Алгоритм SMO, реализация SVM с нуля и сравнение с библиотечной версией

```text
Продолжи главу «Метод опорных векторов (SVM)» после ядрового трюка. Теперь вводится код на Python.
Стиль: научный, с подробными выкладками; код должен быть рабочим и воспроизводимым.
**Единый блок импортов** в начале этого фрагмента: `numpy`, `matplotlib.pyplot`, `sklearn.svm`, `sklearn.datasets`, `sklearn.model_selection`, `time`. Далее импорты не повторять.

**Разделы:**

1. **Проблема масштабируемости двойственной задачи**
   - Матрица ядра требует $O(n^2)$ памяти; стандартные QP‑солверы неприменимы при больших $n$.
   - Идея декомпозиции: на каждом шаге оптимизируется небольшая рабочая группа переменных.
   - Алгоритм SMO (Sequential Minimal Optimization) использует группу из ровно двух множителей, так как ограничение $\sum \alpha_i y_i = 0$ позволяет выразить один через другой и решать аналитически.

2. **Аналитическое решение для двух множителей**
   - Пусть на текущем шаге выбраны индексы $i$ и $j$. Зафиксируем все $\alpha_k$, $k\neq i,j$. Ограничение $\sum \alpha_k y_k = 0$ даёт $\alpha_i y_i + \alpha_j y_j = \text{const}$.
   - Выражаем $\alpha_i$ через $\alpha_j$ и подставляем в целевую функцию, получая квадратичную функцию от одной переменной $\alpha_j$.
   - **Вывод необрезанного обновления** (подробно):
     $$ \alpha_j^{\text{new, unc}} = \alpha_j^{\text{old}} + \frac{y_j (E_i - E_j)}{\eta}, $$
     где $E_i = f(\mathbf{x}_i) - y_i$ — ошибка на $i$-м объекте, $\eta = K_{ii} + K_{jj} - 2K_{ij}$.
   - **Клиппинг (clipping)** полученного значения в границы $[L, H]$, зависящие от $y_i, y_j$ и $C$:
       - если $y_i \ne y_j$: $L = \max(0, \alpha_j^{\text{old}} - \alpha_i^{\text{old}})$, $H = \min(C, C + \alpha_j^{\text{old}} - \alpha_i^{\text{old}})$.
       - если $y_i = y_j$: $L = \max(0, \alpha_i^{\text{old}} + \alpha_j^{\text{old}} - C)$, $H = \min(C, \alpha_i^{\text{old}} + \alpha_j^{\text{old}})$.
     Получаем $\alpha_j^{\text{new}} = \text{clip}(\alpha_j^{\text{new, unc}}, L, H)$.
   - $\alpha_i^{\text{new}} = \alpha_i^{\text{old}} + y_i y_j (\alpha_j^{\text{old}} - \alpha_j^{\text{new}})$.

3. **Обновление смещения $b$**
   - Если $0 < \alpha_i^{\text{new}} < C$, то из KKT: $b = b_i$, где $b_i = -E_i - y_i K_{ii} (\alpha_i^{\text{new}} - \alpha_i^{\text{old}}) - y_j K_{ji} (\alpha_j^{\text{new}} - \alpha_j^{\text{old}}) + b^{\text{old}}$.
     Аналогично для $j$. Усредняем, если оба свободны; иначе используем свободный.

4. **Реализация класса `SimpleSVM` на Python**
   - Конструктор: параметры `C=1.0`, `kernel='rbf'`, `gamma=1.0`, `tol=1e-3`, `max_iter=1000`.
   - Метод `_kernel(X1, X2)`: возвращает матрицу ядра (для `'linear'` — скалярное произведение, для `'rbf'` — экспоненту).
   - Метод `_take_step(i, j)`: реализует логику SMO для пары, описанную выше. Возвращает `True`, если шаг привёл к изменению $\alpha$.
   - Метод `fit(X, y)`: внешний цикл — итерации по всем точкам, поиск нарушителей KKT, выбор второго множителя (эвристика: максимизация $|E_i - E_j|$), вызов `_take_step`. Кеширование ошибок $E$. Критерий останова — число итераций или отсутствие изменений.
   - Метод `decision_function(X)`: вычисляет $f(\mathbf{x})$ без знака.
   - Метод `predict(X)`: возвращает $\mathrm{sign}(f(\mathbf{x}))$.

5. **Верификация реализации**
   - Сгенерировать линейно разделимые 2D-данные, обучить свой SVM с линейным ядром, построить разделяющую прямую и опорные векторы. Сравнить коэффициенты и опорные векторы с `sklearn.svm.SVC(kernel='linear', C=1e10)`.
   - Загрузить breast_cancer, стандартизировать, обучить свой SVM и sklearn SVC (с RBF, одинаковые $C, \gamma$). Сравнить accuracy и число опорных векторов.
   - Сравнить время обучения на выборках разного размера (200, 500, 1000) — построить график.

**Вопросы для самопроверки:**
- Для начинающих (4): почему в SMO выбирают два множителя, что такое кеширование ошибок, как `predict` использует опорные векторы, какие точки становятся опорными.
- Для профессионалов (5): вывод формулы $\alpha_j^{\text{new, unc}}$, обоснование границ $L, H$, как выбирать второй множитель для эффективности, как обновление $b$ зависит от типа опорного вектора, сложность одного шага SMO.
```

### Многоклассовый SVM, SVR, теория обобщения, сравнение с логистической регрессией и заключение

```text
Напиши заключительную часть главы «Метод опорных векторов (SVM)», продолжая после реализации SMO.
Стиль: строгий научный, с интуитивными пояснениями и углублёнными вставками.
Код на Python (sklearn) для демонстрации многоклассовости, SVR и сравнения; импорты (numpy, matplotlib, sklearn) уже были в предыдущей части, не повторять.

**Разделы:**

1. **Многоклассовая классификация SVM**
   - Стратегии сведения многоклассовой задачи к бинарным: One‑vs‑Rest (OvR) и One‑vs‑One (OvO).
     * OvR: обучаем $K$ классификаторов, объекты класса $k$ против всех остальных. Предсказание: класс с максимальным отступом $f_k(\mathbf{x})$.
     * OvO: обучаем $K(K-1)/2$ классификаторов на всех парах классов, голосование большинством.
   - Преимущества и недостатки: OvR склонно к несбалансированности, OvO требует больше моделей, но обучается на меньших подвыборках.
   - Кратко о методе Краммера–Зингера (единая целевая функция, аналог softmax для SVM) — не вдаваясь в детали.
   - Код на `load_digits`: сравнение accuracy, macro F1 и времени обучения `SVC(decision_function_shape='ovo')` и `'ovr'`.

2. **SVM для регрессии (SVR)**
   - Мотивация: предсказание непрерывной величины с сохранением идеи зазора.
   - $\epsilon$-insensitive loss: $L_\epsilon(y, f(\mathbf{x})) = \max(0, |y - f(\mathbf{x})| - \epsilon)$. Ошибки внутри трубки шириной $2\epsilon$ не штрафуются.
   - Прямая задача SVR:
     $$ \min_{\mathbf{w},b} \frac{1}{2}\|\mathbf{w}\|^2 + C \sum_{i=1}^n (\xi_i + \xi_i^*) $$
     $$ \text{s.t. } y_i - \mathbf{w}^T\mathbf{x}_i - b \le \epsilon + \xi_i, \quad \mathbf{w}^T\mathbf{x}_i + b - y_i \le \epsilon + \xi_i^*, \quad \xi_i,\xi_i^* \ge 0. $$
   - Кратко опиши двойственную задачу (аналогично классификации, но появляются два набора множителей $\alpha_i, \alpha_i^*$).
   - Код: пример с `make_regression`, обучить `SVR(kernel='rbf')`, показать график с трубкой и опорными векторами.

3. **Теоретические основы: VC‑размерность и структурная минимизация риска**
   - Напомним, что SVM минимизирует сумму эмпирического риска (hinge loss) и регуляризатора.
   - **Определение VC‑размерности** класса функций.
   - Теорема: с вероятностью $1-\delta$ риск $R(f) \le R_{\text{emp}}(f) + \sqrt{\frac{h(\log(2n/h)+1) - \log(\delta/4)}{n}}$, где $h$ — VC‑размерность.
   - Для линейных классификаторов с зазором $\rho$ VC‑размерность ограничена $\min\left(D, \left\lceil \frac{R^2}{\rho^2} \right\rceil\right) + 1$, где $R$ — радиус минимальной охватывающей сферы данных.
   - Следовательно, максимизация зазора $\rho$ (минимизация $\|\mathbf{w}\|$) напрямую уменьшает VC‑размерность и улучшает обобщающую способность. Это и есть принцип структурной минимизации риска (SRM).

   - **Углублённый взгляд: доказательство оценки VC‑размерности через зазор.**  
     Набросок: покажи, что если данные лежат в шаре радиуса $R$ и разделяются с зазором $\rho$, то количество способов разбить $n$ точек не превосходит $\left(\frac{e n R}{\rho}\right)^{O(1/\rho^2)}$, откуда следует ограничение на VC‑размерность.

4. **Сравнение SVM и логистической регрессии**
   - Функции потерь: hinge loss $ \max(0, 1 - y f) $ vs log loss $\log(1 + \exp(-y f))$. Графики и обсуждение: hinge loss равен нулю для $y f \ge 1$, log loss всегда положителен и более чувствителен к выбросам.
   - Интерпретируемость: LR даёт калиброванные вероятности, SVM — только отступ. Для вероятностей в SVM применяют Platt scaling (логистическая регрессия на выходах SVM).
   - Вычислительная эффективность: SVM зависит от числа опорных векторов; LR обучается быстрее на больших данных.
   - Качество: SVM может выигрывать при чёткой разделимости и правильном ядре; LR — при большом количестве признаков и для разреженных данных.
   - **Итоговая сравнительная таблица** по 8 критериям.

5. **Мост к нейронным сетям и kernel methods**
   - SVM с ядром RBF можно рассматривать как двухслойную сеть: скрытый слой — радиальные базисные функции с центрами в опорных векторах, выходной слой — линейная комбинация.
   - Недостаток: число опорных векторов растёт с размером данных, предсказание медленное.
   - Глубокие нейронные сети учат представление признаков автоматически, не требуя ручного выбора ядра.

6. **Заключение и резюме главы**
   - Обобщить ключевые концепции: зазор, опорные векторы, двойственность, ядра, SMO, SRM.
   - **Итоговая таблица «SVM vs Logistic Regression»** и **«Hard‑margin vs Soft‑margin vs SVR»**.

7. **Вопросы для самопроверки (объединяют всю главу):**
   - Для начинающих (6): что максимизирует SVM, роль $C$ и $\gamma$, зачем ядра, отличие OvO от OvR, чем SVR отличается от линейной регрессии, почему SVM не даёт вероятностей напрямую.
   - Для профессионалов (5): как максимизация зазора связана с VC‑размерностью, доказательство оценки VC‑размерности через зазор, вывод двойственной задачи SVR, платтовская калибровка, сравнение hinge loss и log loss с точки зрения статистической устойчивости.

8. **Задания повышенной сложности**
   1. Вывести двойственную задачу SVR и проанализировать условия KKT.
   2. Реализовать Platt scaling для своего SVM из предыдущей части и оценить калибровку.
   3. Провести эксперимент: построить learning curves для SVM и логистической регрессии, проанализировать компромисс смещения и дисперсии.
   4. Доказать, что для линейно разделимой выборки решение hard‑margin SVM единственно.
   5. Вывести выражение для зазора в терминах опорных векторов и показать, что он равен $2 / \|\mathbf{w}\|$.
```

### Bias‑variance tradeoff: теоретический фундамент

```text
Напиши первую часть главы «Кросс‑валидация, bias‑variance tradeoff и выбор гиперпараметров» для учебного пособия по машинному обучению, которое даёт глубокую математическую базу.

Стиль: основной текст — строгий и ясный, с интуитивными аналогиями; для профессионалов — вставки «Углублённый взгляд» с доказательствами и деталями. Используй длинные тире (—) и кавычки‑ёлочки («»). Обозначение математического ожидания — $\mathbb{E}$, дисперсии — $\mathrm{Var}$.

Начни с мостика: «В предыдущих главах мы строили модели — от линейной регрессии до SVM — и оценивали их качество на тестовой выборке. Но почему одна модель работает лучше другой? Почему добавление признаков иногда улучшает, а иногда ухудшает предсказания? Ответ кроется в фундаментальном разложении ошибки на смещение, дисперсию и неустранимый шум. Понимание этого компромисса — ключ к осознанному выбору сложности модели и стратегии валидации.»

**Разделы (строго по порядку, без кода):**

1. **Постановка задачи и ожидаемая ошибка**
   - Обучающая выборка $\mathcal{D} = \{(\mathbf{x}_i, y_i)\}_{i=1}^n$, истинная (неизвестная) зависимость $y = f(\mathbf{x}) + \varepsilon$, где $\varepsilon$ — случайный шум с $\mathbb{E}[\varepsilon]=0$ и $\mathrm{Var}(\varepsilon)=\sigma^2$.
   - Модель $\hat{f}(\mathbf{x}; \mathcal{D})$, обученная на конкретной выборке.
   - Ожидаемая ошибка на новой точке $\mathbf{x}_0$:
     $$ \text{Err}(\mathbf{x}_0) = \mathbb{E}_{\mathcal{D},\varepsilon} \bigl[ (y_0 - \hat{f}(\mathbf{x}_0; \mathcal{D}))^2 \bigr], $$
     где математическое ожидание берётся по всем возможным обучающим выборкам и реализациям шума.

2. **Разложение ожидаемой квадратичной ошибки**
   - Вывод разложения (пошагово, с промежуточными равенствами):
     $$ \text{Err}(\mathbf{x}_0) = \sigma^2 + \bigl( \mathbb{E}_{\mathcal{D}}[\hat{f}(\mathbf{x}_0)] - f(\mathbf{x}_0) \bigr)^2 + \mathbb{E}_{\mathcal{D}}\bigl[ (\hat{f}(\mathbf{x}_0) - \mathbb{E}_{\mathcal{D}}[\hat{f}(\mathbf{x}_0)])^2 \bigr]. $$
   - Три слагаемых:
     * $\sigma^2$ — **неустранимый шум** (irreducible error);
     * Квадрат смещения (bias²) — насколько в среднем предсказание модели отклоняется от истинной функции;
     * Дисперсия (variance) — насколько предсказания варьируются от выборки к выборке.
   - Интерпретация: смещение характеризует систематическую ошибку модели (недообучение), дисперсия — чувствительность к конкретному набору данных (переобучение).

   - **Углублённый взгляд: доказательство разложения.**  
     Приведи полный вывод, начиная с
     $$ (y_0 - \hat{f})^2 = (y_0 - f + f - \mathbb{E}[\hat{f}] + \mathbb{E}[\hat{f}] - \hat{f})^2, $$
     раскрывая квадрат и используя $\mathbb{E}[\varepsilon]=0$, независимость $\varepsilon$ и $\mathcal{D}$, а также $\mathbb{E}[\hat{f} - \mathbb{E}[\hat{f}]]=0$.

3. **Интуиция и геометрическая аналогия**
   - Аналогия со стрельбой по мишени: смещение — систематическое отклонение центра попаданий от цели; дисперсия — разброс попаданий. Четыре комбинации (низкое/высокое смещение и дисперсия).
   - Связь со сложностью модели:
     * Простая модель (например, линейная регрессия с малым числом признаков) — высокое смещение, низкая дисперсия.
     * Сложная модель (глубокое дерево, полином высокой степени) — низкое смещение, высокая дисперсия.
   - Компромисс: с ростом сложности смещение убывает, дисперсия растёт; существует оптимальная сложность, минимизирующая сумму.

4. **Bias‑variance для классификации**
   - Ошибка классификации (0‑1 loss) не допускает аддитивного разложения.
   - Тем не менее, феномены смещения и дисперсии проявляются и здесь: например, метод $k$ ближайших соседей: при $k=1$ — низкое смещение, высокая дисперсия; при большом $k$ — наоборот.
   - Краткое упоминание альтернативных разложений (например, для log‑loss).

   - **Углублённый взгляд: bias‑variance разложение для 0‑1 loss (по Domingos, 2000).**  
     Приведи основные идеи: ошибка раскладывается на смещение, дисперсию и ковариационный член; в отличие от MSE, здесь возможно отрицательное влияние дисперсии.

5. **Практические следствия**
   - Диагностика: высокая ошибка на обучении → высокое смещение (недообучение); низкая ошибка на обучении, но высокая на тесте → высокая дисперсия (переобучение).
   - Роль регуляризации: уменьшает дисперсию, но может увеличить смещение. Выбор $\lambda$ управляет компромиссом.
   - Роль объёма выборки: большие $n$ снижают дисперсию, поэтому сложные модели могут быть оправданы.

**В конце части приведи вопросы для самопроверки:**
- Для начинающих (5): что измеряет bias, что — variance, как выглядит переобучение с точки зрения bias‑variance, почему регуляризация уменьшает дисперсию, какую ошибку невозможно устранить.
- Для профессионалов (4): полный вывод разложения MSE, доказательство независимости $\varepsilon$ и $\mathbb{E}[\hat{f}]$, особенности разложения для классификации, почему с ростом $n$ дисперсия убывает.

**Важно:** никакого кода. Только теория и математические выкладки.
```

### Кросс‑валидация для временных рядов и зависимых данных

```text
Продолжи главу «Кросс‑валидация, bias‑variance tradeoff и выбор гиперпараметров» после раздела о k‑fold CV.
Стиль: строгий, с интуицией и «Углублённым взглядом». Код на Python, импорты уже были ранее, не повторять.
Особое внимание удели временным рядам и пространственно‑временным данным.

**Разделы:**

1. **Проблема зависимых данных**
   - Стандартное k‑fold предполагает независимость наблюдений. При нарушении (временные ряды, пространственная автокорреляция, группы) случайное перемешивание приводит к утечке информации из будущего в прошлое, и оценка ошибки становится оптимистичной.
   - Принцип: обучающие данные должны предшествовать тестовым во времени.

2. **Кросс‑валидация на временных рядах**
   - **Скользящий контроль (time series split):** последовательное расширение обучающей выборки. На каждом шаге обучаемся на данных до момента $t$, тестируем на следующем периоде $t+1, \dots, t+h$.
   - Формально: для $i$-го разбиения обучающая выборка — первые $i$ блоков, тестовая — $(i+1)$-й блок. Количество разбиений `n_splits = n_samples - train_size`.
   - Реализация `TimeSeriesSplit` из sklearn.
   - Обсуждение: оценка смещена, так как модели обучаются на разном объёме исторических данных; но это отражает реальный сценарий.
   - **Скользящий контроль с фиксированным окном:** обучающая выборка имеет постоянный размер (последние $w$ наблюдений), что больше подходит для стационарных рядов.

3. **Блоковая кросс‑валидация для зависимых данных**
   - Если данные не являются строго временным рядом, но сгруппированы (например, измерения по пациентам, датчикам), необходимо сохранять группы в одном фолде, чтобы избежать утечки.
   - **Group k‑fold:** все наблюдения одной группы попадают либо в обучение, либо в тест. `GroupKFold` в sklearn.
   - **Стратификация по группам:** `StratifiedGroupKFold`.

4. **Пространственно‑временные данные**
   - Пример: прогноз погоды по координатам. Обычная CV может разместить соседние точки в обучении и тесте, что даёт неоправданно высокую оценку из‑за пространственной автокорреляции.
   - Решение: **пространственная кросс‑валидация** — разбиение по блокам координат, удалённым друг от друга.

5. **Демонстрация на Python**
   - Сгенерировать синтетический временной ряд (например, AR(1) процесс) и обучить простую линейную регрессию с временным признаком.
   - Сравнить обычный 5‑fold CV и TimeSeriesSplit: показать, что 5‑fold даёт завышенную accuracy, а TimeSeriesSplit — более реалистичную.
   - Показать GroupKFold на искусственных данных с пациентами.
   - Визуализировать схему разбиений (train/test индексы) для каждого метода.

**Вопросы для самопроверки:**
- Для начинающих (4): почему обычный k‑fold нельзя применять к временным рядам, как работает TimeSeriesSplit, зачем нужны блоки в групповой CV, что такое утечка данных.
- Для профессионалов (4): выведите, что оценка ошибки при случайном k‑fold для AR(1) процесса смещена вниз; объясните, как выбрать длину окна в скользящем контроле; сравните дисперсию оценки при блоковой CV и обычной CV; как модифицировать CV для пространственных данных.
```

```text
Продолжи главу «Кросс‑валидация, bias‑variance tradeoff и выбор гиперпараметров» после разделов о методах CV.
Стиль: строгий, с интуитивными объяснениями и «Углублённым взглядом». Код на Python, импорты (numpy, matplotlib, sklearn, scipy, optuna) добавить в текущий блок; если optuna не установлен, можно использовать sklearn GaussianProcessRegressor для демонстрации байесовской оптимизации вручную.

**Разделы:**

1. **Постановка задачи выбора гиперпараметров**
   - Гиперпараметры: параметры, не обучаемые из данных (например, $\lambda$, $C$, $\gamma$), их настройка.
   - Цель: найти $\boldsymbol{\lambda}^* = \arg\min_{\boldsymbol{\lambda}} \text{Err}(\boldsymbol{\lambda})$, где Err — ошибка на валидации (кросс‑валидации).
   - Проблема: оценка Err дорогая (требует обучения модели). Нужны методы, минимизирующие число вычислений.

2. **Grid search (поиск по сетке)**
   - Полный перебор по заданной сетке значений. Гарантирует нахождение глобального минимума на сетке, но страдает от «проклятия размерности»: число точек растёт экспоненциально с числом гиперпараметров.
   - Рекомендации: начинать с широкой сетки, затем сужать вокруг лучших значений.
   - Код: `GridSearchCV` на примере SVM с RBF ядром (параметры $C$, $\gamma$). Построить heatmap точности.

3. **Random search (случайный поиск)**
   - Выбор $N$ случайных комбинаций из заданных распределений.
   - **Теорема (Bergstra & Bengio, 2012):** если только малое подмножество гиперпараметров действительно важно, случайный поиск с высокой вероятностью найдёт хорошую конфигурацию быстрее, чем grid search. Формально: пусть $p$ — доля пространства, где качество выше порога. Тогда вероятность того, что хотя бы одна из $N$ случайных точек попадёт в «хорошую» область, равна $1 - (1-p)^N$. Это не зависит от размерности, в отличие от grid search, где с ростом размерности плотность сетки падает.
   - **Углублённый взгляд: доказательство эффективности random search.**  
     Приведи простое доказательство вероятностной гарантии и объясни, почему grid search требует экспоненциального числа точек для сохранения плотности.
   - Код: `RandomizedSearchCV` с теми же параметрами, сравнение времени и качества с grid search.

4. **Байесовская оптимизация**
   - Идея: построить вероятностную модель (суррогатную модель) зависимости $\text{Err}(\boldsymbol{\lambda})$ — обычно гауссовский процесс. На каждой итерации выбирать точку, максимизирующую функцию приобретения (например, Expected Improvement), которая балансирует exploitation (где предсказание низкое) и exploration (где неопределённость высокая).
   - Основные шаги:
     1. Выбрать начальные точки (например, случайно), вычислить для них ошибку.
     2. Обучить гауссовский процесс на наблюдаемых точках.
     3. Найти $\boldsymbol{\lambda}_{\text{new}} = \arg\max \text{EI}(\boldsymbol{\lambda})$.
     4. Вычислить $\text{Err}(\boldsymbol{\lambda}_{\text{new}})$, добавить точку в обучающую выборку GP, повторить.
   - Преимущества: эффективно при дорогих вычислениях, хорошо работает с небольшим числом гиперпараметров (до ~20).
   - Демонстрация: ручная реализация байесовской оптимизации с использованием `GaussianProcessRegressor` из sklearn и Expected Improvement, одномерный пример $C$ для SVM. Показать итеративный процесс на графике (истинная функция, предсказания GP, функция приобретения).
   - Альтернативно: краткий код с библиотекой Optuna (если допустимо), но лучше показать ручную реализацию для понимания.

5. **Сравнение методов**
   - Кривые сходимости (лучшее найденное качество vs число итераций) для grid, random и байесовской оптимизации на задаче SVM.
   - Обсуждение, когда что выбирать: grid — для 1–2 параметров и быстрой оценки модели; random — для высокой размерности и умеренного бюджета; байесовская — для дорогих оценок.

**Вопросы для самопроверки:**
- Для начинающих (4): чем random search лучше grid при большом числе гиперпараметров, что такое суррогатная модель, почему байесовская оптимизация экономит вычисления, что такое функция приобретения.
- Для профессионалов (4): доказательство гарантии random search, математическое определение Expected Improvement, как гауссовский процесс предсказывает неопределённость, проблема масштабирования байесовской оптимизации на много параметров.
```

### Введение в байесовскую парадигму и наивный байесовский классификатор (теория)

```text
Напиши первую часть главы «Байесовские методы» для учебного пособия по машинному обучению, которое даёт глубокую математическую базу.

Стиль: основной текст — строгий и ясный, с интуитивными аналогиями; для профессионалов — вставки «Углублённый взгляд» с доказательствами и дополнительными деталями.
Используй длинные тире (—) и кавычки‑ёлочки («»). Обозначение математического ожидания — $\mathbb{E}$, дисперсии — $\mathrm{Var}$.

Начни с мостика: «В классическом машинном обучении мы искали точечные оценки параметров — например, методом максимального правдоподобия. Но что, если мы хотим не просто «лучшую» оценку, а полное распределение возможных значений параметров, учитывающее нашу априорную уверенность и ограниченность данных? Байесовский подход даёт именно это — он трактует параметры модели как случайные величины и обновляет наши представления о них по мере поступления данных.»

**Разделы (строго по порядку, без кода):**

1. **Байесовская парадигма**
   - Совместное распределение данных и параметров: $p(\mathcal{D}, \boldsymbol{\theta}) = p(\mathcal{D} \mid \boldsymbol{\theta}) p(\boldsymbol{\theta})$.
   - Априорное распределение $p(\boldsymbol{\theta})$ — отражает знания до наблюдения данных.
   - Функция правдоподобия $p(\mathcal{D} \mid \boldsymbol{\theta})$.
   - Апостериорное распределение по теореме Байеса:
     $$ p(\boldsymbol{\theta} \mid \mathcal{D}) = \frac{p(\mathcal{D} \mid \boldsymbol{\theta}) p(\boldsymbol{\theta})}{p(\mathcal{D})}, \quad p(\mathcal{D}) = \int p(\mathcal{D} \mid \boldsymbol{\theta}) p(\boldsymbol{\theta}) d\boldsymbol{\theta}. $$
   - Объясни роль нормировочной константы $p(\mathcal{D})$ (маргинальное правдоподобие, evidence).
   - Предсказательное распределение для нового наблюдения $\mathbf{x}_*$:
     $$ p(y_* \mid \mathbf{x}_*, \mathcal{D}) = \int p(y_* \mid \mathbf{x}_*, \boldsymbol{\theta}) p(\boldsymbol{\theta} \mid \mathcal{D}) d\boldsymbol{\theta}. $$
   - Отличие от частотного подхода: не точечные оценки, а распределения, естественный учёт неопределённости.

2. **Наивный байесовский классификатор: вероятностная модель**
   - Задача классификации, признаки $\mathbf{x} = (x_1,\dots,x_D)$, класс $C_k$.
   - Цель: $P(C_k \mid \mathbf{x})$. По теореме Байеса:
     $$ P(C_k \mid \mathbf{x}) = \frac{P(C_k) P(\mathbf{x} \mid C_k)}{P(\mathbf{x})} \propto P(C_k) P(\mathbf{x} \mid C_k). $$
   - Наивное предположение об условной независимости признаков при фиксированном классе:
     $$ P(\mathbf{x} \mid C_k) = \prod_{j=1}^D P(x_j \mid C_k). $$
   - Тогда
     $$ P(C_k \mid \mathbf{x}) \propto P(C_k) \prod_{j=1}^D P(x_j \mid C_k). $$
   - Вывод: это предположение резко сокращает число параметров (вместо оценки полного совместного распределения оцениваем одномерные условные распределения).

3. **Оценивание параметров**
   - Априорные вероятности классов $P(C_k)$ — частоты в обучающей выборке (MLE) или сглаженные оценки.
   - Для дискретных признаков: $P(x_j = v \mid C_k) = \frac{N_{kjv}}{N_k}$ — MLE.
   - Для непрерывных признаков: предполагаем гауссовское распределение $P(x_j \mid C_k) = \mathcal{N}(x_j \mid \mu_{jk}, \sigma^2_{jk})$. Оцениваем $\mu_{jk}$ и $\sigma^2_{jk}$ как выборочное среднее и дисперсию (MLE) или с поправкой.
   - **Сглаживание Лапласа** для категориальных признаков:
     $$ P(x_j = v \mid C_k) = \frac{N_{kjv} + \alpha}{N_k + \alpha V}, $$
     где $V$ — число возможных значений признака, $\alpha \ge 0$. Обсуди, что $\alpha=1$ — сглаживание Лапласа, $\alpha<1$ — сглаживание Лидстоуна.
   - **Углублённый взгляд: байесовский вывод для мультиномиального распределения с априорным Дирихле.**  
     Покажи, что если $P(\boldsymbol{\theta}_k) \sim \text{Dir}(\boldsymbol{\alpha})$, то апостериорное также является распределением Дирихле, а MAP‑оценка или среднее апостериорного дают формулу сглаживания. Это объясняет параметр $\alpha$ как псевдо‑счётчики.

4. **Решающее правило**
   - Классификация по максимуму апостериорной вероятности (MAP):
     $$ \hat{y} = \arg\max_k \left[ \log P(C_k) + \sum_{j=1}^D \log P(x_j \mid C_k) \right]. $$
   - Логарифмирование для численной устойчивости.

5. **Почему наивный байесовский классификатор работает даже при нарушении независимости?**
   - Обсуждение: хотя оценки вероятностей смещены, решающая граница часто корректна. Приведи пример с двумя коррелированными признаками, покажи, что произведение условных вероятностей даёт верное ранжирование классов, если зависимости действуют симметрично.

**В конце части приведи вопросы для самопроверки:**
- Для начинающих (5): что даёт априорное распределение в байесовском подходе, в чём наивность NB, какие бывают виды NB, зачем сглаживание Лапласа, как вычисляется предсказание.
- Для профессионалов (4): докажите, что сглаживание Лапласа есть байесовская оценка с априорным Дирихле, объясните устойчивость NB к нарушению независимости, почему NB быстр, как выбрать $\alpha$ в сглаживании Лапласа.

**Важно:** в этой части кода нет.
```

### Байесовская линейная регрессия

```text
Продолжи главу «Байесовские методы» после раздела о наивном байесовском классификаторе.
Теперь переходим к байесовской линейной регрессии.
Стиль: строгий математический, с подробными выводами; «Углублённый взгляд» для тонких моментов.
Код на Python — с использованием только numpy, scipy.stats; реализовать класс с нуля.

**Разделы:**

1. **Вероятностная модель линейной регрессии**
   - Напомним: $t = \mathbf{w}^T \boldsymbol{\phi}(\mathbf{x}) + \varepsilon$, $\varepsilon \sim \mathcal{N}(0, \sigma^2)$.
   - Будем считать $\sigma^2$ известной (вопрос оценивания — в следующем промпте).
   - Правдоподобие: $p(\mathbf{t} \mid \Phi, \mathbf{w}) = \mathcal{N}(\mathbf{t} \mid \Phi \mathbf{w}, \sigma^2 I)$.

2. **Априорное распределение и сопряжённость**
   - Выбираем сопряжённое априорное: $p(\mathbf{w}) = \mathcal{N}(\mathbf{w} \mid \mathbf{0}, \tau^2 I)$. Объясни, что выбор нулевого среднего типичен, а $\tau^2$ контролирует силу регуляризации.
   - Дай определение сопряжённости: априорное и апостериорное принадлежат одному семейству.

3. **Вывод апостериорного распределения**
   - Апостериор: $p(\mathbf{w} \mid \Phi, \mathbf{t}) \propto p(\mathbf{t} \mid \Phi, \mathbf{w}) p(\mathbf{w})$.
   - Логарифм апостериорной плотности (с точностью до константы):
     $$ \log p(\mathbf{w} \mid \Phi, \mathbf{t}) \propto -\frac{1}{2\sigma^2} \|\mathbf{t} - \Phi \mathbf{w}\|^2 - \frac{1}{2\tau^2} \|\mathbf{w}\|^2. $$
   - Раскрываем квадратичную форму, собираем члены:
     $$ \log p(\mathbf{w} \mid \Phi, \mathbf{t}) \propto -\frac{1}{2} \mathbf{w}^T \left( \frac{1}{\sigma^2} \Phi^T \Phi + \frac{1}{\tau^2} I \right) \mathbf{w} + \frac{1}{\sigma^2} \mathbf{w}^T \Phi^T \mathbf{t}. $$
   - Отсюда:
     $$ \mathbf{S}_N^{-1} = \frac{1}{\sigma^2} \Phi^T \Phi + \frac{1}{\tau^2} I, \quad \mathbf{m}_N = \frac{1}{\sigma^2} \mathbf{S}_N \Phi^T \mathbf{t}. $$
   - Апостериорное распределение: $p(\mathbf{w} \mid \Phi, \mathbf{t}) = \mathcal{N}(\mathbf{w} \mid \mathbf{m}_N, \mathbf{S}_N)$.

   - **Углублённый взгляд: математическое доказательство сопряжённости.**  
     Покажи, что произведение двух гауссовских плотностей (с точностью до нормировки) даёт гауссовскую плотность. Приведи явный вывод через дополнение квадрата.

4. **Связь с гребневой регрессией (Ridge)**
   - MAP‑оценка (максимум апостериорной плотности) совпадает с минимизатором регуляризованной MSE: $\hat{\mathbf{w}}_{\text{MAP}} = \mathbf{m}_N$.
   - Параметр регуляризации $\lambda = \frac{\sigma^2}{\tau^2}$.
   - Таким образом, Ridge — это байесовская оценка с гауссовским априорным.

5. **Предсказательное распределение**
   - Для нового $\mathbf{x}_*$: $p(t_* \mid \mathbf{x}_*, \Phi, \mathbf{t}) = \int p(t_* \mid \mathbf{x}_*, \mathbf{w}) p(\mathbf{w} \mid \Phi, \mathbf{t}) d\mathbf{w}$.
   - Оба распределения гауссовские, интеграл даёт:
     $$ p(t_* \mid \mathbf{x}_*, \Phi, \mathbf{t}) = \mathcal{N}\bigl(t_* \mid \mathbf{m}_N^T \boldsymbol{\phi}(\mathbf{x}_*), \sigma^2 + \boldsymbol{\phi}(\mathbf{x}_*)^T \mathbf{S}_N \boldsymbol{\phi}(\mathbf{x}_*) \bigr). $$
   - Выведи эту формулу, используя свойства линейных гауссовских моделей. Объясни, что дисперсия предсказания состоит из шума $\sigma^2$ и неопределённости в весах $\boldsymbol{\phi}_*^T \mathbf{S}_N \boldsymbol{\phi}_*$ — вдали от данных дисперсия растёт.

6. **Реализация на Python**
   - Сгенерировать синтетические данные: $y = 3x - 2 + \varepsilon$, $n=30$, $x\sim U(-3,3)$, добавить полиномиальные базисные функции до 5‑й степени.
   - Реализовать класс `BayesianLinearRegression` с методами `fit(X, t, sigma2, tau2)` и `predict(X)`. Для `predict` вычислять среднее и дисперсию предсказания.
   - Визуализировать: данные, среднее предсказание (красная линия), полосу $\pm 2$ стандартных отклонения (серая область), несколько семплов из апостериорного распределения весов (пунктирные линии). Сравнить с `Ridge(alpha=sigma2/tau2)` — показать, что среднее совпадает с Ridge.

**Вопросы для самопроверки:**
- Для начинающих (5): что даёт априорное распределение на веса, какой вид имеет апостериор в байесовской линейной регрессии, что такое предсказательное распределение, чем оно отличается от точечного прогноза, как связана дисперсия $\tau^2$ с регуляризацией.
- Для профессионалов (4): докажите сопряжённость гауссовского априорного и гауссовского правдоподобия, выведите формулы для $\mathbf{S}_N$ и $\mathbf{m}_N$, покажите, что предсказательная дисперсия растёт вдали от данных, объясните, почему MAP‑оценка совпадает с Ridge.
```

```text
Продолжи главу «Байесовские методы» после байесовской линейной регрессии.
Теперь переходим к байесовской логистической регрессии.
Стиль: строгий математический, с подробными выводами; «Углублённый взгляд» для тонких моментов.
Код на Python — реализация с нуля, сравнение с sklearn.

**Разделы:**

1. **Постановка задачи и проблема отсутствия сопряжённости**
   - Модель: $p(y=1 \mid \mathbf{x}, \mathbf{w}) = \sigma(\mathbf{w}^T \boldsymbol{\phi}(\mathbf{x}))$.
   - Правдоподобие — произведение бернуллиевских вероятностей.
   - Гауссовское априорное $p(\mathbf{w}) = \mathcal{N}(\mathbf{w} \mid \mathbf{0}, \tau^2 I)$.
   - Апостериорное распределение $p(\mathbf{w} \mid \mathcal{D})$ не является гауссовским, аналитически не вычислимо. Необходимы приближённые методы: лапласовская аппроксимация, вариационный вывод, MCMC.

2. **Лапласовская аппроксимация**
   - Идея: аппроксимировать апостериор гауссовским распределением, центр в моде (MAP‑оценке), ковариация — обратный гессиан логарифма апостериора.
   - MAP‑оценка: $\mathbf{w}_{\text{MAP}} = \arg\max_{\mathbf{w}} \bigl[ \sum_{i=1}^n (y_i \log \sigma_i + (1-y_i)\log(1-\sigma_i)) - \frac{1}{2\tau^2}\|\mathbf{w}\|^2 \bigr]$.
   - Гессиан логарифма апостериора:
     $$ \mathbf{H} = -\nabla\nabla \log p(\mathbf{w} \mid \mathcal{D}) = \Phi^T \mathbf{D} \Phi + \frac{1}{\tau^2} I, \quad \mathbf{D} = \operatorname{diag}\bigl(\sigma_i(1-\sigma_i)\bigr). $$
   - Аппроксимирующее распределение: $q(\mathbf{w}) = \mathcal{N}(\mathbf{w} \mid \mathbf{w}_{\text{MAP}}, \mathbf{S}_N), \quad \mathbf{S}_N = \mathbf{H}^{-1}$.

3. **Предсказательное распределение**
   - Точное: $p(y_*=1 \mid \mathbf{x}_*) = \int \sigma(\mathbf{w}^T \boldsymbol{\phi}_*) q(\mathbf{w}) d\mathbf{w}$ — не берётся аналитически.
   - Приближение через пробит-функцию (приближение МакКаллаха):
     $$ \int \sigma(a) \mathcal{N}(a \mid \mu, s^2) da \approx \sigma\!\left( \kappa(s^2) \mu \right), \quad \kappa(s^2) = \left(1 + \frac{\pi s^2}{8}\right)^{-1/2}. $$
   - Выведи или мотивируй это приближение (сравнение формы сигмоиды и пробит-функции).
   - Альтернатива: сэмплирование Монте‑Карло из $q(\mathbf{w})$, усреднение $\sigma(\mathbf{w}^T \boldsymbol{\phi}_*)$.

   - **Углублённый взгляд: доказательство приближения пробит-функцией.**  
     Покажи, что свёртка логистической функции с гауссовой плотностью хорошо аппроксимируется логистической функцией от масштабированного среднего, приведи вывод коэффициента $\kappa$.

4. **Реализация на Python**
   - Реализовать класс `BayesianLogisticRegressionLaplace`:
     * `fit(X, y, tau2=1.0)`: найти $\mathbf{w}_{\text{MAP}}$ с помощью `scipy.optimize.minimize` (например, BFGS), вычислить $\mathbf{H}$, $\mathbf{S}_N$.
     * `predict_proba(X)`: использовать приближение пробит-функции и для сравнения — сэмплирование (например, 1000 семплов).
     * `predict(X, threshold=0.5)`.
   - Сравнить на breast_cancer с обычной логистической регрессией (`sklearn.linear_model.LogisticRegression`). Показать, что при большом $\tau^2$ MAP близка к MLE, при малом — коэффициенты сжаты.
   - Визуализировать неопределённость: для одномерной задачи (зависимость от одного признака) показать среднюю вероятность и доверительный интервал.

**Вопросы для самопроверки:**
- Для начинающих (4): почему в логистической регрессии нет аналитического апостериора, что делает лапласовская аппроксимация, чем отличается предсказание с учётом неопределённости весов от точечного, как влияет $\tau^2$ на результат.
- Для профессионалов (3): выведите гессиан логарифма апостериора в логистической регрессии, докажите приближение пробит-функцией, объясните недостатки лапласовской аппроксимации (например, при мультимодальном апостериоре).
```

### Решающие деревья (CART): теория, критерии расщепления и вероятностный смысл

```text
Напиши первую часть главы «Решающие деревья и ансамбли» для учебного пособия по машинному обучению, которое даёт глубокую математическую базу.

Стиль: основной текст — строгий и ясный, с интуитивными пояснениями; для профессионалов — вставки «Углублённый взгляд» с доказательствами и деталями.
Используй длинные тире (—) и кавычки‑ёлочки («»). Обозначение математического ожидания — $\mathbb{E}$, дисперсии — $\mathrm{Var}$.

Начни с мостика: «Линейные модели накладывают жёсткую параметрическую форму на зависимость. Решающие деревья предлагают принципиально иной подход — они рекурсивно разбивают пространство признаков на подобласти, в каждой из которых предсказание просто. Эта идея приводит к кусочно‑постоянным функциям, способным улавливать сложные нелинейные закономерности без предположений о глобальной структуре.»

**Разделы (строго по порядку, без кода):**

1. **Геометрическая и функциональная интерпретация**
   - Дерево решений как кусочно‑постоянная аппроксимация. Разбиение $\mathbb{R}^D$ на гиперпрямоугольники с гранями, параллельными осям координат.
   - Формально: для заданного разбиения $\{R_1,\dots,R_M\}$ пространства модель имеет вид
     $$ f(\mathbf{x}) = \sum_{m=1}^M c_m \cdot \mathbf{1}_{\{\mathbf{x} \in R_m\}}, $$
     где $c_m$ — предсказание в регионе $R_m$.
   - Рекурсивное бинарное дерево: каждый внутренний узел содержит условие $x_j \le \theta$, порождающее два дочерних узла.

2. **Жадное построение дерева для регрессии**
   - Для каждого узла ищем расщепление $(j, \theta)$, минимизирующее сумму квадратов ошибок в двух образовавшихся регионах:
     $$ Q(j,\theta) = \min_{c_1}\sum_{i\in R_1(j,\theta)}(y_i-c_1)^2 + \min_{c_2}\sum_{i\in R_2(j,\theta)}(y_i-c_2)^2. $$
   - Оптимальные $c_1, c_2$ — выборочные средние в соответствующих регионах.
   - Эквивалентная форма: выбор расщепления, максимизирующего уменьшение дисперсии
     $$ \Delta = \sigma^2_{\text{parent}} - \frac{m_L}{m}\sigma^2_L - \frac{m_R}{m}\sigma^2_R, $$
     где $\sigma^2_{\text{node}} = \frac{1}{m}\sum_{i\in\text{node}}(y_i-\bar{y})^2$.

3. **Критерии impurity для классификации**
   - Пусть в узле $m$ объектов, $\hat{p}_k$ — доля класса $k$. Определим impurity-функцию $I(\hat{p}_1,\dots,\hat{p}_K)$.
   - Три стандартные меры:
     * **Ошибка классификации:** $I_{\text{err}} = 1 - \max_k \hat{p}_k$.
     * **Энтропия:** $I_{\text{ent}} = -\sum_k \hat{p}_k \log_2 \hat{p}_k$ (измеряет среднее количество информации).
     * **Индекс Джини:** $I_{\text{Gini}} = \sum_k \hat{p}_k (1-\hat{p}_k) = 1 - \sum_k \hat{p}_k^2$ (вероятность ошибки при случайной классификации по распределению $\hat{p}_k$).
   - Показать, что $I_{\text{Gini}}$ и $I_{\text{ent}}$ — строго вогнутые функции распределения, достигают минимума (0) на вырожденных распределениях и максимума на равномерном.
   - Выбор расщепления — максимизация **информационного выигрыша** (Information Gain):
     $$ \text{Gain} = I_{\text{parent}} - \frac{m_L}{m} I_L - \frac{m_R}{m} I_R. $$

   - **Углублённый взгляд: связь impurity с вероятностными моделями и байесовским риском.**  
     Покажи, что минимизация энтропии соответствует максимизации взаимной информации между признаками и классом. Индекс Джини можно рассматривать как среднеквадратичную ошибку предсказания вероятностей класса (Brier score для мультиклассового случая).

4. **Почему impurity-меры предпочитают многозначные признаки?**
   - Энтропия и Джини смещены в пользу признаков с большим количеством уникальных значений (легко получить чистые листья). Обсуждение: Information Gain Ratio в алгоритме C4.5, который нормирует Gain на энтропию признака (Split Information):
     $$ \text{GainRatio} = \frac{\text{Gain}}{-\sum_{v} \frac{m_v}{m} \log_2 \frac{m_v}{m}}. $$

5. **Проблема переобучения и регуляризация**
   - Дерево без ограничений может идеально запомнить обучающую выборку (нулевая ошибка), но будет обладать высокой дисперсией.
   - Pre‑pruning: остановка по `min_samples_split`, `min_samples_leaf`, `max_depth`, минимальному уменьшению impurity.
   - Post‑pruning (cost‑complexity pruning): для каждого поддерева $T$ определим
     $$ R_\alpha(T) = R(T) + \alpha |T|, $$
     где $R(T)$ — суммарная ошибка на обучающей выборке, $|T|$ — число листьев. При возрастании $\alpha$ получаем вложенную последовательность поддеревьев; оптимальное $\alpha$ выбирается кросс‑валидацией.

**Вопросы для самопроверки:**
- Для начинающих (5): что такое impurity-мера, почему энтропия предпочтительнее ошибки классификации, зачем обрезать деревья, как работает pre‑pruning, что такое Information Gain Ratio.
- Для профессионалов (4): докажите вогнутость энтропии и индекса Джини, объясните связь индекса Джини с ожидаемой ошибкой классификации при случайном голосовании, выведите градиент энтропии по параметрам узла, обсудите NP‑полноту задачи построения оптимального дерева минимального размера.

**Важно:** в этой части кода нет.
```

### Бэггинг: математические основы, смещение и дисперсия

```text
Продолжи главу «Решающие деревья и ансамбли» после реализации одиночного дерева.
Стиль: строгий математический, с подробными доказательствами и «Углублённым взглядом». Код на Python (sklearn, numpy).

**Разделы:**

1. **Идея бэггинга (Bootstrap Aggregating)**
   - Бэггинг снижает дисперсию предсказаний, усредняя модели, обученные на бутстреп‑выборках.
   - Генерация бутстреп‑выборки: из $n$ объектов выбираем $n$ с возвращением. Примерно $1 - e^{-1} \approx 63.2\%$ уникальных объектов.
   - Ансамбль: $\hat{f}_{\text{bag}}(\mathbf{x}) = \frac{1}{B}\sum_{b=1}^B \hat{f}_b(\mathbf{x})$.

2. **Математическое обоснование снижения дисперсии**
   - Пусть $\hat{f}_1,\dots,\hat{f}_B$ — предсказания моделей, имеющие одинаковую дисперсию $\sigma^2$ и среднюю попарную корреляцию $\rho$.
   - Дисперсия среднего:
     $$ \mathrm{Var}(\bar{f}) = \frac{1}{B^2}\left( \sum_i \mathrm{Var}(\hat{f}_i) + \sum_{i\neq j} \mathrm{Cov}(\hat{f}_i,\hat{f}_j) \right) = \frac{1}{B}\sigma^2 + \frac{B-1}{B}\rho\sigma^2 = \rho\sigma^2 + \frac{1-\rho}{B}\sigma^2. $$
     При $\rho<1$ имеем $\mathrm{Var}(\bar{f}) < \sigma^2$. При $B\to\infty$ дисперсия стремится к $\rho\sigma^2$.
   - **Углублённый взгляд: асимптотическая дисперсия бэггинга.**  
     Докажи, что бутстреп-оценки асимптотически имеют корреляцию $\rho$, зависящую от устойчивости базового алгоритма. Для неустойчивых алгоритмов (деревья) $\rho$ мала, и бэггинг даёт большое снижение дисперсии.

3. **Бэггинг для классификации**
   - Голосование большинством (hard voting) или усреднение вероятностей (soft voting). Привести формулу вероятности ошибки ансамбля в зависимости от ошибок отдельных моделей (если они независимы).

4. **Out‑of‑Bag (OOB) оценка**
   - Каждый объект не входит примерно в 36.8% бутстреп‑выборок. OOB‑предсказание — усреднение только по тем деревьям, которые не видели объект.
   - OOB‑ошибка является почти несмещённой оценкой ошибки обобщения и часто близка к кросс‑валидационной.

5. **Код и эксперименты**
   - На breast_cancer: обучить `BaggingClassifier` с `DecisionTreeClassifier` без ограничения глубины. Построить график зависимости accuracy от числа моделей $B$ для train, test и OOB.
   - Показать, что с ростом $B$ тестовая точность растёт и насыщается, а OOB‑ошибка хорошо отслеживает тестовую.

**Вопросы для самопроверки:**
- Для начинающих (4): зачем нужен бэггинг, что такое бутстреп-выборка, почему OOB-ошибка является честной оценкой, как бэггинг влияет на смещение и дисперсию.
- Для профессионалов (4): докажите формулу дисперсии ансамбля для равной корреляции, объясните, почему бэггинг не снижает смещение, сравните OOB-ошибку с кросс‑валидацией по смещению и дисперсии, выведите вероятность ошибки ансамбля при независимых классификаторах и покажите, что она экспоненциально убывает с $B$.
```

## ПРОМТ 4  
### Случайный лес: декорреляция деревьев, важность признаков и теория

```text
Продолжи главу «Решающие деревья и ансамбли» после бэггинга.
Стиль: строгий, с математическими обоснованиями и «Углублённым взглядом». Код на Python (sklearn).

**Разделы:**

1. **От бэггинга к случайному лесу**
   - Дополнительная декорреляция: при каждом расщеплении в дереве рассматривается не все $D$ признаков, а случайное подмножество размера $m$. Типичные значения: $m = \lfloor\sqrt{D}\rfloor$ (классификация), $m = \lfloor D/3\rfloor$ (регрессия).
   - Математическая мотивация: в бэггинге деревья могут быть похожи из‑за доминирующих признаков; случайный выбор признаков вынуждает деревья использовать разные признаки, снижая $\rho$ в формуле дисперсии.

2. **Анализ bias‑variance для случайного леса**
   - Объясни, что случайный лес, как и бэггинг, не увеличивает смещение по сравнению с одиночным несмещённым деревом (при $m=D$ получается обычный бэггинг). С ростом $m$ корреляция растёт, с уменьшением $m$ — падает, но может расти смещение, если важные признаки исключаются.
   - Выбор $m$ — компромисс; на практике рекомендуемые значения работают хорошо.

3. **Важность признаков в случайном лесе**
   - **Mean Decrease in Impurity (MDI):** для каждого признака суммируется уменьшение критерия (Gini/Entropy) по всем узлам всех деревьев, взвешенное на число объектов в узле. Недостаток: смещение к признакам с большим числом категорий.
   - **Permutation Importance:** для каждого признака случайно перемешиваем его значения (разрушая связь с откликом) и измеряем падение метрики. Устойчива к категориям, но может завышать важность коррелированных признаков.
   - Формальное определение: $\text{FI}_j = \mathbb{E}[\text{Err}(\tilde{X}_j) - \text{Err}(X)]$, где $\tilde{X}_j$ — матрица с переставленным $j$-м столбцом.
   - **Углублённый взгляд: связь MDI с дисперсией предсказаний.**  
     Покажи, что для регрессии с MSE суммарное уменьшение impurity по всем узлам леса, делённое на число деревьев, асимптотически сходится к объяснённой дисперсии, связанной с соответствующим признаком.

4. **Эксперименты**
   - На breast_cancer: обучить `RandomForestClassifier` и построить столбчатую диаграмму важностей (MDI и permutation). Сравнить их, обсудить различия.
   - Построить график зависимости точности от $m$ (при фиксированном $B=500$), показать оптимум.
   - Показать, что случайный лес превосходит одиночное дерево и бэггинг.

**Вопросы для самопроверки:**
- Для начинающих (4): чем случайный лес отличается от бэггинга, зачем случайно выбирать признаки, как измерить важность признака, что лучше: много глубоких деревьев или мелких.
- Для профессионалов (4): выведите формулу дисперсии ансамбля с учётом корреляции, индуцированной общими сильными признаками; объясните, почему MDI смещён; сравните permutation importance и SHAP; как выбор $m$ влияет на смещение и дисперсию.
```

## ПРОМПТ 5  
### Градиентный бустинг: функциональный градиентный спуск и вывод для произвольных функций потерь

```text
Продолжи главу «Решающие деревья и ансамбли» после случайного леса.
Стиль: строгий математический, с подробными выводами, без кода (реализация будет в следующем промпте). «Углублённый взгляд» — для доказательств.

**Разделы:**

1. **Идея градиентного бустинга как функциональной оптимизации**
   - Цель: найти функцию $F(\mathbf{x})$, минимизирующую ожидаемые потери $\mathbb{E}_{X,Y}[L(Y, F(X))]$.
   - Представляем $F$ в виде аддитивного разложения: $F_M(\mathbf{x}) = \sum_{m=0}^M \nu h_m(\mathbf{x})$, где $h_m$ — слабые модели (деревья), $\nu$ — learning rate.
   - На каждом шаге $m$ ищем $h_m$, наилучшим образом уменьшающую потери. Функциональный градиентный спуск: $h_m$ приближает направление наискорейшего убывания $L$ в функциональном пространстве.

2. **Вывод псевдо‑остатков (обобщённый градиент)**
   - Рассмотрим разложение $L(y_i, F_{m-1}(\mathbf{x}_i) + f(\mathbf{x}_i))$ по Тейлору до первого порядка:
     $$ L(y_i, F_{m-1}(\mathbf{x}_i) + f(\mathbf{x}_i)) \approx L(y_i, F_{m-1}(\mathbf{x}_i)) + g_i f(\mathbf{x}_i), $$
     где $g_i = \left.\frac{\partial L(y_i, F)}{\partial F}\right|_{F=F_{m-1}(\mathbf{x}_i)}$.
   - Чтобы максимально уменьшить потери, $f$ должна аппроксимировать отрицательный градиент: $f(\mathbf{x}_i) \propto -g_i$. Поэтому в качестве целевых значений для обучения $h_m$ берут псевдо‑остатки $r_{im} = -g_i$.
   - При квадратичной ошибке $L = \frac{1}{2}(y-F)^2$: $g_i = -(y_i - F_{m-1}(\mathbf{x}_i))$, откуда $r_{im} = y_i - F_{m-1}(\mathbf{x}_i)$ — обычные остатки.
   - При log loss (бинарная классификация, $y\in\{-1,1\}$, $L = \log(1+\exp(-2yF))$):
     $$ g_i = \frac{-2y_i}{1+\exp(2y_i F_{m-1}(\mathbf{x}_i))}, \quad r_{im} = \frac{2y_i}{1+\exp(2y_i F_{m-1}(\mathbf{x}_i))}. $$
   - При логистической регрессии ($y\in\{0,1\}$, $L = -[y\log p + (1-y)\log(1-p)]$, $p = \sigma(F)$):
     $$ g_i = p_i - y_i, \quad r_{im} = y_i - p_i. $$

3. **Выбор оптимального шага $\nu$ (learning rate)**
   - Постоянный малый learning rate $\nu$ (shrinkage) улучшает обобщение за счёт более плавного спуска, но требует больше итераций. Связь с регуляризацией: маленький $\nu$ действует как $L_2$-подобный штраф.
   - Можно также решать задачу линейного поиска: $\nu_m = \arg\min_\nu \sum_i L(y_i, F_{m-1}(\mathbf{x}_i) + \nu h_m(\mathbf{x}_i))$.

4. **Алгоритм градиентного бустинга (Generic Gradient Boosting)**
   - Инициализация: $F_0(\mathbf{x}) = \arg\min_\gamma \sum_i L(y_i, \gamma)$ (константа).
   - Для $m=1$ до $M$:
     1. Вычислить $r_{im} = -\partial L(y_i, F) / \partial F \big|_{F=F_{m-1}(\mathbf{x}_i)}$.
     2. Обучить регрессионное дерево $h_m$ на $\{(\mathbf{x}_i, r_{im})\}$.
     3. Найти $\nu_m$ (или использовать фиксированный $\nu$).
     4. Обновить $F_m(\mathbf{x}) = F_{m-1}(\mathbf{x}) + \nu \, h_m(\mathbf{x})$.

   - **Углублённый взгляд: Ньютоновский бустинг (второй порядок).**  
     Разложение до второго порядка:
     $$ L \approx L(F_{m-1}) + g_i f + \frac{1}{2} h_i f^2. $$
     Минимум по $f$ достигается при $f = -g_i / h_i$. Это используется в XGBoost для повышения точности. Выведи формулы для $g_i$ и $h_i$ для MSE и log loss.

**Вопросы для самопроверки:**
- Для начинающих (4): что такое градиентный бустинг и в чём его отличие от бэггинга, зачем нужен learning rate, что такое псевдо‑остатки, почему бустинг склонен к переобучению.
- Для профессионалов (4): выведите псевдо‑остатки для log loss и покажите их связь с остатками логистической регрессии; объясните, как shrinkage действует как регуляризация; докажите, что градиентный бустинг является функциональным градиентным спуском в гильбертовом пространстве с ядром, задаваемым деревьями; сравните сходимость градиентного и Ньютоновского бустинга.
```

## ПРОМПТ 7  
### Современный градиентный бустинг: XGBoost, LightGBM, CatBoost — алгоритмические инновации и математика

```text
Продолжи главу «Решающие деревья и ансамбли» после базовой реализации GBM.
Стиль: строгий, с детальным описанием алгоритмических и математических усовершенствований. Код — сравнение трёх библиотек на одном датасете. Импорты: `xgboost`, `lightgbm`, `catboost`, `sklearn`.

**Разделы:**

1. **XGBoost: регуляризация и Newton boosting**
   - Целевая функция с регуляризацией: $\text{Obj} = \sum_i L(y_i, \hat{y}_i) + \sum_k \Omega(f_k)$, где $\Omega(f) = \gamma T + \frac{1}{2}\lambda \|\mathbf{w}\|^2$ ($T$ — число листьев, $\mathbf{w}$ — веса листьев).
   - Использование второй производной (Newton boosting): на шаге $t$ минимизируем
     $$ \sum_{i} \left[ g_i f_t(\mathbf{x}_i) + \frac{1}{2} h_i f_t^2(\mathbf{x}_i) \right] + \gamma T + \frac{1}{2}\lambda \sum_{j=1}^T w_j^2. $$
   - Для заданной структуры дерева оптимальные веса листьев $w_j^* = -\frac{\sum_{i \in I_j} g_i}{\sum_{i \in I_j} h_i + \lambda}$, а соответствующее оптимальное значение целевой функции:
     $$ \text{Obj}^* = -\frac{1}{2} \sum_{j=1}^T \frac{(\sum_{i \in I_j} g_i)^2}{\sum_{i \in I_j} h_i + \lambda} + \gamma T. $$
     Этот показатель используется для оценки качества расщепления (аналог impurity).
   - Block structure, сжатые столбцы, эффективное разреженное расщепление.

2. **LightGBM: GOSS и EFB**
   - Gradient‑based One‑Side Sampling (GOSS): при построении гистограмм оставляются все объекты с большими градиентами и случайная подвыборка объектов с малыми градиентами (с умножением на коэффициент для компенсации). Математически: пусть $a$ — доля сохраняемых объектов с большими градиентами, $b$ — с малыми; тогда объекты с малыми градиентами умножаются на $(1-a)/b$ при вычислении статистик.
   - Exclusive Feature Bundling (EFB): объединение взаимоисключающих признаков (например, one‑hot encoded) в один признак без потери информации, что снижает размерность.
   - Leaf‑wise growth: дерево растёт не по уровням, а выбирая лист с максимальным уменьшением потерь — более асимметричные и глубокие деревья, лучшая точность.

3. **CatBoost: ordered boosting и обработка категориальных признаков**
   - Проблема target leakage: если использовать ту же выборку для вычисления target statistics (среднее целевой по категории), возникает смещение. Ordered boosting: искусственно создаются «перестановочные» обучающие наборы, и для каждого объекта используется модель, обученная на объектах, предшествующих ему в перестановке.
   - Вычисление target statistics со сглаживанием: $\hat{x}_{k} = \frac{\sum_{i=1}^n [x_{i,k}=x_{i,k}] y_i + a p}{n_k + a}$, где $p$ — глобальное среднее, $a$ — коэффициент сглаживания.
   - Симметричные деревья: на каждом уровне используется один и тот же признак и порог для всех узлов. Ускоряет предсказание.

4. **Сравнительный эксперимент**
   - Датасет: California Housing или flight delays. Обучить XGBoost, LightGBM, CatBoost с сопоставимыми параметрами (число итераций 1000, learning rate 0.05, глубина 6). Измерить время обучения и RMSE/accuracy. Построить бар‑чарты.
   - Обсудить, что влияет на выбор: CatBoost — при множестве категориальных признаков; LightGBM — на очень больших выборках; XGBoost — исторически популярен, сильная регуляризация.

**Вопросы для самопроверки:**
- Для начинающих (4): какие три библиотеки градиентного бустинга самые популярные, в чём отличие LightGBM от XGBoost по структуре дерева, зачем CatBoost хорош для категориальных признаков, что такое early stopping.
- Для профессионалов (4): выведите оптимальные веса листьев и значение целевой функции для XGBoost; объясните, как GOSS корректирует смещение оценки градиента; опишите алгоритм EFB и условия, при которых признаки могут быть объединены; сравните смещение и дисперсию leaf‑wise и level‑wise роста.
```

### Интерпретация ансамблей: SHAP, PDP, ICE и заключение главы

### Kernel trick, теорема Мерсера и свойства ядер

```text
Напиши первую часть главы «Ядерные методы» для учебного пособия по машинному обучению, которое даёт глубокую математическую базу.

Стиль: основной текст — строгий и ясный, с геометрической интуицией; для профессионалов — вставки «Углублённый взгляд» с доказательствами и деталями.
Используй длинные тире (—) и кавычки‑ёлочки («»). Обозначение нормы — $\|\cdot\|$, математического ожидания — $\mathbb{E}$, дисперсии — $\mathrm{Var}$.

Начни с мостика: «Линейные модели эффективны и интерпретируемы, но реальные данные редко бывают линейно разделимы. Мы уже видели, как в SVM используется ядерный трюк для неявного перехода в спрямляющее пространство. В этой главе мы изучим ядерные методы систематически: как превратить любой линейный алгоритм в нелинейный, не покидая уютного мира выпуклой оптимизации, и как гауссовские процессы позволяют работать с неопределённостью напрямую в функциональном пространстве.»

**Разделы (строго по порядку, без кода):**

1. **Мотивация: от линейного к нелинейному**
   - Напомни проблему: многие данные линейно неразделимы, но после нелинейного преобразования $\phi: \mathbb{R}^d \to \mathcal{H}$ становятся линейно разделимыми.
   - Пример: окружность на плоскости, отображение $\phi(x_1, x_2) = (x_1, x_2, x_1^2 + x_2^2)$.
   - Проблема: размерность $\mathcal{H}$ может быть огромной или бесконечной, явно работать с $\phi(\mathbf{x})$ невозможно.

2. **Kernel trick (ядерный трюк)**
   - Заметим, что многие линейные алгоритмы (SVM, регрессия, PCA) зависят от данных только через скалярные произведения $\langle \phi(\mathbf{x}_i), \phi(\mathbf{x}_j) \rangle$.
   - Определим функцию ядра $K(\mathbf{x}, \mathbf{x}') = \langle \phi(\mathbf{x}), \phi(\mathbf{x}') \rangle_{\mathcal{H}}$.
   - Если мы можем вычислить $K$ напрямую, не строя $\phi$, то получаем нелинейное обобщение без увеличения вычислительной сложности.
   - Пример: полиномиальное ядро $K(\mathbf{x}, \mathbf{x}') = (\mathbf{x}^T \mathbf{x}' + c)^d$. Покажи явно для $d=2, c=0$: $\phi(\mathbf{x})$ содержит все мономы второй степени с коэффициентами, и скалярное произведение вычисляется за $O(d)$ вместо $O(d^2)$.

3. **Теорема Мерсера**
   - Формулировка: функция $K: \mathcal{X}\times\mathcal{X} \to \mathbb{R}$ является ядром (т.е. соответствует скалярному произведению в некотором гильбертовом пространстве) тогда и только тогда, когда она симметрична и положительно полуопределена (п.п.о.).
   - Определение п.п.о.: для любого набора $\{\mathbf{x}_1,\dots,\mathbf{x}_m\}$ матрица Грама $K_{ij} = K(\mathbf{x}_i,\mathbf{x}_j)$ положительно полуопределена.
   - Интуиция: это условие гарантирует, что $K$ задаёт допустимую меру близости, соответствующую некоторому евклидову пространству.
   - **Углублённый взгляд: доказательство (набросок).** Покажи, как из п.п.о. строится гильбертово пространство (пополнение линейной оболочки функций $K(\cdot, \mathbf{x})$ со скалярным произведением, определённым через ядро). Следствие: любое п.п.о. ядро можно использовать в SVM, регрессии и т.д.

4. **Конструирование новых ядер из старых**
   - Свойства замкнутости: сумма ядер, произведение ядер, умножение на положительную константу, предел — всё это даёт ядра. Это позволяет строить сложные ядра для специфических структур данных (тексты, графы, изображения).

5. **Обзор популярных ядер**
   - Линейное: $K(\mathbf{x},\mathbf{x}') = \mathbf{x}^T \mathbf{x}'$.
   - Полиномиальное: $K(\mathbf{x},\mathbf{x}') = (\gamma \mathbf{x}^T \mathbf{x}' + r)^d$.
   - RBF (гауссово): $K(\mathbf{x},\mathbf{x}') = \exp(-\gamma \|\mathbf{x} - \mathbf{x}'\|^2)$. Обсуди, что $\gamma$ контролирует «ширину» соседства.
   - Сигмоидальное: $K(\mathbf{x},\mathbf{x}') = \tanh(\gamma \mathbf{x}^T \mathbf{x}' + r)$ — не всегда п.п.о., но используется.
   - Ядра для специальных данных: строковые (спектральные), графовые (random walk), вероятностные (ядро Фишера).

   - **Углублённый взгляд: ядро RBF и его спрямляющее пространство.**  
     Используя разложение экспоненты в ряд Тейлора, покажи, что $\phi(\mathbf{x})$ бесконечномерно и содержит все мономы $\mathbf{x}^\alpha$ с убывающими весами. Это объясняет универсальность RBF.

6. **Теорема о представлении (Representer Theorem)**
   - Для широкого класса задач минимизации эмпирического риска с регуляризацией в RKHS решение имеет вид $f(\cdot) = \sum_{i=1}^n \alpha_i K(\cdot, \mathbf{x}_i)$.
   - Приведи формулировку и интуитивное обоснование: оптимальная функция лежит в линейной оболочке ядерных функций, центрированных на обучающих точках.
   - Это ключевой факт, который будет использоваться в ядерной логистической регрессии и KPCA.

**В конце части приведи вопросы для самопроверки:**
- Для начинающих (5): что такое kernel trick, зачем он нужен, примеры ядер, что означает положительная полуопределённость, как строить новые ядра.
- Для профессионалов (4): докажите, что полиномиальное ядро степени $d$ является п.п.о., объясните связь между RKHS и reproducing property, сформулируйте Representer Theorem и докажите для случая квадратичных потерь, почему RBF-ядро соответствует бесконечномерному пространству.

**Важно:** никакого кода.
```

```text
Продолжи главу «Ядерные методы» после раздела о SVM.
Стиль: строгий математический с интуитивными пояснениями; «Углублённый взгляд» для доказательств.
Код на Python; импорты (numpy, matplotlib, sklearn.linear_model, sklearn.kernel_approximation, sklearn.pipeline) добавить к предыдущим.

**Разделы:**

1. **Линейная логистическая регрессия и её ограничения**
   - Напомни: модель $p(y=1|\mathbf{x}) = \sigma(\mathbf{w}^T \mathbf{x})$, линейная граница.
   - Если данные нелинейны, как её обобщить? Вариант: добавить базисные функции, но их число быстро растёт.

2. **Ядровая версия через Representer Theorem**
   - Согласно теореме о представлении, решение задачи минимизации эмпирического риска с $L_2$-регуляризацией в RKHS имеет вид $f(\mathbf{x}) = \sum_{i=1}^n \alpha_i K(\mathbf{x}_i, \mathbf{x})$.
   - Подставим это представление в функцию потерь логистической регрессии:
     $$ \min_{\boldsymbol{\alpha}} \left[ \sum_{i=1}^n \log(1 + \exp(-y_i \sum_{j} \alpha_j K(\mathbf{x}_j, \mathbf{x}_i))) + \frac{\lambda}{2} \boldsymbol{\alpha}^T \mathbf{K} \boldsymbol{\alpha} \right], $$
     где $\mathbf{K}$ — матрица Грама $K_{ij} = K(\mathbf{x}_i, \mathbf{x}_j)$, $y_i \in \{-1, +1\}$.
   - Преимущества: размерность оптимизации равна числу наблюдений $n$, а не числу признаков.

3. **Алгоритмы обучения**
   - Можно применить методы оптимизации (градиентный спуск, L‑BFGS) к $\boldsymbol{\alpha}$.
   - На практике используют аппроксимации для масштабируемости: Nystroem method, Random Fourier Features (RFF).
   - Объясни идею RFF: сэмплируем случайные признаки $z(\mathbf{x})$ такие, что $z(\mathbf{x})^T z(\mathbf{x}') \approx K(\mathbf{x}, \mathbf{x}')$, затем обучаем обычную логистическую регрессию на $z(\mathbf{x})$.

4. **Реализация и сравнение**
   - Сгенерируй нелинейные данные (moons). Построй `LogisticRegression` с полиномиальными признаками (2-й степени) и с `RBFSampler` + `LogisticRegression`. Сравни точность и форму границы с SVM с RBF.
   - Покажи, как `KernelRidge` (ядровая ридж-регрессия) работает для регрессии.

   - **Углублённый взгляд: вывод двойственной задачи для ядерной логистической регрессии.**  
     Покажи, что можно получить двойственную задачу, подобную SVM, но с ограничениями $0 \le \alpha_i \le C$, и разница лишь в функции потерь (log loss vs hinge loss).

5. **Сравнение с SVM**
   - Обсуди: SVM разрежен (многие $\alpha_i=0$), ядерная логистическая регрессия — нет. SVM даёт отступ, логистическая регрессия — вероятности. Выбор зависит от потребности.

**Вопросы для самопроверки:**
- Для начинающих (4): чем ядерная логистическая регрессия отличается от обычной, что такое Representer Theorem, как Random Fourier Features аппроксимируют ядро, какие преимущества у ядерной логистической регрессии перед SVM.
- Для профессионалов (3): выведите двойственную задачу для ядерной логистической регрессии, докажите, что решение имеет вид $\sum \alpha_i K(\mathbf{x}_i, \mathbf{x})$, сравните сложность обучения SVM и ядерной логистической регрессии по памяти и времени.
```

## ПРОМТ 4  
### Ядерный PCA (KPCA)

```text
Продолжи главу «Ядерные методы» после ядерной логистической регрессии.
Стиль: строгий математический, с подробными выводами и «Углублённым взглядом».
Код на Python (numpy, matplotlib, sklearn.decomposition.KernelPCA). Импорты не повторять.

**Разделы:**

1. **Мотивация: PCA для нелинейных данных**
   - Обычный PCA находит направления максимальной дисперсии в исходном пространстве. Для данных с нелинейной структурой (например, спирали, концентрические круги) это неэффективно.
   - Идея: отобразить данные в спрямляющее пространство $\mathcal{H}$ с помощью $\phi$ и выполнить PCA там. Так как $\mathcal{H}$ может быть огромным, используем ядерный трюк.

2. **Вывод KPCA**
   - Пусть $\phi(\mathbf{x}_i)$ — образы, предполагаем, что они центрированы (можно вычесть среднее). Ковариационная матрица в $\mathcal{H}$: $C = \frac{1}{n} \sum_{i=1}^n \phi(\mathbf{x}_i) \phi(\mathbf{x}_i)^T$.
   - Задача на собственные значения: $C \mathbf{v} = \lambda \mathbf{v}$. Так как $\mathbf{v}$ лежит в оболочке $\phi(\mathbf{x}_i)$, $\mathbf{v} = \sum_{i=1}^n a_i \phi(\mathbf{x}_i)$.
   - Подставляем, домножаем слева на $\phi(\mathbf{x}_k)^T$, получаем $\frac{1}{n} \sum_i a_i \langle \phi(\mathbf{x}_k), \phi(\mathbf{x}_i) \rangle \langle \phi(\mathbf{x}_i), \phi(\mathbf{x}_j) \rangle$... Выводим эквивалентную проблему:
     $$ n \lambda \mathbf{a} = \tilde{\mathbf{K}} \mathbf{a}, $$
     где $\tilde{\mathbf{K}}$ — центрированная ядерная матрица: $\tilde{\mathbf{K}} = \mathbf{K} - \mathbf{1}_n \mathbf{K} - \mathbf{K} \mathbf{1}_n + \mathbf{1}_n \mathbf{K} \mathbf{1}_n$, $\mathbf{1}_n$ — матрица $1/n$.
   - Таким образом, KPCA сводится к собственным значениям центрированной матрицы Грама.
   - Проекция новой точки $\mathbf{x}$ на $k$-ю главную компоненту: $\langle \mathbf{v}_k, \phi(\mathbf{x}) \rangle = \sum_{i=1}^n a_{ki} K(\mathbf{x}_i, \mathbf{x})$.

3. **Реконструкция в исходном пространстве (pre‑image)**
   - В отличие от обычного PCA, обратное отображение из $\mathcal{H}$ в исходное пространство неоднозначно и сложно. Можно решать задачу оптимизации для поиска прообраза, но это вычислительно затратно. На практике KPCA чаще используют для визуализации и как этап предобработки.

4. **Примеры и реализация**
   - Сгенерировать нелинейные данные: `make_circles`, `make_swiss_roll` (3D). Применить KPCA с RBF ядром. Визуализировать проекции на первые две компоненты — данные должны разделиться или развернуться в линейно разделимую форму.
   - Сравнить с обычным PCA: на тех же данных PCA не сможет разделить круги.
   - Реализовать KPCA с нуля (на numpy) и через `KernelPCA` из sklearn. Показать, что результаты совпадают.

5. **Выбор ядра и его параметров**
   - Для RBF ядра $\gamma$ управляет «масштабом»: при малых $\gamma$ данные стягиваются, при больших — переобучается. Показать на примере.
   - Аналог собственных значений: можно оценить долю объяснённой дисперсии в $\mathcal{H}$.

**Вопросы для самопроверки:**
- Для начинающих (4): чем KPCA отличается от обычного PCA, зачем нужно центрирование ядерной матрицы, как выбирать $\gamma$ в RBF, какие применения KPCA.
- Для профессионалов (3): выведите формулу центрирования ядерной матрицы, докажите, что решение задачи на собственные значения эквивалентно $\tilde{\mathbf{K}} \mathbf{a} = n\lambda \mathbf{a}$, объясните, почему реконструкция (pre‑image) сложна, и приведите примеры методов.
```

## ПРОМТ 5  
### Гауссовские процессы (GP): регрессия и связь с ядерными методами

```text
Продолжи главу «Ядерные методы» после KPCA.
Стиль: строгий математический, с геометрической интуицией и «Углублённым взглядом».
Код на Python (sklearn.gaussian_process, numpy, matplotlib). Импорты не повторять.

**Разделы:**

1. **Гауссовский процесс как распределение над функциями**
   - Определение: GP — это набор случайных величин, любое конечное подмножество которых имеет совместное нормальное распределение.
   - Полностью задаётся функцией среднего $m(\mathbf{x})$ и ковариационной функцией (ядром) $k(\mathbf{x}, \mathbf{x}')$.
   - Обозначение: $f(\mathbf{x}) \sim \mathcal{GP}(m(\mathbf{x}), k(\mathbf{x}, \mathbf{x}'))$.
   - Интуиция: ядро определяет гладкость и свойства функций (периодичность, стационарность и т.д.). Примеры рисования семплов из GP с разными ядрами.

2. **Регрессия с гауссовским процессом (GPR)**
   - Модель: $y_i = f(\mathbf{x}_i) + \varepsilon_i$, $\varepsilon_i \sim \mathcal{N}(0, \sigma_n^2)$.
   - Совместное распределение наблюдений $\mathbf{y}$ и предсказания $f_*$ на новой точке $\mathbf{x}_*$:
     $$ \begin{bmatrix} \mathbf{y} \\ f_* \end{bmatrix} \sim \mathcal{N}\left( \mathbf{0}, \begin{bmatrix} \mathbf{K} + \sigma_n^2 I & \mathbf{k}_* \\ \mathbf{k}_*^T & k(\mathbf{x}_*, \mathbf{x}_*) \end{bmatrix} \right), $$
     где $\mathbf{K}$ — матрица Грама, $\mathbf{k}_*$ — вектор ядер между $\mathbf{x}_*$ и обучающими точками.
   - Условное распределение $f_* | \mathbf{y}$ также гауссовское:
     $$ \mu_* = \mathbf{k}_*^T (\mathbf{K} + \sigma_n^2 I)^{-1} \mathbf{y}, $$
     $$ \sigma^2_* = k(\mathbf{x}_*, \mathbf{x}_*) - \mathbf{k}_*^T (\mathbf{K} + \sigma_n^2 I)^{-1} \mathbf{k}_*. $$
   - Выведи эти формулы, используя свойства многомерного нормального распределения.
   - Среднее $\mu_*$ — это предсказание, дисперсия $\sigma^2_*$ — неопределённость, которая растёт вдали от данных.

3. **Обучение гиперпараметров**
   - Гиперпараметры ядра (например, длина шкалы $l$ в RBF) и шум $\sigma_n^2$ можно выбирать, максимизируя логарифмическое маржинальное правдоподобие (evidence):
     $$ \log p(\mathbf{y} | X, \boldsymbol{\theta}) = -\frac{1}{2} \mathbf{y}^T (\mathbf{K} + \sigma_n^2 I)^{-1} \mathbf{y} - \frac{1}{2} \log |\mathbf{K} + \sigma_n^2 I| - \frac{n}{2} \log 2\pi. $$
   - Это автоматически балансирует сложность модели и качество подгонки (Occam's razor). Используем градиентный подъём.

4. **Реализация на Python**
   - Сгенерировать синтетические данные: $y = x \sin(x) + \text{шум}$.
   - Использовать `GaussianProcessRegressor` из sklearn с RBF ядром и белым шумом. Подобрать гиперпараметры через `fit`.
   - Построить график: данные, предсказанное среднее, 95% доверительный интервал. Показать, как интервал расширяется в областях без данных.
   - Сравнить с Ridge-регрессией и SVM-регрессией: GP даёт не только предсказание, но и оценку неопределённости.

5. **Связь с ядерными методами**
   - Покажи, что предсказательное среднее GPR эквивалентно ядерной ридж-регрессии (Kernel Ridge Regression). Действительно, $\boldsymbol{\alpha} = (\mathbf{K} + \sigma_n^2 I)^{-1} \mathbf{y}$, и предсказание $\mu_* = \sum_i \alpha_i k(\mathbf{x}_i, \mathbf{x}_*)$. Это в точности решение задачи минимизации $\|\mathbf{y} - \mathbf{K} \boldsymbol{\alpha}\|^2 + \sigma_n^2 \boldsymbol{\alpha}^T \mathbf{K} \boldsymbol{\alpha}$.
   - Но GPR дополнительно предоставляет дисперсию предсказания и позволяет обучать гиперпараметры через evidence.

6. **Когда использовать GP?**
   - Преимущества: калиброванная неопределённость, малое количество гиперпараметров, интерпретируемость.
   - Недостатки: кубическая сложность $O(n^3)$ от числа наблюдений, необходимость обращения матрицы. Методы аппроксимации (sparse GP, inducing points) — кратко упомянуть.

**Вопросы для самопроверки:**
- Для начинающих (5): что такое гауссовский процесс, какие параметры он оценивает, чем GPR лучше обычной регрессии, как интерпретировать доверительную полосу, что такое evidence.
- Для профессионалов (4): выведите формулы для $\mu_*$ и $\sigma^2_*$, покажите, что GPR эквивалентен ядерной ридж-регрессии, объясните, как максимизация evidence автоматически настраивает сложность, назовите методы масштабирования GP на большие данные.
```

### Метод главных компонент (PCA) и связь с сингулярным разложением (SVD)

```text
Напиши первую часть главы «Снижение размерности и обучение без учителя» для учебного пособия по машинному обучению с глубокой математической базой.

Стиль: основной текст — строгий и ясный, с геометрической интуицией; для профессионалов — вставки «Углублённый взгляд» с доказательствами и дополнительными деталями.
Используй длинные тире (—) и кавычки‑ёлочки («»). Обозначение математического ожидания — $\mathbb{E}$, дисперсии — $\mathrm{Var}$, следа матрицы — $\mathrm{tr}$.

Начни с мостика: «До сих пор мы в основном работали с размеченными данными. Однако часто меток нет, а размерность признакового пространства огромна — например, тысячи экспрессий генов или пикселей изображения. Как выделить наиболее информативные направления, вдоль которых данные меняются сильнее всего? Метод главных компонент (PCA) решает эту задачу, находя ортогональные направления максимальной дисперсии, и тем самым позволяет сжать данные, удалить шум и визуализировать многомерные облака точек.»

**Разделы (строго по порядку, без кода):**

1. **Постановка задачи снижения размерности**
   - Данные $\mathbf{X} \in \mathbb{R}^{n \times D}$ (центрированы: каждый признак имеет нулевое среднее). Цель: найти линейное отображение на $k$-мерное подпространство ($k < D$), максимально сохраняющее вариацию данных.
   - Формулировка максимизации дисперсии проекции: ищем направление $\mathbf{w}_1$, такое что $\|\mathbf{w}_1\| = 1$ и дисперсия проекций $\mathbf{X} \mathbf{w}_1$ максимальна:
     $$ \mathbf{w}_1 = \arg\max_{\|\mathbf{w}\|=1} \frac{1}{n-1}\|\mathbf{X w}\|^2 = \arg\max_{\|\mathbf{w}\|=1} \mathbf{w}^T \mathbf{S} \mathbf{w}, $$
     где $\mathbf{S} = \frac{1}{n-1} \mathbf{X}^T \mathbf{X}$ — выборочная ковариационная матрица.

2. **Собственные векторы ковариационной матрицы**
   - Задача сводится к поиску собственных векторов $\mathbf{S}$: $\mathbf{S} \mathbf{w} = \lambda \mathbf{w}$.
   - Первый главный компонент $\mathbf{w}_1$ — собственный вектор с наибольшим собственным значением $\lambda_1$, которое равно дисперсии вдоль этого направления.
   - Второй главный компонент ортогонален первому и соответствует второму по величине собственному значению, и т.д.
   - Показать, что сумма всех собственных значений равна суммарной дисперсии данных: $\sum_{j=1}^D \lambda_j = \mathrm{tr}(\mathbf{S})$.

3. **Связь с сингулярным разложением (SVD)**
   - Вместо формирования ковариационной матрицы можно напрямую применить SVD к центрированной матрице данных: $\mathbf{X} = \mathbf{U} \mathbf{\Sigma} \mathbf{V}^T$, где $\mathbf{U} \in \mathbb{R}^{n \times D}$, $\mathbf{\Sigma} = \mathrm{diag}(\sigma_1,\dots,\sigma_D)$, $\mathbf{V} \in \mathbb{R}^{D \times D}$ ортогональна.
   - Тогда $\mathbf{S} = \frac{1}{n-1} \mathbf{V} \mathbf{\Sigma}^T \mathbf{\Sigma} \mathbf{V}^T$, следовательно, собственные векторы — столбцы $\mathbf{V}$, собственные значения $\lambda_j = \sigma_j^2 / (n-1)$.
   - Главные компоненты (проекции) — столбцы $\mathbf{U} \mathbf{\Sigma}$.
   - Преимущества SVD: численная устойчивость без явного вычисления $\mathbf{X}^T \mathbf{X}$.

   - **Углублённый взгляд: геометрический смысл SVD.**  
     Покажи, что $\mathbf{X} \mathbf{v}_j = \sigma_j \mathbf{u}_j$, где $\mathbf{v}_j$ — направление максимальной вариации, $\sigma_j \mathbf{u}_j$ — координаты объектов на этом направлении. Наибольшие сингулярные числа соответствуют самым «энергичным» осям эллипсоида данных.

4. **Выбор числа главных компонент**
   - Критерии: доля объяснённой дисперсии ($\sum_{j=1}^k \lambda_j / \sum_{j=1}^D \lambda_j \ge 90\%$), график «каменистой осыпи» (scree plot) — ищем «локоть», где убывание собственных значений замедляется.
   - Баланс между сжатием и сохранением информации.

5. **Интерпретация главных компонент**
   - Главные компоненты — линейные комбинации исходных признаков: $\mathbf{w}_j^T \mathbf{x}$. Коэффициенты $\mathbf{w}_j$ (нагрузки) показывают, какие признаки вносят наибольший вклад в компоненту.
   - Ограничения PCA: линейность, чувствительность к масштабу признаков (требуется стандартизация), игнорирование нелинейных структур (позже будет KPCA).

**В конце части приведи вопросы для самопроверки:**
- Для начинающих (5): что максимизирует PCA, что такое главная компонента, как выбрать число компонент, зачем стандартизировать данные перед PCA, как связаны собственные значения и дисперсия.
- Для профессионалов (4): докажите, что первый главный компонент является собственным вектором ковариационной матрицы с наибольшим собственным значением; выведите связь между SVD и PCA; объясните, почему SVD предпочтительнее для численной устойчивости; как влияет отсутствие центрирования на результат.

**Важно:** никакого кода.
```

## ПРОМТ 3  
### Кластеризация: K‑средних (K‑means)

```text
Продолжи главу «Снижение размерности и обучение без учителя» после PCA.
Теперь переходим к кластеризации. Стиль: строгий математический, с доказательствами и кодом.
Импорты (numpy, matplotlib, sklearn.cluster, sklearn.datasets, sklearn.metrics) добавить к текущим.

**Разделы:**

1. **Постановка задачи кластеризации**
   - Даны неразмеченные данные $\mathbf{X} \in \mathbb{R}^{n \times D}$. Разбить на $K$ групп (кластеров), чтобы объекты внутри группы были «похожи», а между группами — «различны».
   - Формализация для K‑means: минимизация внутрикластерной суммы квадратов (within‑cluster sum of squares, WCSS):
     $$ J = \sum_{k=1}^K \sum_{i \in C_k} \|\mathbf{x}_i - \boldsymbol{\mu}_k\|^2, $$
     где $C_k$ — множество индексов объектов в $k$-м кластере, $\boldsymbol{\mu}_k$ — его центроид.

2. **Алгоритм Ллойда**
   - Инициализация центроидов (случайно или K‑means++).
   - Повторять до сходимости:
     1. **Шаг присваивания:** каждый объект приписывается к ближайшему центроиду (по евклидову расстоянию).
        $$ C_k^{(t)} = \{ i : \|\mathbf{x}_i - \boldsymbol{\mu}_k^{(t)}\|^2 \le \|\mathbf{x}_i - \boldsymbol{\mu}_j^{(t)}\|^2 \;\forall j \}. $$
     2. **Шаг обновления:** центроиды пересчитываются как средние объектов в кластере.
        $$ \boldsymbol{\mu}_k^{(t+1)} = \frac{1}{|C_k^{(t)}|} \sum_{i \in C_k^{(t)}} \mathbf{x}_i. $$
   - Доказательство монотонного убывания $J$ на каждой итерации.

3. **Сходимость и локальные минимумы**
   - Алгоритм сходится к локальному минимуму за конечное число шагов (так как число возможных разбиений конечно, а $J$ убывает).
   - Из‑за невыпуклости результат зависит от начальной инициализации. Метод K‑means++ выбирает начальные центроиды далеко друг от друга, что улучшает качество и скорость сходимости.

4. **Выбор числа кластеров $K$**
   - Метод локтя: построить график $J$ от $K$, искать «локоть», где убывание замедляется.
   - Силуэтный коэффициент (подробнее в следующем промпте).
   - Gap statistic (идея, без реализации).

5. **Реализация с нуля и сравнение**
   - Напиши класс `KMeans` с методами `fit`, `predict`. Используй евклидово расстояние и инициализацию случайными точками или K‑means++ (опционально).
   - Сравни с `sklearn.cluster.KMeans` на `make_blobs`. Покажи визуализацию кластеров и центроидов.
   - На breast_cancer (без меток) примени K‑means с $K=2$, сравни полученные кластеры с истинными метками (внешняя оценка — ARI, NMI).

**Вопросы для самопроверки:**
- Для начинающих (5): как работает K‑means, что такое центроид, как выбрать $K$, почему плохая инициализация влияет на результат, что такое K‑means++.
- Для профессионалов (4): докажите, что на шаге обновления среднее арифметическое действительно минимизирует сумму квадратов расстояний до центроида; покажите, что алгоритм сходится за конечное число шагов; обсудите ограничения K‑means (форма кластеров — сферы, чувствительность к масштабу); как K‑means связан с EM‑алгоритмом для смеси гауссиан.
```

## ПРОМПТ 4  
### Иерархическая кластеризация и DBSCAN

```text
Продолжи главу «Снижение размерности и обучение без учителя» после K‑means.
Стиль: строгий, с геометрической интуицией, кодом на Python и сравнением методов.
Импорты (sklearn.cluster, scipy.cluster.hierarchy) уже доступны.

**Разделы:**

1. **Иерархическая кластеризация**
   - Подходы: агломеративный (снизу вверх) и дивизимный (сверху вниз). Основное внимание — агломеративной.
   - Алгоритм: каждый объект — отдельный кластер; на каждом шаге объединяем два ближайших кластера, пока не останется один.
   - Меры расстояния между кластерами (linkage):
     * **Single linkage:** расстояние между ближайшими точками (склонность к «цепочкам»).
     * **Complete linkage:** расстояние между самыми дальними точками (компактные кластеры).
     * **Average linkage:** среднее расстояние.
     * **Ward's method:** минимизация увеличения внутрикластерной дисперсии при объединении.
   - Дендрограмма — визуализация иерархии. Выбор числа кластеров — «разрезание» дендрограммы на нужной высоте.

2. **Преимущества и недостатки**
   - Нет необходимости задавать $K$ заранее, дендрограмма даёт наглядную картину.
   - Вычислительная сложность $O(n^3)$ для наивной реализации, $O(n^2 \log n)$ для некоторых linkage.
   - Чувствительность к шуму и выбросам (особенно single linkage).

3. **DBSCAN (Density‑Based Spatial Clustering of Applications with Noise)**
   - Идея: кластеры — области высокой плотности, разделённые областями низкой плотности.
   - Параметры: $\varepsilon$ (радиус окрестности) и `min_samples` (минимальное число точек для образования плотной области).
   - Классификация точек:
     * **Core point:** имеет не менее `min_samples` соседей в $\varepsilon$-окрестности.
     * **Border point:** не является core, но достижима из core point.
     * **Noise:** не принадлежит ни одному кластеру.
   - Алгоритм: для каждой непосещённой точки, если она core, начинаем новый кластер и рекурсивно добавляем все $\varepsilon$-достижимые точки.
   - Преимущества: не требует задания числа кластеров, находит кластеры произвольной формы, устойчив к шуму.
   - Недостатки: чувствителен к параметрам $\varepsilon$ и `min_samples`, плохо работает с кластерами разной плотности.

4. **Реализация и сравнение на Python**
   - Сгенерируй данные с невыпуклыми кластерами: `make_moons`, `make_blobs` с длинными вытянутыми кластерами.
   - Примени `KMeans`, `AgglomerativeClustering` (с разными linkage), `DBSCAN`. Визуализируй результаты разными цветами.
   - Для DBSCAN покажи, как noise-точки помечаются отдельно.
   - Построй дендрограмму с помощью `scipy.cluster.hierarchy.dendrogram` и покажи, как разрез даёт кластеры.

**Вопросы для самопроверки:**
- Для начинающих (5): чем иерархическая кластеризация отличается от K‑means, что такое дендрограмма, как работает DBSCAN, какие у него преимущества, что такое core point.
- Для профессионалов (4): сравните single и complete linkage с точки зрения устойчивости к шуму; объясните, как параметр $\varepsilon$ в DBSCAN связан с понятием плотности; выведите сложность алгоритма DBSCAN; обсудите, почему DBSCAN не работает хорошо при большой размерности (проклятие размерности).
```

```text
Напиши заключительную часть главы «Снижение размерности и обучение без учителя».
Стиль: строгий, с математическими определениями метрик, кодом для их вычисления и итоговым сравнением.

**Разделы:**

1. **Проблема оценки качества кластеризации**
   - В отличие от обучения с учителем, истинные метки обычно неизвестны. Метрики делятся на внутренние (оценивают компактность и разделимость) и внешние (сравнивают с эталонной разметкой, если она есть).

2. **Внешние метрики (с истинными метками)**
   - Adjusted Rand Index (ARI): измеряет согласие двух разбиений с поправкой на случайность. Диапазон $[-1, 1]$, 1 — идеальное совпадение.
   - Normalized Mutual Information (NMI): нормированная взаимная информация, $[0,1]$.
   - Кратко приведи формулы и код `sklearn.metrics.adjusted_rand_score`, `normalized_mutual_info_score`.

3. **Внутренние метрики (без истинных меток)**
   - **Силуэтный коэффициент (Silhouette score):** для каждого объекта $i$ вычисляется
     $$ s(i) = \frac{b(i) - a(i)}{\max\{a(i), b(i)\}}, $$
     где $a(i)$ — среднее расстояние до точек своего кластера, $b(i)$ — минимальное среднее расстояние до точек другого кластера. $s(i) \in [-1, 1]$: близко к 1 — хорошее соответствие, около 0 — точка между кластерами, отрицательное — вероятно, точка в чужом кластере.
   - Средний силуэт по всем точкам — оценка кластеризации.
   - **Индекс Дэвиса–Болдина (Davies‑Bouldin index):** среднее по кластерам максимального отношения суммы внутрикластерных расстояний к расстоянию между центроидами. Чем меньше, тем лучше.
   - Показать код `silhouette_score`, `davies_bouldin_score` из sklearn.

4. **Сравнение K‑means, иерархической и DBSCAN**
   - Построить таблицу: требуемое число кластеров, форма кластеров, устойчивость к шуму, масштабируемость, интерпретируемость.
   - На нескольких синтетических датасетах (сферы, луны, кольца) вычислить и сравнить силуэт и Davies‑Bouldin для каждого метода.
   - Обсудить, когда что применять: K‑means — когда кластеры примерно сферические и одинакового размера; иерархическая — для малых выборок и когда нужна дендрограмма; DBSCAN — когда ожидаются кластеры произвольной формы и есть шум.

5. **Заключение и мост к следующим темам**
   - Снижение размерности и кластеризация — фундамент анализа данных без учителя. PCA даёт линейные направления вариации, кластеризация — группировку объектов. Более сложные методы (t‑SNE, UMAP) позволяют визуализировать нелинейные многообразия; глубокие автоэнкодеры — нелинейное снижение размерности.
   - В следующей части мы перейдём к нейронным сетям, где идеи оптимизации и представлений признаков получат новое развитие.

**Вопросы для самопроверки (объединяют всю главу):**
- Для начинающих (6): что такое PCA и для чего он нужен, как работает K‑means, в чём отличие DBSCAN от K‑means, что показывает дендрограмма, как измерить качество кластеризации без истинных меток, какой метод выбрать для данных с шумом.
- Для профессионалов (5): докажите, что PCA минимизирует ошибку реконструкции при проекции на подпространство; выведите связь между K‑means и матричной факторизацией; объясните, почему силуэтный коэффициент может быть обманчив при несферических кластерах; опишите ограничения DBSCAN в высоких размерностях; сравните вычислительную сложность PCA, K‑means и иерархической кластеризации.

**Задания повышенной сложности**
1. Реализовать PCA через итерационный алгоритм (степенной метод) и сравнить скорость с SVD на больших матрицах.
2. Реализовать K‑means++ инициализацию и доказать, что она даёт $O(\log K)$-приближение к оптимальному значению WCSS.
3. Применить DBSCAN к географическим данным (координаты преступлений), подобрать $\varepsilon$ с помощью k‑distance графика.
4. Сравнить ARI и NMI для оценки кластеризации при разном числе кластеров, построить графики.
5. Используя PCA для уменьшения размерности до 2D, сравнить визуально кластеры, найденные K‑means и DBSCAN, на данных с нелинейной структурой (например, спираль).
```

```text
Напиши первую часть главы «Метрические методы: метод k‑ближайших соседей» для учебного пособия, адресованного одновременно начинающим и профессионалам.

Стиль: основной текст интуитивный, с яркими аналогиями; для профессионалов — углублённые вставки «Углублённый взгляд» со строгими выводами и деталями.
Используй длинные тире (—) и кавычки‑ёлочки («»). Обозначение дисперсии — $\mathrm{Var}$.

Начни с мостика: «В предыдущих главах мы строили параметрические модели — прямые, гиперплоскости, логистические кривые. Метод k‑ближайших соседей (kNN) принципиально иной: он не создаёт глобальной формулы, а принимает решение локально, опираясь на сходство объектов».

**Разделы (строго по порядку):**

1. **Интуиция и основная идея**
   - Яркий пример: «риэлтор оценивает квартиру по ценам ближайших проданных квартир».
   - Ключевой принцип: похожие объекты имеют похожие отклики.
   - Формальная постановка: обучающая выборка $\{(\mathbf{x}^{(i)}, y^{(i)})\}_{i=1}^n$, для нового объекта $\mathbf{x}_{\text{new}}$ ищем $k$ ближайших в смысле некоторой метрики $d(\mathbf{x}_{\text{new}}, \mathbf{x}^{(i)})$.
   - Ленивое обучение (lazy learning): нет фазы обучения, модель просто запоминает выборку.
   - Два режима: регрессия и классификация.

2. **kNN‑регрессия**
   - Предсказание как среднее (или взвешенное среднее) откликов $k$ соседей:
     $\hat{y}(\mathbf{x}_{\text{new}}) = \frac{1}{k}\sum_{i \in N_k(\mathbf{x}_{\text{new}})} y^{(i)}$.
   - Влияние $k$: $k=1$ — нулевая ошибка на обучении, но высокая дисперсия; $k=n$ — постоянное среднее, высокая систематическая ошибка. Компромисс смещения и дисперсии.
   - *[Вставить рисунок 1: сглаживание регрессионной кривой при увеличении k на искусственных данных]*.

3. **kNN‑классификация**
   - Правило большинства: $\hat{y} = \arg\max_c \sum_{i \in N_k} \mathbb{1}[y^{(i)} = c]$.
   - Для чётного $k$ — проблема ничьей, рекомендация нечётного $k$ для бинарного случая.
   - Оценка апостериорной вероятности: $\hat{P}(c | \mathbf{x}) = \frac{1}{k}\sum_{i \in N_k} \mathbb{1}[y^{(i)}=c]$.
   - Решающая граница: кусочно‑линейна для евклидовой метрики (интуитивно — диаграмма Вороного). *[Вставить рисунок 2: решающая граница 1‑NN и 5‑NN на двумерном примере]*.

4. **Метрики расстояний**
   - Евклидово расстояние: $d_2(\mathbf{a},\mathbf{b}) = \sqrt{\sum_{j}(a_j - b_j)^2}$.
   - Манхэттенское расстояние: $d_1 = \sum_j |a_j - b_j|$. Устойчивее к выбросам? Обсудить.
   - Расстояние Минковского $d_p$ как обобщение.
   - Косинусное расстояние (сходство): $1 - \frac{\mathbf{a}^T\mathbf{b}}{\|\mathbf{a}\|\|\mathbf{b}\|}$ — фокусируется на угле, а не на длине векторов. Важно для текстовых данных.
   - Расстояние Махаланобиса: $d_M = \sqrt{(\mathbf{a}-\mathbf{b})^T S^{-1}(\mathbf{a}-\mathbf{b})}$, учитывает корреляции признаков. *[Углублённый взгляд: связь с whitening‑преобразованием и оценка ковариационной матрицы S]*.
   - **Углублённый взгляд: проблема масштаба признаков.** Показать, что евклидово расстояние доминируется признаком с большей дисперсией. Вывод: стандартизация практически обязательна.

5. **Взвешенный kNN**
   - Идея: более близкие соседи важнее. Веса $w_i = 1/d(\mathbf{x}_{\text{new}}, \mathbf{x}^{(i)})^p$ или ядерные веса.
   - Взвешенное голосование в классификации, взвешенное среднее в регрессии.
   - Преимущество: меньшая чувствительность к выбору $k$.

6. **Выбор $k$**
   - $k$ — гиперпараметр. Маленькое $k$ — переобучение, сложная граница. Большое $k$ — недообучение, упрощение.
   - Выбор через кросс‑валидацию (без кода на этом этапе, только концептуально).
   - Эвристика: $k \approx \sqrt{n}$ как отправная точка (историческая рекомендация).

В конце первой части — **вопросы для самопроверки**:
- Для начинающих (6): что такое ленивое обучение, как влияет k на регрессию, правило большинства при классификации, в чём разница евклидова и манхэттенского расстояния, зачем стандартизировать признаки, что дают веса соседей.
- Для профессионалов (5): компромисс смещения и дисперсии для kNN, связь расстояния Махаланобиса с PCA, почему косинусное расстояние предпочтительнее евклидова для текстов, как обобщить kNN на многомерный отклик, как оценить апостериорные вероятности и насколько они калиброваны.

Никакого кода.
```

### Теоретические основы: от правила Байеса к LDA и QDA

```text
Напиши первую часть главы «Линейный и квадратичный дискриминантный анализ (LDA/QDA)» для учебного пособия, адресованного одновременно начинающим и профессионалам.

Стиль: основной текст интуитивный, с яркими аналогиями; для профессионалов — углублённые вставки «Углублённый взгляд» со строгими выводами и доказательствами.
Используй длинные тире (—) и кавычки-ёлочки («»). Обозначение дисперсии — $\mathrm{Var}$.

Начни с мостика: «Метод k‑ближайших соседей не делал предположений о форме распределения признаков — он просто опирался на сходство объектов. Теперь мы переходим к генеративному подходу: построим вероятностную модель того, как признаки распределены внутри каждого класса. Это даст не только классификатор, но и хорошо калиброванные вероятности, а также мощный инструмент визуализации — линейные дискриминанты».

Разделы строго по порядку:

1. **Байесовский классификатор**
   - Пусть $K$ классов, $\pi_k = P(Y=k)$ — априорная вероятность класса $k$, $p(\mathbf{x} \mid Y=k)$ — функция правдоподобия (плотность признаков для класса $k$).
   - Апостериорная вероятность по формуле Байеса:
     $$P(Y=k \mid \mathbf{x}) = \frac{\pi_k \, p(\mathbf{x} \mid Y=k)}{\sum_{j=1}^K \pi_j \, p(\mathbf{x} \mid Y=j)}.$$
   - Правило максимальной апостериорной вероятности (MAP): $\hat{y} = \arg\max_k P(Y=k \mid \mathbf{x})$.
   - Дискриминантные функции $\delta_k(\mathbf{x}) = \log P(Y=k \mid \mathbf{x})$ или любая монотонная функция. Тогда классификатор $\hat{y} = \arg\max_k \delta_k(\mathbf{x})$.
   - Граница между классами $i$ и $j$: $\{\mathbf{x} \mid \delta_i(\mathbf{x}) = \delta_j(\mathbf{x})\}$.
   - Почему работа с логарифмами удобнее: сумма вместо произведения, отбрасывание общих констант.

2. **Многомерное нормальное распределение (краткое напоминание)**
   - Плотность $p(\mathbf{x}) = \frac{1}{(2\pi)^{d/2}|\Sigma|^{1/2}} \exp\left(-\frac{1}{2}(\mathbf{x}-\boldsymbol{\mu})^T \Sigma^{-1} (\mathbf{x}-\boldsymbol{\mu})\right)$.
   - Параметры: $\boldsymbol{\mu} \in \mathbb{R}^d$, $\Sigma \in \mathbb{R}^{d \times d}$ (симметричная, положительно определённая).
   - Геометрический смысл: эллипсоиды постоянной плотности. Собственные векторы $\Sigma$ задают направления осей, собственные значения — растяжение.
   - Квадратичная форма $(\mathbf{x}-\boldsymbol{\mu})^T \Sigma^{-1} (\mathbf{x}-\boldsymbol{\mu})$ — расстояние Махаланобиса.

3. **Квадратичный дискриминантный анализ (QDA)**
   - Предположение: $p(\mathbf{x} \mid Y=k) = \mathcal{N}(\boldsymbol{\mu}_k, \Sigma_k)$. Каждый класс имеет свою ковариационную матрицу.
   - Логарифм апостериорной вероятности (с точностью до констант):
     $$\delta_k(\mathbf{x}) = -\frac{1}{2} \log|\Sigma_k| - \frac{1}{2}(\mathbf{x}-\boldsymbol{\mu}_k)^T \Sigma_k^{-1} (\mathbf{x}-\boldsymbol{\mu}_k) + \log \pi_k.$$
   - Это квадратичная функция от $\mathbf{x}$. Граница между классами $i$ и $j$:
     $$-\frac{1}{2}\mathbf{x}^T(\Sigma_i^{-1} - \Sigma_j^{-1})\mathbf{x} + \dots = 0 \quad \text{— уравнение квадрики.}$$
   - Какие фигуры могут получаться: эллипсы, гиперболы, параболы, пары прямых.
   - Число параметров QDA для $K$ классов и $d$ признаков: $K \cdot d$ средних + $K \cdot \frac{d(d+1)}{2}$ уникальных элементов ковариаций. Риск переобучения при малых $n$.

4. **Линейный дискриминантный анализ (LDA)**
   - Дополнительное предположение: $\Sigma_1 = \Sigma_2 = \dots = \Sigma_K = \Sigma$ — общая ковариационная матрица.
   - Тогда квадратичный член $\mathbf{x}^T \Sigma^{-1} \mathbf{x}$ в $\delta_k$ не зависит от $k$ и сокращается. Дискриминантная функция становится линейной:
     $$\delta_k(\mathbf{x}) = \mathbf{x}^T \Sigma^{-1} \boldsymbol{\mu}_k - \frac{1}{2}\boldsymbol{\mu}_k^T \Sigma^{-1} \boldsymbol{\mu}_k + \log \pi_k.$$
   - Разделяющая граница между классами $i$ и $j$ — гиперплоскость:
     $$\mathbf{x}^T \Sigma^{-1}(\boldsymbol{\mu}_i - \boldsymbol{\mu}_j) = \frac{1}{2}(\boldsymbol{\mu}_i^T \Sigma^{-1} \boldsymbol{\mu}_i - \boldsymbol{\mu}_j^T \Sigma^{-1} \boldsymbol{\mu}_j) - \log\frac{\pi_i}{\pi_j}.$$
   - Интерпретация: если классы равновероятны, граница — прямая, перпендикулярная $\Sigma^{-1}(\boldsymbol{\mu}_i - \boldsymbol{\mu}_j)$ и проходящая через середину отрезка между центрами (в смысле расстояния Махаланобиса).
   - Число параметров LDA: $K \cdot d + \frac{d(d+1)}{2}$ — значительно меньше, чем у QDA. Устойчивость, особенно при малых выборках.
   - **Углублённый взгляд: связь LDA с линейной регрессией.** При бинарной классификации ($y \in \{0,1\}$) и сбалансированных классах коэффициенты LDA пропорциональны коэффициентам линейной регрессии, предсказывающей индикатор класса. Показать это через теорему Байеса и линейную структуру дискриминантной функции.

5. **Оценка параметров по выборке**
   - Оценка априорных вероятностей: $\hat{\pi}_k = n_k / n$, где $n_k$ — число объектов класса $k$.
   - Оценка векторов средних: $\hat{\boldsymbol{\mu}}_k = \frac{1}{n_k} \sum_{i: y_i=k} \mathbf{x}^{(i)}$.
   - Оценка общей ковариационной матрицы для LDA (pooled covariance):
     $$\hat{\Sigma} = \frac{1}{n-K} \sum_{k=1}^K \sum_{i: y_i=k} (\mathbf{x}^{(i)} - \hat{\boldsymbol{\mu}}_k)(\mathbf{x}^{(i)} - \hat{\boldsymbol{\mu}}_k)^T.$$
     Пояснить, почему деление на $n-K$, а не $n$: несмещённость.
   - Для QDA: $\hat{\Sigma}_k = \frac{1}{n_k-1} \sum_{i: y_i=k} (\mathbf{x}^{(i)} - \hat{\boldsymbol{\mu}}_k)(\mathbf{x}^{(i)} - \hat{\boldsymbol{\mu}}_k)^T$.
   - Обсудить вырожденность $\hat{\Sigma}_k$, если $n_k \le d$ (особенно для QDA).

6. **Сравнение LDA и QDA: компромисс смещения и дисперсии**
   - LDA — более смещённая модель (жёсткое предположение об одинаковости ковариаций), но с меньшей дисперсией.
   - QDA — гибче, может точнее описать границу, но требует больше данных для оценки множества параметров.
   - Рекомендация: LDA как baseline; QDA, если данных много и есть основания полагать, что разброс классов различается.

В конце первой части — **вопросы для самопроверки**:
- Для начинающих (7): Что максимизирует байесовский классификатор? Запишите дискриминантную функцию LDA. Чем граница QDA отличается от границы LDA? Что такое расстояние Махаланобиса? Почему в LDA меньше параметров, чем в QDA? Как оценивается общая ковариационная матрица? Как интерпретировать коэффициенты LDA?
- Для профессионалов (5): Выведите явную формулу границы двух классов в LDA при равных априорных вероятностях. Докажите эквивалентность LDA и линейной регрессии для бинарной классификации. Как связана граница LDA с собственными векторами $\Sigma^{-1}$? Что происходит с QDA, когда $n_k \le d$? Сравните байесовскую ошибку LDA и QDA в случае, когда истинные ковариационные матрицы различны.

Никакого кода.
```
